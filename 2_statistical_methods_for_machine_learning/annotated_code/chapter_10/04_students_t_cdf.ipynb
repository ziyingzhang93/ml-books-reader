## Summary / 摘要



The t-distribution CDF is used to compute cumulative probabilities and determine critical values for t-tests. This notebook demonstrates how to compute and interpret the t-distribution CDF.



t 分布 CDF 用于计算累积概率和确定 t 检验的临界值。本笔记本演示如何计算和解释 t 分布 CDF。



**Key Insight / 关键见解:**



The t-distribution CDF is essential for hypothesis testing. At α=0.05 significance level (two-tailed), we find critical values where CDF=0.025 and CDF=0.975.

## Step 1 — Import Libraries / 导入库

```python
from numpy import arange

from matplotlib import pyplot

from scipy.stats import t, norm



# 创建样本空间

# Create sample space

sample_space = arange(-5, 5, 0.001)
```

## Step 2 — Compute t-distribution CDF / 计算 t 分布 CDF

```python
# 使用自由度 = 样本空间长度 - 1

# Use degrees of freedom = len(sample_space) - 1

dof = len(sample_space) - 1



# 计算 t 分布的 CDF

# Compute t-distribution CDF

cdf_t = t.cdf(sample_space, dof)



# 为了比较，也计算标准高斯分布的CDF

# For comparison, also compute standard Gaussian CDF

cdf_gaussian = norm.cdf(sample_space)



print(f"Degrees of freedom: {dof}")

print(f"CDF at x=0: {cdf_t[len(sample_space)//2]:.4f}")
```

## Step 3 — Plot t and Gaussian CDFs / 绘制 t 和高斯 CDF

```python
# 绘制 CDF

# Plot CDFs

pyplot.figure(figsize=(10, 6))

pyplot.plot(sample_space, cdf_t, linewidth=2, label=f"t-distribution (df={dof})")

pyplot.plot(sample_space, cdf_gaussian, linewidth=2, linestyle='--', label="Gaussian")

pyplot.xlabel('Value / 值')

pyplot.ylabel('Cumulative Probability / 累积概率')

pyplot.title("Student's t CDF vs Gaussian CDF / Student t CDF 与高斯 CDF")

pyplot.axhline(y=0.5, color='gray', linestyle=':', alpha=0.5)

pyplot.legend()

pyplot.grid(True, alpha=0.3)

pyplot.show()
```

## Step 4 — Critical Values for Hypothesis Testing / 假设检验的临界值

```python
# 对于 α = 0.05 的显著性水平（双尾）

# For α = 0.05 significance level (two-tailed)

alpha = 0.05

alpha_half = alpha / 2



# 找到临界值（两尾）

# Find critical values (two-tailed)

critical_value_lower = t.ppf(alpha_half, dof)

critical_value_upper = t.ppf(1 - alpha_half, dof)



print(f"Significance level (α): {alpha}")

print(f"Lower critical value (df={dof}): {critical_value_lower:.4f}")

print(f"Upper critical value (df={dof}): {critical_value_upper:.4f}")



# 比较高斯分布的临界值

# Compare with Gaussian critical values

gaussian_critical = norm.ppf(1 - alpha_half)

print(f"\nGaussian critical value: {gaussian_critical:.4f}")

print(f"\nDifference (t - Gaussian): {critical_value_upper - gaussian_critical:.4f}")

print("t 分布临界值更极端（further from 0）/ t-distribution has more extreme critical values")
```

## Step 5 — Visualize Rejection Regions / 可视化拒绝区域

```python
# 对于不同的自由度，绘制拒绝区域

# For different degrees of freedom, plot rejection regions

pyplot.figure(figsize=(12, 6))



for dof_val in [5, 10, 30, 100]:

    cdf = t.cdf(sample_space, dof_val)

    pyplot.plot(sample_space, cdf, linewidth=2, label=f"df={dof_val}")



# 添加高斯分布

pyplot.plot(sample_space, norm.cdf(sample_space), linewidth=2, linestyle='--', 

            color='black', label="Gaussian")



# 标记显著性水平

pyplot.axhline(y=0.025, color='red', linestyle='--', alpha=0.5, label='α/2 = 0.025')

pyplot.axhline(y=0.975, color='red', linestyle='--', alpha=0.5, label='1-α/2 = 0.975')



pyplot.xlabel('Value / 值')

pyplot.ylabel('Cumulative Probability / 累积概率')

pyplot.title("t-distribution CDF for Different Degrees of Freedom / 不同自由度的 t 分布 CDF")

pyplot.legend()

pyplot.grid(True, alpha=0.3)

pyplot.show()
```

## Learning Notes / 学习笔记



- **Statistical Concept / 统计学概念**: The t-distribution CDF is used to determine p-values and critical regions. With small degrees of freedom, the critical values are further from zero (more conservative), accounting for increased sampling variability. / t 分布 CDF 用于确定 p 值和临界区域。自由度小时，临界值离零更远（更保守），考虑了增加的抽样变异性。



- **ML Application / 机器学习应用**: t-tests compare group means with unknown population variance. The t-distribution is fundamental to constructing confidence intervals for regression coefficients, model parameters, and effect size estimates in machine learning. / t 检验在总体方差未知时比较组均值。t 分布对于构造机器学习中回归系数、模型参数和效应量估计的置信区间至关重要。

## Complete Code / 完整代码一览

```python
from numpy import arange

from matplotlib import pyplot

from scipy.stats import t, norm



sample_space = arange(-5, 5, 0.001)

dof = len(sample_space) - 1



# Compute CDFs

cdf_t = t.cdf(sample_space, dof)

cdf_gaussian = norm.cdf(sample_space)



# Plot CDFs

pyplot.figure(figsize=(10, 6))

pyplot.plot(sample_space, cdf_t, linewidth=2, label=f"t-distribution (df={dof})")

pyplot.plot(sample_space, cdf_gaussian, linewidth=2, linestyle='--', label="Gaussian")

pyplot.xlabel('Value')

pyplot.ylabel('Cumulative Probability')

pyplot.title("Student's t CDF vs Gaussian CDF")

pyplot.axhline(y=0.5, color='gray', linestyle=':', alpha=0.5)

pyplot.legend()

pyplot.grid(True, alpha=0.3)

pyplot.show()



# Find critical values

alpha = 0.05

alpha_half = alpha / 2

critical_value = t.ppf(1 - alpha_half, dof)

gaussian_critical = norm.ppf(1 - alpha_half)



print(f"t-distribution critical value (df={dof}): {critical_value:.4f}")

print(f"Gaussian critical value: {gaussian_critical:.4f}")
```
