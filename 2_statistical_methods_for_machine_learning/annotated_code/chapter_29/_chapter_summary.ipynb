# Chapter 29: Chi-Squared Test
# 第29章：卡方检验

## Theme | 主题
Testing independence: do two categorical variables depend on each other?
检验独立性：两个分类变量是否相互依赖？

## Evolution Roadmap | 演变路线图
```
Contingency Table (Observed counts)
└─ Compute Expected Counts (under independence H0)
   └─ Chi-Squared Statistic
      χ² = Σ (O - E)² / E
      └─ Compare to χ²(df) distribution
         └─ p-value & conclusion
```

## Progression Logic | 进度逻辑

### Stage 1: Contingency Table (列联表)
**English:** Two categorical variables arranged in rows and columns. Example: rows = {smoker, non-smoker}, columns = {lung cancer, no cancer}. Entries = observed counts.
**中文:** 两个分类变量排列成行和列。例如：行 = {吸烟者、非吸烟者}，列 = {肺癌、无癌}。条目 = 观察计数。

### Stage 2: Null Hypothesis (零假设)
**English:** H0: the two variables are independent (no association). Under H0, expected count for cell (i,j) = (row i total) * (col j total) / grand total.
**中文:** H0：两个变量是独立的(无关联)。在H0下，单元格(i,j)的预期计数 = (行i总计) * (列j总计) / 总计。

### Stage 3: Chi-Squared Statistic (卡方统计)
**English:** χ² = Σ (Observed - Expected)² / Expected, summed over all cells. Large χ² → strong departure from independence.
**中文:** χ² = Σ (观察 - 预期)² / 预期，在所有单元格上求和。较大的χ² → 强烈偏离独立性。

### Stage 4: Degrees of Freedom (自由度)
**English:** df = (# rows - 1) * (# columns - 1). E.g., 2x2 table has df = 1.
**中文:** df = (# 行 - 1) * (# 列 - 1)。例如，2x2表有df = 1。

### Stage 5: p-Value & Interpretation (p值与解释)
**English:** Compare χ² to χ²(df) distribution. Small p-value (<0.05) rejects H0: variables are associated. Large p-value fails to reject H0: evidence of independence.
**中文:** 将χ²与χ²(df)分布进行比较。小p值(<0.05)拒绝H0：变量相关联。大p值未能拒绝H0：独立的证据。

## ML Relevance | ML相关性

1. **Feature Independence (特征独立性)**: Test if features and target are associated, guiding feature selection.
2. **Categorical Data (分类数据)**: Applies to any categorical variables, not just numeric data.
3. **Data Exploration (数据探索)**: Identify associations in contingency tables (e.g., customer segment vs. purchase category).
4. **Model Assumptions (模型假设)**: Some models assume feature independence; χ² test validates this assumption.
