## Summary / 摘要

Box-Cox transformation y = (X^λ - 1)/λ optimally transforms skewed data toward normality. The parameter λ is chosen by maximum likelihood to minimize deviation from normality. scipy.stats.boxcox automatically selects optimal λ and returns transformed data. Works for positive-valued data (X > 0). After transformation, data are closer to normal, enabling parametric methods. Shows practical workflow: visual/statistical assessment of non-normality → apply transformation → re-assess → apply standard methods.
