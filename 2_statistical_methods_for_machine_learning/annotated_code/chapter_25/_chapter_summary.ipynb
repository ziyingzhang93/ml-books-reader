# Chapter 25: Non-Gaussian Data
# 第25章：非高斯数据

## Theme | 主题
Real data is messy: understanding why normality fails and how to fix it via transformations.
真实数据是混乱的：理解为什么正态性失败以及如何通过变换来修复。

## Evolution Roadmap | 演变路线图
```
WHY DATA IS NON-GAUSSIAN:
  Small Sample → high variability
  └─ Large Sample → population non-normality
     └─ Low Resolution (discretization artifacts)
        └─ Outliers (contamination or real extreme events)
           └─ Long Tail (exponential, power-law decay)
              └─ Truncation (artificial boundaries)

HOW TO FIX:
  Exponential Distribution → log transformation
  └─ Power-Law Distribution → Box-Cox transformation
     └─ Right-Skewed → Yeo-Johnson transformation (handles negatives)
```

## Progression Logic | 进度逻辑

### Stage 1: Small Sample (小样本)
**English:** Limited data exhibits high sampling variability, masking true distribution. Histogram may look non-normal simply due to sparse coverage.
**中文:** 有限数据表现出高采样变异性，掩盖真实分布。直方图可能看起来非正态只是因为稀疏覆盖。

### Stage 2: Large Sample (大样本)
**English:** With more data, the true population distribution emerges. If it deviates from normal (e.g., exponential, Poisson), transformation becomes necessary.
**中文:** 更多数据后，真实总体分布出现。如果它偏离正态(例如指数、泊松)，变换变得必要。

### Stage 3: Low Resolution (低分辨率)
**English:** Rounding or discretization artifacts (e.g., survey ratings 1-5) create discrete bumps. Transformation smooths these.
**中文:** 四舍五入或离散化工件(例如调查评级1-5)创建离散凹凸。变换平滑这些。

### Stage 4: Outliers (异常值)
**English:** Extreme values, whether measurement error or real, inflate variance and skewness. Log or robust transformations suppress extreme values.
**中文:** 极端值，无论是测量误差还是真实，都会膨胀方差和偏斜。日志或鲁棒变换抑制极端值。

### Stage 5: Long Tail (长尾)
**English:** Exponential or power-law data has right tail extending far beyond the mean. Log (exponential) or Box-Cox (power-law) transformations compress the tail.
**中文:** 指数或幂律数据的右尾远超均值。Log(指数)或Box-Cox(幂律)变换压缩尾部。

### Stage 6: Truncation (截断)
**English:** Artificial boundaries (e.g., non-negative prices, limited survey range) create artificial non-normality. Domain knowledge may help (e.g., assume latent normal + truncation).
**中文:** 人工边界(例如非负价格、有限调查范围)创建人工非正态性。领域知识可能有帮助(例如假设潜在正态+截断)。

### Stage 7: Exponential Distribution (指数分布)
**English:** Strongly right-skewed (e.g., wait times, failure times). Log transformation: y' = log(y) yields approximately normal y'.
**中文:** 强右偏斜(例如等待时间、失败时间)。Log变换：y' = log(y)产生大约正态的y'。

### Stage 8: Box-Cox Transformation (Box-Cox变换)
**English:** Optimal power transformation: y' = (y^λ - 1) / λ for λ estimated from data. Finds best λ to normalize y. Yeo-Johnson variant handles negative values.
**中文:** 最优幂变换：y' = (y^λ - 1) / λ对于从数据估计的λ。找到最佳λ来规范化y。Yeo-Johnson变体处理负值。

## ML Relevance | ML相关性

1. **Statistical Inference (统计推断)**: Transformation enables use of parametric tests and methods that assume normality.
2. **Linear Regression (线性回归)**: Residuals should be normal. Transform target variable if residuals violate normality.
3. **Variance Stabilization (方差稳定)**: Heteroscedastic errors (variance changes with x) often become homoscedastic after transformation.
4. **Interpretability (可解释性)**: Log transformation has natural interpretation (percentage change) and appears frequently in exponential growth models.
