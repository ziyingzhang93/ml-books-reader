# Chapter 8: Central Limit Theorem
# 第8章：中心极限定理

## Theme | 主题
One sample starts as jagged; repeated sampling reveals the hidden Gaussian sampling distribution.
一个样本开始是锯齿状的；重复采样揭示隐藏的高斯采样分布。

## Evolution Roadmap | 演变路线图
```
Single Simulation (One sample mean)
└─ 1000 Simulations Histogram
   (Sample means form a Gaussian, regardless of parent distribution)
```

## Progression Logic | 进度逻辑

### Stage 1: One Sample (一个样本)
**English:** Draw a single random sample from any distribution (normal, uniform, exponential, etc.), compute its mean.
**中文:** 从任何分布(正态、均匀、指数等)中抽取单个随机样本，计算其均值。

### Stage 2: Repeated Sampling (重复采样)
**English:** Repeat the process 1000 times (or more), collecting sample means into a histogram.
**中文:** 重复该过程1000次(或更多)，将样本均值收集到直方图中。

### Stage 3: Distribution Emergence (分布出现)
**English:** The histogram of sample means forms a bell curve (Gaussian), even if the parent distribution was non-normal. This is the Central Limit Theorem.
**中文:** 样本均值的直方图形成钟形曲线(高斯)，即使母分布是非正态的。这就是中心极限定理。

## ML Relevance | ML相关性

1. **Hypothesis Testing (假设检验)**: CLT justifies using z-tests and t-tests even on non-normal data (with large samples).
2. **Confidence Intervals (置信区间)**: CLT enables construction of CIs around sample means without knowing the true distribution.
3. **Sampling Distribution (采样分布)**: CLT shows that SE = σ / sqrt(n), linking sample size to estimate precision.
4. **Model Robustness (模型鲁棒性)**: Large training sets often have sample means that are approximately Gaussian-distributed, making inference more reliable.
