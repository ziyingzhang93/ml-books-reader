## Summary / 摘要

**English:** This notebook simulates rolling a fair six-sided die 50 times and calculates the sample mean. A fair die has expected value 3.5. This single simulation is the foundation for the Central Limit Theorem: we'll repeat this many times in the next notebook to see how sample means distribute.

**中文:** 本笔记本模拟投掷一个公平的六面骰子50次并计算样本均值。公平骰子的期望值为3.5。这个单次模拟是中心极限定理的基础：我们在下一个笔记本中会重复多次来查看样本均值如何分布。

**Formula / 公式:**
$$E[X] = \sum_{x=1}^{6} x \cdot P(X = x) = \sum_{x=1}^{6} x \cdot \frac{1}{6} = \frac{21}{6} = 3.5$$

## Step 1 — Import Libraries / 导入库

```python
# Import random integer generation and mean calculation / 导入随机整数生成和均值计算
from numpy.random import seed
from numpy.random import randint
from numpy import mean
```

## Step 2 — Set Seed / 设置种子

```python
# Set seed for reproducibility / 设置种子以保证可重现性
seed(1)
```

## Step 3 — Simulate Dice Rolls / 模拟骰子投掷

```python
# Generate 50 random die rolls (values 1-6) / 生成50次随机骰子投掷（值1-6）
# randint(1, 7, 50) generates 50 integers in [1, 7), i.e., 1-6 / randint(1, 7, 50)生成50个[1, 7)中的整数，即1-6
rolls = randint(1, 7, 50)
print("50 dice rolls:")
print(rolls)
```

## Step 4 — Calculate Mean / 计算均值

```python
# Calculate the sample mean of the rolls / 计算投掷的样本均值
# For a fair die, the expected value is 3.5 / 对于公平骰子，期望值为3.5
sample_mean = mean(rolls)

print(f"\nSample mean: {sample_mean:.2f}")
print(f"Expected value of one die: 3.5")
print(f"Difference from expected: {abs(sample_mean - 3.5):.2f}")
```

## Learning Notes / 学习笔记

- **Statistical Concept / 统计学概念:** A fair six-sided die has probability 1/6 for each outcome {1, 2, 3, 4, 5, 6}. The expected value is $E[X] = \sum x \cdot P(X=x) = 3.5$. The variance is $\text{Var}(X) = \sum (x - 3.5)^2 \cdot (1/6) = 35/12 \approx 2.917$. The sample mean of 50 rolls will vary around 3.5 due to random variation. As we increase the number of simulations (next notebook), the distribution of sample means will approach a Gaussian distribution centered at 3.5—this is the Central Limit Theorem in action.

- **ML Application / 机前machine learning:** Simulating random processes (like dice rolls) is fundamental to Monte Carlo methods used in RL, Bayesian inference, and generative models. Understanding how sample statistics vary across independent trials motivates ensemble methods and bootstrap estimation. Neural network training uses stochastic gradient descent, which involves sampling mini-batches—each mini-batch's loss is analogous to a single die roll's outcome. Confidence intervals and uncertainty estimates rely on understanding the distribution of sample statistics across different random realizations.

## Complete Code / 完整代码一览

```python
# ===== Section 1: Imports =====
from numpy.random import seed
from numpy.random import randint
from numpy import mean

# ===== Section 2: Set Seed =====
# Set seed for reproducibility / 设置种子以保证可重现性
seed(1)

# ===== Section 3: Simulate Dice Rolls =====
# Roll a fair die 50 times / 投掷一个公平骰子50次
rolls = randint(1, 7, 50)
print("50 dice rolls:")
print(rolls)

# ===== Section 4: Calculate and Interpret Mean =====
# Compute sample mean / 计算样本均值
sample_mean = mean(rolls)

print(f"\nSample mean: {sample_mean:.2f}")
print(f"Expected value of one die: 3.5")
print(f"Difference from expected: {abs(sample_mean - 3.5):.2f}")
```
