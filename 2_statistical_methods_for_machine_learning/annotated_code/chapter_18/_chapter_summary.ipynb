# Chapter 18: Cross-Validation
# 第18章：交叉验证

## Theme | 主题
Systematic resampling for model evaluation: estimate generalization error without overfitting.
系统重采样用于模型评估：估计泛化误差而不过度拟合。

## Evolution Roadmap | 演变路线图
```
Dataset (all observed data)
└─ k-Fold Partitioning
   └─ Iterate: Train on (k-1) folds, evaluate on 1 fold
      └─ Collect k performance scores
         └─ Average score + CI estimates generalization error
```

## Progression Logic | 进度逻辑

### Stage 1: Dataset Partitioning (数据集分割)
**English:** Divide data into k roughly equal folds (typical k=5 or 10). Each fold is held out once as a test set.
**中文:** 将数据分为k个大约相等的折叠(典型k=5或10)。每个折叠作为测试集保留一次。

### Stage 2: Iteration (迭代)
**English:** For fold i: train model on folds 1..k excluding i, evaluate on fold i. Record performance metric (accuracy, RMSE, AUC, etc.).
**中文:** 对于折叠i：在除i外的折叠1..k上训练模型，在折叠i上评估。记录性能指标(准确性、RMSE、AUC等)。

### Stage 3: Performance Collection (性能收集)
**English:** After k iterations, collect k performance scores. Compute mean (point estimate of generalization) and std (uncertainty).
**中文:** 在k次迭代后，收集k个性能分数。计算平均值(泛化的点估计)和标准差(不确定性)。

### Stage 4: Stratification (分层)
**English:** For classification, stratify folds by class label to ensure each fold is representative. For regression, stratify by target quantiles.
**中文:** 对于分类，按类别标签对折叠进行分层以确保每个折叠具有代表性。对于回归，按目标分位数进行分层。

## ML Relevance | ML相关性

1. **Model Selection (模型选择)**: k-fold CV estimates generalization error for different models, guiding selection.
2. **Hyperparameter Tuning (超参数调整)**: Nested CV: outer loop for evaluation, inner loop for hyperparameter selection.
3. **Small Sample Protection (小样本保护)**: CV maximizes data use compared to single train/test split.
4. **Variance Reduction (方差降低)**: Average of k scores is more stable than single-split estimate.
