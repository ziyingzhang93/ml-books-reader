## Summary / 摘要



Before analyzing correlation, we need test data with known relationships. This notebook demonstrates how to generate two correlated variables and visualize their relationship using scatter plots.



在分析相关性之前，我们需要具有已知关系的测试数据。本笔记本演示如何生成两个相关变量并使用散点图可视化其关系。



**Correlation Concept / 相关性概念:**


Correlation measures the linear relationship between two variables, ranging from -1 (perfect negative correlation) to +1 (perfect positive correlation).

## Step 1 — Import Libraries and Set Random Seed / 导入库并设置随机种子

```python
from numpy import random, mean, std

from matplotlib import pyplot



# 设置随机种子以实现可重复性

# Set random seed for reproducibility

random.seed(1)
```

## Step 2 — Generate Correlated Variables / 生成相关变量

```python
# 生成 1000 个相关变量的观察值

# Generate 1000 observations of correlated variables



# 第一个变量：均值=100，标准差=20

# First variable: mean=100, std=20
data1 = 20 * random.randn(1000) + 100



# 第二个变量：与 data1 成线性关系，加上噪声

# Second variable: linear relationship with data1 plus noise
data2 = data1 + (10 * random.randn(1000) + 50)



# 计算描述性统计量

# Compute descriptive statistics

print(f"Data 1 - mean={mean(data1):.3f}, stdv={std(data1):.3f}")
print(f"Data 2 - mean={mean(data2):.3f}, stdv={std(data2):.3f}")
```

## Step 3 — Visualize the Dataset / 可视化数据集

```python
# 创建散点图

# Create scatter plot

pyplot.figure(figsize=(10, 6))

pyplot.scatter(data1, data2, alpha=0.5, s=20)

pyplot.xlabel('Variable 1 (data1) / 变量 1')

pyplot.ylabel('Variable 2 (data2) / 变量 2')

pyplot.title('Scatter Plot of Correlated Variables / 相关变量的散点图')

pyplot.grid(True, alpha=0.3)

pyplot.show()



# 观察：两个变量呈正线性关系

# Observation: both variables show positive linear relationship
```

## Step 4 — Summary Statistics / 汇总统计

```python
import numpy as np



# 计算更详细的统计量

# Compute detailed statistics

print("Data 1 Statistics / 数据 1 统计:")
print(f"  Mean / 均值: {mean(data1):.4f}")
print(f"  Std Dev / 标准差: {std(data1):.4f}")
print(f"  Min / 最小值: {data1.min():.4f}")
print(f"  Max / 最大值: {data1.max():.4f}")
print(f"  Median / 中位数: {np.median(data1):.4f}")


print("\nData 2 Statistics / 数据 2 统计:")
print(f"  Mean / 均值: {mean(data2):.4f}")
print(f"  Std Dev / 标准差: {std(data2):.4f}")
print(f"  Min / 最小值: {data2.min():.4f}")
print(f"  Max / 最大值: {data2.max():.4f}")
print(f"  Median / 中位数: {np.median(data2):.4f}")


print(f"\nNumber of observations / 观察数: {len(data1)}")
```

## Step 5 — Distribution Visualization / 分布可视化

```python
# 绘制直方图和分布

# Plot histograms and distributions

pyplot.figure(figsize=(12, 5))



pyplot.subplot(1, 2, 1)

pyplot.hist(data1, bins=30, alpha=0.7, edgecolor='black')

pyplot.xlabel('Value / 值')

pyplot.ylabel('Frequency / 频率')

pyplot.title('Distribution of Variable 1 / 变量 1 的分布')

pyplot.grid(True, alpha=0.3, axis='y')



pyplot.subplot(1, 2, 2)

pyplot.hist(data2, bins=30, alpha=0.7, edgecolor='black', color='orange')

pyplot.xlabel('Value / 值')

pyplot.ylabel('Frequency / 频率')

pyplot.title('Distribution of Variable 2 / 变量 2 的分布')

pyplot.grid(True, alpha=0.3, axis='y')



pyplot.tight_layout()

pyplot.show()
```

## Complete Code / 完整代码一览

```python
from numpy import random, mean, std

from matplotlib import pyplot

import numpy as np



# Set random seed

random.seed(1)



# Generate correlated variables

data1 = 20 * random.randn(1000) + 100

data2 = data1 + (10 * random.randn(1000) + 50)



# Print statistics

print(f"Data 1 - mean={mean(data1):.3f}, std={std(data1):.3f}")
print(f"Data 2 - mean={mean(data2):.3f}, std={std(data2):.3f}")


# Create scatter plot

pyplot.figure(figsize=(10, 6))

pyplot.scatter(data1, data2, alpha=0.5, s=20)

pyplot.xlabel('Variable 1')

pyplot.ylabel('Variable 2')

pyplot.title('Scatter Plot of Correlated Variables')

pyplot.grid(True, alpha=0.3)

pyplot.show()



# Distribution histograms

pyplot.figure(figsize=(12, 5))

pyplot.subplot(1, 2, 1)

pyplot.hist(data1, bins=30, alpha=0.7)

pyplot.title('Distribution of Variable 1')

pyplot.subplot(1, 2, 2)

pyplot.hist(data2, bins=30, alpha=0.7, color='orange')

pyplot.title('Distribution of Variable 2')

pyplot.tight_layout()

pyplot.show()
```
