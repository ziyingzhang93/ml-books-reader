## Summary / 摘要

**English:** This notebook creates a line plot to visualize the sine function over a continuous interval. Line plots are ideal for showing relationships between variables, trends over time, and continuous functions.

**中文:** 本笔记本创建一个线图以在连续区间上可视化正弦函数。线图适合显示变量之间的关系、时间趋势和连续函数。

## Step 1 — Import Libraries / 导入库

```python
# Import mathematical function and plotting library / 导入数学函数和绘图库
from numpy import sin
from matplotlib import pyplot
```

## Step 2 — Create X-axis Values / 创建X轴值

```python
# Generate x-values from 0 to 10 with step 0.1 / 生成0到10的x值，步长为0.1
# This creates 100 evenly-spaced points for smooth visualization / 这创建100个均匀分布的点以便平滑可视化
x = [x*0.1 for x in range(100)]
```

## Step 3 — Calculate Y-axis Values / 计算Y轴值

```python
# Compute sine of each x value / 计算每个x值的正弦
# sin() expects values in radians; our range [0, 10) covers ~1.6 periods / sin()期望弧度值；我们的范围[0, 10)覆盖~1.6个周期
y = sin(x)
```

## Step 4 — Create and Display Line Plot / 创建并显示线图

```python
# Create a line plot connecting points (x, y) / 创建连接点(x, y)的线图
pyplot.plot(x, y)

# Add labels and title / 添加标签和标题
pyplot.xlabel('x (radians) / x（弧度）')
pyplot.ylabel('sin(x)')
pyplot.title('Sine Function / 正弦函数')
pyplot.grid(True)

# Display the plot / 显示图表
pyplot.show()
```

## Learning Notes / 学习笔记

- **Statistical Concept / 统计学概念:** Line plots represent continuous data and functional relationships. They are particularly useful for time series data, where the x-axis represents time and the y-axis represents a measured variable. The connecting lines suggest continuity and trend. Line plots are fundamental for exploratory data analysis (EDA) when investigating how variables evolve.

- **ML Application / 机器学习应用:** Line plots are essential for monitoring training progress in machine learning. Plotting loss or accuracy over training epochs reveals convergence behavior, overfitting, and underfitting. Time series forecasting models are evaluated using line plots to visually compare predicted vs. actual values. Feature visualization in neural networks often uses line plots to display learned patterns over input ranges.

## Complete Code / 完整代码一览

```python
# ===== Section 1: Imports =====
from numpy import sin
from matplotlib import pyplot

# ===== Section 2: Prepare Data =====
# Create x-values from 0 to 10 with fine resolution / 生成0到10的x值，细分辨率
x = [x*0.1 for x in range(100)]

# Calculate corresponding y-values using sine function / 使用正弦函数计算对应的y值
y = sin(x)

# ===== Section 3: Create Line Plot =====
# Plot line connecting all (x, y) points / 绘制连接所有(x, y)点的线
pyplot.plot(x, y)

# Add labels and grid for clarity / 添加标签和网格以便清晰
pyplot.xlabel('x (radians) / x（弧度）')
pyplot.ylabel('sin(x)')
pyplot.title('Sine Function / 正弦函数')
pyplot.grid(True)

# Display plot / 显示图表
pyplot.show()
```
