# Chapter 27: Nonparametric Correlation
# 第27章：非参数相关

## Theme | 主题
Beyond Pearson: rank-based correlation for monotonic relationships that don't require linearity.
超越皮尔逊：基于等级的相关性，用于不需要线性的单调关系。

## Evolution Roadmap | 演变路线图
```
Dataset (two variables x, y)
└─ Transform to Ranks
   └─ Spearman ρ (Spearman's rank correlation)
      └─ Kendall τ (Kendall's tau)
         (even more robust to outliers)
```

## Progression Logic | 进度逻辑

### Stage 1: Motivation for Rank Correlation (排名相关的动机)
**English:** Pearson r measures linear association. If relationship is monotonic but nonlinear (e.g., exponential, logarithmic), Pearson r underestimates. Rank correlation detects monotonic trend.
**中文:** Pearson r测量线性关联。如果关系是单调但非线性的(例如指数、对数)，Pearson r低估。排名相关检测单调趋势。

### Stage 2: Spearman ρ (Spearman's Rho) (Spearman ρ)
**English:** Rank both x and y, then compute Pearson r on ranks: ρ = corr(rank(x), rank(y)). Ranges [-1, 1]. Robust to outliers: extreme values still get extreme ranks, but magnitude doesn't matter.
**中文:** 对x和y进行排名，然后计算等级的Pearson r：ρ = corr(rank(x), rank(y))。范围[-1、1]。对异常值鲁棒：极端值仍获得极端等级，但幅度无关紧要。

### Stage 3: Kendall τ (Kendall's Tau) (Kendall τ)
**English:** Count concordant (agreeing) and discordant (disagreeing) pairs. τ = (# concordant - # discordant) / total pairs. Even more robust than Spearman, smaller standard error for large n.
**中文:** 计算协调(同意)和不协调(不同意)对。τ = (# 协调 - # 不协调) / 总对。比Spearman更加鲁棒，对于大n具有较小的标准误差。

### Stage 4: When to Use Each (何时使用每个)
**English:** Pearson r: linear relationships, no outliers. Spearman ρ: monotonic relationships, moderate outliers. Kendall τ: severe outliers, smaller sample size.
**中文:** Pearson r：线性关系，无异常值。Spearman ρ：单调关系，中等异常值。Kendall τ：严重异常值、较小样本量。

## ML Relevance | ML相关性

1. **Feature Selection (特征选择)**: Spearman/Kendall correlation with target identifies monotonic predictive features missed by Pearson.
2. **Outlier Robustness (异常值鲁棒性)**: Nonparametric correlation is more stable when data contains extreme values.
3. **Nonlinear Relationships (非线性关系)**: Captures monotonic trends that are nonlinear (e.g., log, exponential).
4. **Distribution-Free (无分布)**: No distributional assumptions; works on ordinal data too.
