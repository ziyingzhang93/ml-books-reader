## Summary / 摘要

Kendall's tau ($\tau$) is another rank-based correlation coefficient that measures the ordinal association between two variables. It counts the number of concordant and discordant pairs, making it particularly useful for small sample sizes and ordinal data. Kendall's tau often provides different insights than Spearman's rho, especially in the presence of tied ranks.

Kendall的tau($\tau$)是另一个基于排序的相关系数，测度两个变量之间的序数关联。它计算一致对和不一致对的数量，对于小样本量和序数数据特别有用。Kendall的tau经常比Spearman的rho提供不同的见解，尤其是在存在并列排序时。

### Key Formula / 关键公式

$$\tau = \frac{n_c - n_d}{\frac{1}{2}n(n-1)}$$

where $n_c$ is concordant pairs, $n_d$ is discordant pairs, and $n$ is sample size.

## Step 1 — Generate Data / 生成数据

```python
from numpy.random import rand, seed

# Set random seed for reproducibility / 设置随机种子以保证可重现性
seed(1)

# Generate two variables with monotonic relationship / 生成具有单调关系的两个变量
data1 = rand(1000) * 20
data2 = data1 + (rand(1000) * 10)
```

## Step 2 — Calculate Kendall's Correlation / 计算Kendall相关

```python
from scipy.stats import kendalltau

# Calculate Kendall's tau correlation coefficient and p-value
# kendalltau returns: (tau coefficient, p-value)
# 计算Kendall的tau相关系数和p值
# kendalltau返回: (tau系数, p值)
coef, p = kendalltau(data1, data2)

print('Kendall correlation coefficient: %.3f' % coef)
```

## Step 3 — Interpret Statistical Significance / 解释统计显著性

```python
# Interpret the p-value using significance level alpha = 0.05
# H0: No ordinal association (tau = 0)
# H1: Ordinal association exists (tau ≠ 0)
# 使用显著性水平alpha = 0.05解释p值
# H0: 无序数关联(tau = 0)
# H1: 存在序数关联(tau ≠ 0)
alpha = 0.05

if p > alpha:
    print('Samples are uncorrelated (fail to reject H0) p=%.3f' % p)
else:
    print('Samples are correlated (reject H0) p=%.3f' % p)
```

## Learning Notes / 学习笔记

- **Statistical Concept**: Kendall's tau is based on the probability that two random observations have the same ordering. It typically has a lower absolute value than Spearman's rho for the same data, but it has better statistical properties and a clearer interpretation as a probability. The coefficient ranges from -1 to +1 with same interpretation as Spearman's.

  **统计概念**: Kendall的tau基于两个随机观察具有相同排序的概率。对于相同数据，它通常的绝对值比Spearman的rho低，但它具有更好的统计属性和更清晰的概率解释。系数范围从-1到+1，解释与Spearman的相同。

- **ML Application**: In ranking and preference learning tasks, Kendall's tau is more appropriate than Spearman's rho. It's useful in collaborative filtering, information retrieval ranking evaluation, and understanding ordinal relationships in ordinal regression models. The pair-counting interpretation makes it suitable for ranking disagreement analysis.

  **ML应用**: 在排名和偏好学习任务中，Kendall的tau比Spearman的rho更适合。它在协同过滤、信息检索排名评估和理解序数回归模型中的序数关系时很有用。对对计数的解释使其适合排名分歧分析。

## Complete Code / 完整代码一览

```python
from numpy.random import rand, seed
from scipy.stats import kendalltau

seed(1)
data1 = rand(1000) * 20
data2 = data1 + (rand(1000) * 10)

coef, p = kendalltau(data1, data2)
print('Kendall correlation coefficient: %.3f' % coef)

alpha = 0.05
if p > alpha:
    print('Samples are uncorrelated (fail to reject H0) p=%.3f' % p)
else:
    print('Samples are correlated (reject H0) p=%.3f' % p)
```
