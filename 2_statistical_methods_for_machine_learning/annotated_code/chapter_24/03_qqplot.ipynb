## Summary / 摘要

A Q-Q (Quantile-Quantile) plot compares sample quantiles against theoretical normal quantiles. Points should fall on a diagonal line for normally distributed data. Deviations from the line indicate departures from normality: curved patterns suggest skewness, divergence at the tails suggests heavy or light tails. The statsmodels.graphics.gofplots.qqplot function provides a convenient implementation. Q-Q plots are more objective than histograms and sensitive to tail behavior, making them ideal for assessing distributional assumptions.

Q-Q（分位数-分位数）图将样本分位数与理论正态分位数进行比较。对于正态分布数据，点应落在对角线上。偏离线表示偏离正态性：曲线模式表示偏度，尾部散度表示重尾或轻尾。statsmodels.graphics.gofplots.qqplot函数提供了方便的实现。Q-Q图比直方图更客观，对尾部行为敏感，使其理想用于评估分布假设。

## Step 1 — Import Libraries and Generate Data / 导入库和生成数据

```python3
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from statsmodels.graphics.gofplots import qqplot

np.random.seed(42)
data = np.random.normal(50, 15, 100)

print(f"Data generated: n={len(data)}")
```

## Step 2 — Create Q-Q Plot / 创建Q-Q图

```python3
# Create Q-Q plot with confidence band
# 创建带有置信带的Q-Q图
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Q-Q plot with fit line
# 图1: Q-Q图与拟合线
qqplot(data, line='45', ax=axes[0])
axes[0].set_title('Q-Q Plot: Sample vs. Normal Distribution')
axes[0].set_xlabel('Theoretical Quantiles')
axes[0].set_ylabel('Sample Quantiles')
axes[0].grid(True, alpha=0.3)

# Plot 2: Q-Q plot with best fit
# 图2: Q-Q图与最佳拟合
qqplot(data, line='s', ax=axes[1])
axes[1].set_title('Q-Q Plot: With Fitted Line')
axes[1].set_xlabel('Theoretical Quantiles')
axes[1].set_ylabel('Sample Quantiles')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print(f"Q-Q Plot interpretation:")
print(f"  Points near diagonal line → Normal distribution")
print(f"  Curved pattern → Skewness")
print(f"  S-shape → Departure from normality")
```

```python3
## Learning Notes / 学习笔记

- **Statistical Concept**: Q-Q plots are based on comparing empirical quantiles (sample percentiles) to theoretical quantiles from the normal distribution. For perfectly normal data, all points lie exactly on the reference line. The slope of the fitted line represents the standard deviation; the intercept represents the mean. Systematic deviations reveal specific non-normality patterns: left curves indicate left skew, S-shapes indicate heavy tails.
  
  **统计概念**: Q-Q图基于比较经验分位数（样本百分位数）和正态分布的理论分位数。对于完全正态数据，所有点精确位于参考线上。拟合线的斜率表示标准差；截距表示均值。系统性偏差揭示了特定的非正态性模式：左曲线表示左偏，S形表示重尾。

- **ML Application**: Q-Q plots are essential for model diagnostic. In linear regression, Q-Q plots of residuals validate normality assumptions; departures suggest need for transformation or robust methods. For classification, Q-Q plots of predicted probabilities help assess calibration. Combined with histograms and statistical tests, Q-Q plots provide comprehensive normality assessment. Practitioners often prefer Q-Q plots over formal tests due to visual interpretability and sensitivity to specific deviation types.
```

## Complete Code / 完整代码一览

```python3
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.gofplots import qqplot

np.random.seed(42)
data = np.random.normal(50, 15, 100)

fig, ax = plt.subplots(figsize=(8, 6))
qqplot(data, line='45', ax=ax)
ax.set_title('Q-Q Plot: Sample vs. Normal')
ax.grid(True, alpha=0.3)
plt.show()
```
