## Summary / 摘要

**English:** This notebook demonstrates the Law of Large Numbers: as sample size increases, the sample mean converges to the true population mean. We generate samples of increasing size from N(50,5) and plot the error (sample mean - true mean) against sample size to visualize convergence.

**中文:** 本笔记本演示了大数法则：随着样本量增加，样本均值收敛到真实总体均值。我们从N(50,5)生成不同大小的样本，并绘制误差（样本均值 - 真实均值）对样本量的散点图以可视化收敛。

**Formula / 公式:**
$$\lim_{n \to \infty} \bar{X}_n = \mu \text{ (almost surely)}$$

## Step 1 — Import Libraries / 导入库

```python
# Import random generation, numerical operations, and plotting / 导入随机生成、数值操作和绘图
from numpy.random import seed
from numpy.random import randn
from numpy import mean
from numpy import array
from matplotlib import pyplot
```

## Step 2 — Set Seed and Define Sample Sizes / 设置种子并定义样本大小

```python
# Set seed for reproducibility / 设置种子以保证可重现性
seed(1)

# Create list of increasing sample sizes / 创建递增的样本大小列表
# From 10 to 20000 with step 200 / 从10到20000，步长200
sizes = list()
for x in range(10, 20000, 200):
    sizes.append(x)

print(f"Sample sizes: {sizes[:5]} ... {sizes[-5:]}")
print(f"Total number of experiments: {len(sizes)}")
```

## Step 3 — Generate Samples and Calculate Means / 生成样本并计算均值

```python
# For each sample size, generate samples and compute the mean / 对于每个样本大小，生成样本并计算均值
# samples ~ N(50, 5), so true mean is 50 / 样本 ~ N(50, 5)，所以真实均值为50
# The error is (sample_mean - 50) / 误差是(sample_mean - 50)
means = [mean(5 * randn(size) + 50) for size in sizes]

print(f"First 5 means (should vary): {[f'{m:.2f}' for m in means[:5]]}")
print(f"Last 5 means (should be closer to 50): {[f'{m:.2f}' for m in means[-5:]]}")
```

## Step 4 — Plot Error vs Sample Size / 绘制误差对样本大小的散点图

```python
# Plot the error (sample mean - true mean) vs sample size / 绘制误差（样本均值 - 真实均值）对样本大小的散点图
# The Law of Large Numbers: as n increases, error should shrink to 0 / 大数法则：随着n增加，误差应该缩小到0
pyplot.scatter(sizes, array(means) - 50, alpha=0.5)
pyplot.xlabel('Sample Size / 样本大小')
pyplot.ylabel('Error (Sample Mean - True Mean) / 误差（样本均值 - 真实均值）')
pyplot.title('Law of Large Numbers: Sample Mean Convergence to Population Mean / 大数法则：样本均值收敛到总体均值')
pyplot.grid(True, alpha=0.3)
pyplot.show()
```

## Learning Notes / 学习笔记

- **Statistical Concept / 统计学概念:** The Law of Large Numbers (LLN) is a cornerstone of probability theory and statistics. It states that the sample mean $\bar{X}_n$ converges to the true population mean $\mu$ as the sample size $n \to \infty$. Weak LLN says convergence occurs in probability; strong LLN says almost sure convergence. The error typically decreases as $O(1/\sqrt{n})$, so variance decreases as $O(1/n)$. This justifies using sample statistics to estimate population parameters and explains why larger samples provide more reliable estimates—a fundamental principle in statistical inference.

- **ML Application / 机器学习应用:** The LLN motivates many ML practices: cross-validation uses multiple folds to average estimates (reducing variance); ensemble methods average predictions to approach the true underlying function; federated learning aggregates updates across clients to approximate global gradients. In online learning and streaming data, the LLN justifies that accumulated experience leads to better-than-random performance. Sample complexity bounds in learning theory quantify how large $n$ must be to achieve reliable generalization—directly applying the LLN principle. Regularization and confidence intervals both rely on understanding convergence rates.

## Complete Code / 完整代码一览

```python
# ===== Section 1: Imports =====
from numpy.random import seed
from numpy.random import randn
from numpy import mean
from numpy import array
from matplotlib import pyplot

# ===== Section 2: Set Seed and Define Sample Sizes =====
# Set seed for reproducibility / 设置种子以保证可重现性
seed(1)

# Create list of increasing sample sizes / 创建递增的样本大小列表
sizes = list()
for x in range(10, 20000, 200):
    sizes.append(x)

# ===== Section 3: Generate Samples and Compute Means =====
# For each sample size, compute mean of samples from N(50, 5) / 对于每个样本大小，计算来自N(50, 5)的样本均值
means = [mean(5 * randn(size) + 50) for size in sizes]

# ===== Section 4: Visualize Convergence =====
# Plot error (sample mean - true mean=50) vs sample size / 绘制误差(样本均值 - 真实均值=50)对样本大小
pyplot.scatter(sizes, array(means) - 50, alpha=0.5)
pyplot.xlabel('Sample Size / 样本大小')
pyplot.ylabel('Error (Sample Mean - True Mean) / 误差（样本均值 - 真实均值）')
pyplot.title('Law of Large Numbers: Sample Mean Convergence to Population Mean / 大数法则：样本均值收敛到总体均值')
pyplot.grid(True, alpha=0.3)
pyplot.show()
```
