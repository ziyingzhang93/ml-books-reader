# Chapter 7: Law of Large Numbers
# 第7章：大数法则

## Theme | 主题
From ideal population to empirical convergence: sample means stabilize as sample size grows.
从理想总体到经验收敛：样本均值随样本量增长而稳定。

## Evolution Roadmap | 演变路线图
```
Population (Ideal Distribution)
└─ Sample Mean Convergence Plot
   (Rolling average stabilizes → theoretical mean)
```

## Progression Logic | 进度逻辑

### Stage 1: Theoretical Foundation (理论基础)
**English:** Define the ideal population distribution and its true mean (μ).
**中文:** 定义理想总体分布及其真实均值(μ)。

### Stage 2: Empirical Demonstration (经验演示)
**English:** Generate a large sample and compute rolling mean at intervals (n=10, 100, 1000, ...). Plot shows convergence: smaller n exhibits high variance, larger n approaches μ.
**中文:** 生成大型样本并按间隔计算滚动均值(n=10、100、1000...)。图显示收敛：较小的n显示高方差，较大的n接近μ。

## ML Relevance | ML相关性

1. **Sampling & Estimation (采样与估计)**: LLN justifies using sample statistics to estimate population parameters.
2. **Model Validation (模型验证)**: Performance metrics (accuracy, loss) converge to true error as test set grows.
3. **Statistical Inference (统计推断)**: Larger datasets enable more reliable parameter estimates and narrower confidence intervals.
4. **Computational Efficiency (计算效率)**: Understanding convergence rates helps choose optimal batch sizes and training iterations.
