## Summary / 摘要

**English:** This notebook demonstrates shuffling NumPy arrays in place. NumPy's `shuffle()` function efficiently randomizes the order of array elements along the first axis, essential for data preprocessing and randomized algorithms.

**中文:** 本笔记本演示了原地随机排列NumPy数组。NumPy的`shuffle()`函数有效地沿第一轴随机化数组元素的顺序，对于数据预处理和随机算法至关重要。

## Step 1 — Import Libraries / 导入库

```python
# Import NumPy seed and shuffle functions / 导入NumPy种子和shuffle函数
from numpy.random import seed
from numpy.random import shuffle
```

## Step 2 — Set Seed and Create Array / 设置种子并创建数组

```python
# Set NumPy seed for reproducibility / 设置NumPy种子以保证可重现性
seed(1)

# Create array of integers 0-19 / 创建0-19的整数数组
sequence = [i for i in range(20)]
print(f"Original sequence: {sequence}")
```

## Step 3 — Shuffle Array In Place / 原地随机排列数组

```python
# Shuffle the array in place (modifies original list) / 原地排列数组（修改原始列表）
# numpy.random.shuffle() returns None and modifies the array directly / numpy.random.shuffle()返回None并直接修改数组
shuffle(sequence)

print(f"\nShuffled sequence:   {sequence}")
print(f"\nNote: shuffle() modifies the array in place along the first axis.")
```

## Learning Notes / 学习笔记

- **Statistical Concept / 统计学概念:** NumPy's `shuffle()` randomly permutes the first axis of an array in place. For a 1D array, it shuffles elements; for a 2D array, it shuffles rows (preserving row structure). The function uses the same efficient Fisher-Yates algorithm as Python's `shuffle()`. Each permutation has equal probability. Unlike `numpy.random.permutation()`, which returns a new array, `shuffle()` modifies the input directly, saving memory. This is the standard approach for randomizing data order before train/test splits.

- **ML Application / 机器学习应用:** NumPy's `shuffle()` is ubiquitous in ML pipelines: shuffling features for feature importance estimation, randomizing sample order before gradient descent, separating training and test indices, and implementing curriculum learning. In data preprocessing, shuffling ensures that any ordering bias in the dataset doesn't affect model training. For fairness and bias mitigation, shuffling can be stratified to preserve group distributions. Large-scale implementations rely on `shuffle()` for efficient in-place randomization of millions of samples.

## Complete Code / 完整代码一览

```python
# ===== Section 1: Imports =====
from numpy.random import seed
from numpy.random import shuffle

# ===== Section 2: Create Array =====
# Set NumPy seed for reproducibility / 设置NumPy种子以保证可重现性
seed(1)

# Create array of integers / 创建整数数组
sequence = [i for i in range(20)]
print(f"Original sequence: {sequence}")

# ===== Section 3: Shuffle In Place =====
# Shuffle array in place / 原地随机排列数组
shuffle(sequence)

print(f"\nShuffled sequence:   {sequence}")
print(f"\nNote: shuffle() modifies the array in place along the first axis.")
```
