## Summary / 摘要

**English:** This notebook generates random samples from a standard normal distribution using NumPy's `randn()` function. NumPy's `randn()` returns arrays of Gaussian values and is orders of magnitude faster than Python's `random.gauss()` for large arrays.

**中文:** 本笔记本使用NumPy的`randn()`函数从标准正态分布生成随机样本。NumPy的`randn()`返回高斯值数组，对于大数组比Python的`random.gauss()`快得多。

**Formula / 公式:**
$$Z \sim N(0, 1)$$

## Step 1 — Import Libraries / 导入库

```python
# Import NumPy seed and randn functions / 导入NumPy种子和randn函数
from numpy.random import seed
from numpy.random import randn
```

## Step 2 — Set Seed / 设置种子

```python
# Set NumPy seed for reproducibility / 设置NumPy种子以保证可重现性
seed(1)
```

## Step 3 — Generate Gaussian Array / 生成高斯数组

```python
# Generate 10 random samples from N(0, 1) - standard normal / 从N(0, 1)生成10个随机样本 - 标准正态分布
# numpy.random.randn(*shape) returns array of standard normal values / numpy.random.randn()返回标准正态值数组
values = randn(10)

print("10 random values from standard normal N(0,1) (NumPy array):")
print(values)
print(f"\nArray shape: {values.shape}")
print(f"Mean (approx): {values.mean():.4f}")
print(f"Std dev (approx): {values.std():.4f}")
```

## Learning Notes / 学习笔记

- **Statistical Concept / 统计学概念:** NumPy's `randn(*shape)` generates samples from the standard normal distribution $N(0, 1)$. The function accepts shape as arguments: `randn(10)` returns a 1D array of 10 values, while `randn(3, 4)` returns a 3×4 matrix. The distribution is characterized by mean 0 and variance 1. Any Gaussian $N(\mu, \sigma^2)$ can be generated via linear transformation: $X = \sigma Z + \mu$ where $Z \sim N(0, 1)$. NumPy's implementation uses the Box-Muller transform or ziggurat algorithm for efficiency.

- **ML Application / 机器学习应用:** NumPy's `randn()` is essential for neural network weight initialization (He, Xavier, Glorot initializations start with `randn()`), generative models, and Bayesian inference. In variational autoencoders (VAEs), the latent space is sampled using `randn()`. Stochastic regularization (dropout, noise injection) relies on Gaussian samples. In uncertainty estimation and Bayesian neural networks, predictions are modulated by Gaussian noise. Efficient generation of large Gaussian arrays is fundamental to scaling deep learning and probabilistic models.
