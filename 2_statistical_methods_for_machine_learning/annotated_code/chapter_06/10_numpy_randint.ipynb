## Summary / 摘要

**English:** This notebook generates random integers using NumPy's `randint()` function. NumPy's `randint()` is vectorized and returns arrays of random integers, making it efficient for large-scale discrete sampling.

**中文:** 本笔记本使用NumPy的`randint()`函数生成随机整数。NumPy的`randint()`是向量化的，返回随机整数数组，使其对大规模离散采样高效。

## Step 1 — Import Libraries / 导入库

```python
# Import NumPy seed and randint functions / 导入NumPy种子和randint函数
from numpy.random import seed
from numpy.random import randint
```

## Step 2 — Set Seed / 设置种子

```python
# Set NumPy seed for reproducibility / 设置NumPy种子以保证可重现性
seed(1)
```

## Step 3 — Generate Random Integer Array / 生成随机整数数组

```python
# Generate 20 random integers from [0, 10) / 从[0, 10)生成20个随机整数
# numpy.random.randint(low, high, size) returns array of shape 'size' / numpy.random.randint()返回形状为'size'的数组
# Unlike Python's randint, NumPy's does not include 'high' / 与Python的randint不同，NumPy的不包括'high'
values = randint(0, 10, 20)

print("20 random integers from discrete U[0,10) (NumPy array):")
print(values)
print(f"\nArray shape: {values.shape}")
print(f"Min: {values.min()}, Max: {values.max()}")
```

## Learning Notes / 学习笔记

- **Statistical Concept / 统计学概念:** NumPy's `randint(low, high, size)` generates random integers in the range [low, high) with specified output shape. The integers follow a discrete uniform distribution where each value has equal probability. Note the difference from Python's `random.randint(a, b)`: NumPy uses [low, high) (excludes high), while Python uses [a, b] (includes b). For $k = \text{high} - \text{low}$ possible values, each has probability $1/k$. This is efficient for generating large arrays of random indices.

- **ML Application / 机器学习应用:** NumPy's `randint()` is standard in ML for: generating random batch indices in SGD, creating random augmentations (crop positions, rotation angles as integers), sampling negative examples in contrastive learning, and index-based data shuffling. Random search over integer hyperparameters uses `randint()` to generate candidate values. In graph neural networks, random edge sampling uses integer indices from `randint()`. Efficiently generating large arrays of random integers is essential for scalable stochastic algorithms.
