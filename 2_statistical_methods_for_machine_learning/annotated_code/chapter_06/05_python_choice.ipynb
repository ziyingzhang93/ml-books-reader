## Summary / 摘要

**English:** This notebook demonstrates random selection from a sequence with replacement. The `choice()` function selects one element randomly from a list, useful for random sampling, shuffling choices, and simulation.

**中文:** 本笔记本演示了从序列中有放回地随机选择。`choice()`函数随机从列表中选择一个元素，用于随机采样、打乱选择和模拟。

## Step 1 — Import Libraries / 导入库

```python
# Import seed and choice functions / 导入种子和choice函数
from random import seed
from random import choice
```

## Step 2 — Set Seed and Create Sequence / 设置种子并创建序列

```python
# Set seed for reproducibility / 设置种子以保证可重现性
seed(1)

# Create a sequence of integers 0-19 / 创建0-19的整数序列
sequence = [i for i in range(20)]
print(f"Original sequence: {sequence}")
```

## Step 3 — Make Random Choices / 进行随机选择

```python
# Make 5 random selections FROM the sequence (with replacement) / 从序列中进行5次随机选择（有放回）
# Each element has equal probability of being selected / 每个元素被选中的概率相等
# An element can be selected multiple times / 一个元素可以被多次选中
print("\n5 random choices with replacement:")
for i in range(5):
    selection = choice(sequence)
    print(f"  {i+1}: {selection}")
```

## Learning Notes / 学习笔记

- **Statistical Concept / 统计学概念:** Sampling with replacement means each selection is independent—previous selections don't affect probabilities. If a sequence has $n$ elements, each has probability $1/n$ of being chosen on any draw. Over many draws, each element is chosen approximately $n$ times if we make $n^2$ selections. This contrasts with sampling without replacement (used in `sample()`), where each element can be selected at most once. With replacement is easier to analyze mathematically and is standard for Monte Carlo methods.

- **ML Application / 机器学习应用:** Sampling with replacement is fundamental in bootstrap resampling—a cornerstone technique in statistical learning for estimating confidence intervals and out-of-bag errors. Random Forest and bagging use sampling with replacement to create bootstrap datasets. In reinforcement learning, experience replay uses sampling to select mini-batches from memory with replacement. Importance sampling and stratified sampling leverage random choice for unbiased estimation of model performance and uncertainty quantification.

## Complete Code / 完整代码一览

```python
# ===== Section 1: Imports =====
from random import seed
from random import choice

# ===== Section 2: Create Sequence =====
# Set seed for reproducibility / 设置种子以保证可重现性
seed(1)

# Create sequence of integers / 创建整数序列
sequence = [i for i in range(20)]
print(f"Original sequence: {sequence}")

# ===== Section 3: Make Random Choices =====
# Select 5 elements with replacement / 选择5个有放回的元素
print("\n5 random choices with replacement:")
for i in range(5):
    selection = choice(sequence)
    print(f"  {i+1}: {selection}")
```
