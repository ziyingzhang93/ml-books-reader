## Summary / 摘要

**English:** This notebook demonstrates seeding NumPy's random number generator. NumPy has its own separate PRNG state, independent from Python's `random` module. Setting the NumPy seed ensures reproducible array generation and vectorized random operations.

**中文:** 本笔记本演示了NumPy随机数生成器的种子设置。NumPy有自己独立的PRNG状态，独立于Python的`random`模块。设置NumPy种子确保可重现的数组生成和向量化随机操作。

## Step 1 — Import Libraries / 导入库

```python
# Import NumPy seed and random array generation / 导入NumPy种子和随机数组生成
from numpy.random import seed
from numpy.random import rand
```

## Step 2 — Set NumPy Seed and Generate First Array / 设置NumPy种子并生成第一个数组

```python
# Set NumPy seed / 设置NumPy种子
seed(1)

# Generate random array of shape (3,) / 生成形状(3,)的随机数组
print("First array with seed(1):")
print(rand(3))
```

## Step 3 — Reset NumPy Seed and Verify Reproducibility / 重置NumPy种子并验证可重现性

```python
# Reset NumPy seed to same value / 将NumPy种子重置为相同值
seed(1)

# Generate again - should be identical / 再次生成 - 应该相同
print("\nSecond array with seed(1) again:")
print(rand(3))

print("\nNote: NumPy's seed() is separate from Python's random.seed()")
```

## Learning Notes / 学习笔记

- **Statistical Concept / 统计学概念:** NumPy maintains its own independent PRNG state separate from Python's `random` module. This allows users to seed NumPy independently without affecting Python's built-in random functions. NumPy uses the Mersenne Twister algorithm by default, a high-quality PRNG with a very long period ($2^{19937} - 1$). Modern NumPy (>= 1.17) also supports alternative generators like PCG64 for different quality/speed tradeoffs. Seeding NumPy ensures reproducible scientific computing workflows.

- **ML Application / 机器学习应用:** NumPy seeding is essential for reproducible ML pipelines. Most libraries (scikit-learn, TensorFlow, PyTorch) recommend setting NumPy's seed along with their own seeds for full reproducibility. NumPy's vectorized random operations are faster than Python's `random` module, making them standard in data science. When combined with framework-specific seeds (e.g., `tf.random.set_seed()` in TensorFlow), NumPy seeding ensures all stochastic components are deterministic for experiment replication.

## Complete Code / 完整代码一览

```python
# ===== Section 1: Imports =====
from numpy.random import seed
from numpy.random import rand

# ===== Section 2: Generate First Random Array =====
# Set NumPy seed to initialize generator / 设置NumPy种子以初始化生成器
seed(1)

# Generate random array / 生成随机数组
print("First array with seed(1):")
print(rand(3))

# ===== Section 3: Demonstrate Reproducibility =====
# Reset NumPy seed to same value / 将NumPy种子重置为相同值
seed(1)

# Generate again - should be identical / 再次生成 - 应该相同
print("\nSecond array with seed(1) again:")
print(rand(3))

print("\nNote: NumPy's seed() is separate from Python's random.seed()")
```
