## Summary / 摘要

A confidence interval provides a range estimate for a population parameter (e.g., proportion, mean) based on sample data. For a binomial proportion p with sample size n and success count k, the 95% confidence interval is calculated as: p ± 1.96×√(p(1-p)/n). With n=50 and p=0.2 (10 successes), the interval quantifies uncertainty in the estimated proportion. A 95% CI means that if we repeated sampling many times, approximately 95% of calculated intervals would contain the true population proportion.

置信区间基于样本数据为总体参数（例如比例、均值）提供范围估计。对于二项式比例p、样本大小n和成功计数k，95%置信区间计算为：p ± 1.96×√(p(1-p)/n)。对于n=50和p=0.2（10次成功），该区间量化了估计比例中的不确定性。95%CI意味着如果我们多次重复抽样，大约95%的计算区间将包含真实的总体比例。

## Step 1 — Import Libraries / 导入库

```python3
# Import required libraries
# 导入所需库
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
# 设置随机种子以保证可重复性
np.random.seed(42)
```

## Step 2 — Define Sample Parameters / 定义样本参数

```python3
# Sample parameters
# 样本参数
n = 50  # Sample size / 样本大小
k = 10  # Number of successes / 成功次数
p = k / n  # Sample proportion / 样本比例

# Confidence level
# 置信水平
confidence_level = 0.95  # 95% confidence interval / 95%置信区间
z_critical = 1.96  # Critical value for 95% / 95%的临界值

# Display parameters
# 显示参数
print(f"Sample size (n): {n}")
print(f"Number of successes (k): {k}")
print(f"Sample proportion (p): {p:.4f}")
print(f"Confidence level: {confidence_level*100:.0f}%")
print(f"Z-critical value: {z_critical}")
```

## Step 3 — Calculate Standard Error / 计算标准误差

```python3
# Calculate standard error of the proportion
# 计算比例的标准误差
se = np.sqrt(p * (1 - p) / n)

# Display standard error
# 显示标准误差
print(f"\nStandard Error Calculation:")
print(f"SE = sqrt(p(1-p)/n)")
print(f"SE = sqrt({p:.4f} × {1-p:.4f} / {n})")
print(f"SE = {se:.4f}")
```

## Step 4 — Calculate Confidence Interval / 计算置信区间

```python3
# Calculate margin of error
# 计算误差范围
margin_of_error = z_critical * se

# Calculate confidence interval bounds
# 计算置信区间边界
ci_lower = p - margin_of_error
ci_upper = p + margin_of_error

# Ensure bounds are within [0, 1]
# 确保边界在[0, 1]内
ci_lower = max(0, ci_lower)
ci_upper = min(1, ci_upper)

# Display confidence interval
# 显示置信区间
print(f"\nConfidence Interval Calculation:")
print(f"Margin of Error = Z × SE")
print(f"Margin of Error = {z_critical} × {se:.4f} = {margin_of_error:.4f}")
print(f"\nCI = p ± Margin of Error")
print(f"CI = {p:.4f} ± {margin_of_error:.4f}")
print(f"\n95% Confidence Interval: [{ci_lower:.4f}, {ci_upper:.4f}]")
print(f"Or as percentages: [{ci_lower*100:.2f}%, {ci_upper*100:.2f}%]")
```

## Step 5 — Visualize Confidence Interval / 可视化置信区间

```python3
# Create visualization
# 创建可视化
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Point estimate with error bar
# 图1: 点估计与误差条
axes[0].errorbar(1, p, yerr=margin_of_error, fmt='o', markersize=10, 
                 capsize=15, capthick=2, linewidth=2, color='blue', label='Sample proportion')
axes[0].axhline(ci_lower, color='green', linestyle='--', linewidth=2, label='CI bounds')
axes[0].axhline(ci_upper, color='green', linestyle='--', linewidth=2)
axes[0].fill_between([0.5, 1.5], ci_lower, ci_upper, alpha=0.2, color='green')

axes[0].set_xlim(0.5, 1.5)
axes[0].set_ylim(-0.1, 0.5)
axes[0].set_xticks([1])
axes[0].set_xticklabels(['Sample'])
axes[0].set_ylabel('Proportion')
axes[0].set_title(f'{confidence_level*100:.0f}% Confidence Interval for Proportion')
axes[0].grid(True, alpha=0.3, axis='y')
axes[0].legend()

# Add text annotations
# 添加文本注释
axes[0].text(1.15, p, f'p={p:.4f}', fontsize=10, va='center')
axes[0].text(1.15, ci_upper, f'{ci_upper:.4f}', fontsize=10, va='center')
axes[0].text(1.15, ci_lower, f'{ci_lower:.4f}', fontsize=10, va='center')

# Plot 2: Distribution of the sample proportion
# 图2: 样本比例的分布
x = np.linspace(p - 4*se, p + 4*se, 100)
from scipy import stats
y = stats.norm.pdf(x, p, se)

axes[1].plot(x, y, 'b-', linewidth=2, label='Sampling distribution')
axes[1].axvline(p, color='blue', linestyle='-', linewidth=2, label='Sample proportion')
axes[1].axvline(ci_lower, color='green', linestyle='--', linewidth=2, label='CI bounds')
axes[1].axvline(ci_upper, color='green', linestyle='--', linewidth=2)

# Shade the CI region
# 为CI区域着色
x_fill = x[(x >= ci_lower) & (x <= ci_upper)]
y_fill = stats.norm.pdf(x_fill, p, se)
axes[1].fill_between(x_fill, y_fill, alpha=0.3, color='green')

axes[1].set_xlabel('Proportion')
axes[1].set_ylabel('Density')
axes[1].set_title('Sampling Distribution with 95% CI')
axes[1].grid(True, alpha=0.3)
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python3
## Learning Notes / 学习笔记

- **Statistical Concept**: The confidence interval uses the normal approximation to the binomial distribution, valid when np ≥ 10 and n(1-p) ≥ 10. The z-critical value of 1.96 comes from the standard normal distribution, capturing 95% of the area under the curve. The standard error accounts for both the proportion (p) and sample size (n).
  
  **统计概念**: 置信区间使用二项分布的正态近似，当np ≥ 10且n(1-p) ≥ 10时有效。z临界值1.96来自标准正态分布，捕获曲线下方95%的面积。标准误差同时考虑了比例(p)和样本大小(n)。

- **ML Application**: Confidence intervals quantify uncertainty in model predictions and performance metrics (e.g., classification accuracy, recall). Essential in A/B testing, reporting business metrics, and making decisions about model deployment. Wide intervals indicate high uncertainty; narrow intervals allow confident decision-making.
  
  **ML应用**: 置信区间量化了模型预测和性能指标（例如分类准确性、召回率）中的不确定性。在A/B测试、报告业务指标和制定模型部署决策中至关重要。宽区间表示高不确定性；窄区间允许自信的决策制定。
```

## Complete Code / 完整代码一览

```python3
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

n, k = 50, 10
p = k / n
z_critical = 1.96

se = np.sqrt(p * (1 - p) / n)
margin_of_error = z_critical * se

ci_lower = max(0, p - margin_of_error)
ci_upper = min(1, p + margin_of_error)

print(f"Sample proportion: {p:.4f}")
print(f"Standard error: {se:.4f}")
print(f"95% CI: [{ci_lower:.4f}, {ci_upper:.4f}]")
print(f"Or: [{ci_lower*100:.2f}%, {ci_upper*100:.2f}%]")
```
