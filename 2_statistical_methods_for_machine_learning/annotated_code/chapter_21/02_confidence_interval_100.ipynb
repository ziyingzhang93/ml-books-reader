## Summary / 摘要

This notebook demonstrates the effect of increasing sample size from n=50 to n=100 on the confidence interval width. With the same proportion p=0.2 (now 20 successes out of 100), the larger sample yields a narrower confidence interval, reflecting reduced estimation uncertainty. The comparison illustrates the inverse square root relationship between sample size and standard error: increasing n by 2x reduces SE by √2, making the CI approximately 30% narrower.

本笔记本演示了将样本大小从n=50增加到n=100对置信区间宽度的影响。对于相同的比例p=0.2（现在是100中的20次成功），较大的样本产生更窄的置信区间，反映了估计不确定性的降低。比较说明了样本大小和标准误差之间的倒平方根关系：将n增加2倍将SE减少√2，使CI变窄约30%。

## Step 1 — Import Libraries / 导入库

```python
# Import required libraries
# 导入所需库
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Set random seed for reproducibility
# 设置随机种子以保证可重复性
np.random.seed(42)
```

```python
## Step 2 — Define Sample Parameters for n=100 / 定义n=100的样本参数
```

```python
# Sample parameters for larger sample
# 较大样本的样本参数
n = 100  # Larger sample size / 较大的样本大小
k = 20  # Number of successes / 成功次数
p = k / n  # Sample proportion / 样本比例

# Confidence level
# 置信水平
confidence_level = 0.95  # 95% confidence interval / 95%置信区间
z_critical = 1.96  # Critical value for 95% / 95%的临界值

# Display parameters
# 显示参数
print(f"Sample size (n): {n}")
print(f"Number of successes (k): {k}")
print(f"Sample proportion (p): {p:.4f}")
print(f"Confidence level: {confidence_level*100:.0f}%")
print(f"Z-critical value: {z_critical}")
```

## Step 3 — Calculate Standard Error and CI / 计算标准误差和置信区间

```python
# Calculate standard error
# 计算标准误差
se = np.sqrt(p * (1 - p) / n)

# Calculate margin of error
# 计算误差范围
margin_of_error = z_critical * se

# Calculate confidence interval bounds
# 计算置信区间边界
ci_lower = max(0, p - margin_of_error)
ci_upper = min(1, p + margin_of_error)

# Display results
# 显示结果
print(f"\nStandard Error Calculation:")
print(f"SE = sqrt(p(1-p)/n) = sqrt({p:.4f} × {1-p:.4f} / {n}) = {se:.4f}")
print(f"\nMargin of Error:")
print(f"ME = Z × SE = {z_critical} × {se:.4f} = {margin_of_error:.4f}")
print(f"\n95% Confidence Interval: [{ci_lower:.4f}, {ci_upper:.4f}]")
print(f"Or as percentages: [{ci_lower*100:.2f}%, {ci_upper*100:.2f}%]")
```

```python
## Step 4 — Compare with n=50 / 与n=50比较
```

```python
# Previous results from n=50
# n=50的前期结果
n_prev = 50
k_prev = 10
p_prev = k_prev / n_prev
se_prev = np.sqrt(p_prev * (1 - p_prev) / n_prev)
margin_prev = z_critical * se_prev
ci_lower_prev = max(0, p_prev - margin_prev)
ci_upper_prev = min(1, p_prev + margin_prev)

# Compare the two
# 比较两者
print(f"\nComparison: n=50 vs n=100")
print(f"{'':20} | {'n=50':15} | {'n=100':15}")
print("-" * 55)
print(f"{'Sample proportion':20} | {p_prev:15.4f} | {p:15.4f}")
print(f"{'Standard error':20} | {se_prev:15.4f} | {se:15.4f}")
print(f"{'Margin of error':20} | {margin_prev:15.4f} | {margin_of_error:15.4f}")
print(f"{'CI lower':20} | {ci_lower_prev:15.4f} | {ci_lower:15.4f}")
print(f"{'CI upper':20} | {ci_upper_prev:15.4f} | {ci_upper:15.4f}")
print(f"{'CI width':20} | {ci_upper_prev-ci_lower_prev:15.4f} | {ci_upper-ci_lower:15.4f}")

# Calculate reductions
# 计算减少
se_reduction_pct = (1 - se / se_prev) * 100
width_reduction_pct = (1 - (ci_upper - ci_lower) / (ci_upper_prev - ci_lower_prev)) * 100

print(f"\nStandard error reduction: {se_reduction_pct:.1f}%")
print(f"CI width reduction: {width_reduction_pct:.1f}%")
print(f"Expected reduction (1/sqrt(2)): {(1 - 1/np.sqrt(2))*100:.1f}%")
```

## Step 5 — Visualize Comparison / 可视化比较

```python
# Create side-by-side comparison
# 创建并排比较
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

# Plot 1: Side-by-side error bars
# 图1: 并排误差条
positions = [1, 2]
proportions = [p_prev, p]
errors = [margin_prev, margin_of_error]

axes[0].errorbar(positions, proportions, yerr=errors, fmt='o', markersize=10,
                capsize=15, capthick=2, linewidth=2, color=['blue', 'red'])
axes[0].set_xlim(0.5, 2.5)
axes[0].set_ylim(0, 0.4)
axes[0].set_xticks(positions)
axes[0].set_xticklabels(['n=50', 'n=100'])
axes[0].set_ylabel('Proportion')
axes[0].set_title('95% CI Comparison: n=50 vs n=100')
axes[0].grid(True, alpha=0.3, axis='y')

# Add value labels
# 添加值标签
for i, (pos, prop, err) in enumerate(zip(positions, proportions, errors)):
    axes[0].text(pos, prop + err + 0.02, f'{prop:.4f}', ha='center', fontsize=10)
    axes[0].text(pos, prop - err - 0.02, f'±{err:.4f}', ha='center', fontsize=9)

# Plot 2: Standard error comparison
# 图2: 标准误差比较
sample_sizes = [50, 100]
standard_errors = [se_prev, se]

axes[1].bar(sample_sizes, standard_errors, color=['blue', 'red'], alpha=0.7, edgecolor='black')
axes[1].set_xlabel('Sample Size')
axes[1].set_ylabel('Standard Error')
axes[1].set_title('Standard Error Comparison')
axes[1].set_xticks(sample_sizes)
axes[1].grid(True, alpha=0.3, axis='y')

# Add value labels
for size, se_val in zip(sample_sizes, standard_errors):
    axes[1].text(size, se_val + 0.002, f'{se_val:.4f}', ha='center', fontsize=10)

# Plot 3: Sampling distributions comparison
# 图3: 采样分布比较
x_prev = np.linspace(p_prev - 4*se_prev, p_prev + 4*se_prev, 100)
y_prev = stats.norm.pdf(x_prev, p_prev, se_prev)

x_new = np.linspace(p - 4*se, p + 4*se, 100)
y_new = stats.norm.pdf(x_new, p, se)

axes[2].plot(x_prev, y_prev, 'b-', linewidth=2, label='n=50')
axes[2].plot(x_new, y_new, 'r-', linewidth=2, label='n=100')

# Fill CI regions
x_fill_prev = x_prev[(x_prev >= ci_lower_prev) & (x_prev <= ci_upper_prev)]
y_fill_prev = stats.norm.pdf(x_fill_prev, p_prev, se_prev)
axes[2].fill_between(x_fill_prev, y_fill_prev, alpha=0.2, color='blue')

x_fill_new = x_new[(x_new >= ci_lower) & (x_new <= ci_upper)]
y_fill_new = stats.norm.pdf(x_fill_new, p, se)
axes[2].fill_between(x_fill_new, y_fill_new, alpha=0.2, color='red')

axes[2].set_xlabel('Proportion')
axes[2].set_ylabel('Density')
axes[2].set_title('Sampling Distributions with 95% CI')
axes[2].legend()
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
## Learning Notes / 学习笔记

- **Statistical Concept**: The inverse square root relationship SE = √(p(1-p)/n) shows that doubling sample size reduces standard error by √2 ≈ 1.41. This fundamental law means that to halve the standard error, one must quadruple the sample size. As n increases, the sampling distribution becomes narrower and more concentrated, allowing more precise parameter estimation.
  
  **统计概念**: 倒平方根关系SE = √(p(1-p)/n)表明将样本大小加倍会将标准误差减少√2 ≈ 1.41。这个基本定律意味着要使标准误差减半，必须将样本大小增加四倍。随着n增加，采样分布变得更窄、更集中，允许更精确的参数估计。

- **ML Application**: This comparison demonstrates sample size planning in practice. For online experiments (A/B tests), doubling the sample size narrows confidence intervals by ~30%, enabling faster detection of significant effects. Trade-offs between sample size, experiment duration, and detection power are central to experimental design in machine learning applications.
  
  **ML应用**: 此比较演示了实践中的样本大小规划。对于在线实验（A/B测试），将样本大小增加一倍会将置信区间减少约30%，使得能更快速地检测显著效果。样本大小、实验持续时间和检测功率之间的权衡对于机器学习应用中的实验设计至关重要。
```

## Complete Code / 完整代码一览

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

z_critical = 1.96

# n=50
n_prev, k_prev = 50, 10
p_prev = k_prev / n_prev
se_prev = np.sqrt(p_prev * (1 - p_prev) / n_prev)
margin_prev = z_critical * se_prev
ci_lower_prev = max(0, p_prev - margin_prev)
ci_upper_prev = min(1, p_prev + margin_prev)

# n=100
n, k = 100, 20
p = k / n
se = np.sqrt(p * (1 - p) / n)
margin = z_critical * se
ci_lower = max(0, p - margin)
ci_upper = min(1, p + margin)

print(f"n=50:  CI = [{ci_lower_prev:.4f}, {ci_upper_prev:.4f}], width = {ci_upper_prev - ci_lower_prev:.4f}")
print(f"n=100: CI = [{ci_lower:.4f}, {ci_upper:.4f}], width = {ci_upper - ci_lower:.4f}")
print(f"Width reduction: {(1 - (ci_upper - ci_lower)/(ci_upper_prev - ci_lower_prev))*100:.1f}%")
```
