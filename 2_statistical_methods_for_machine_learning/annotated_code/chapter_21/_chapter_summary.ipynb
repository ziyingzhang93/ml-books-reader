# Chapter 21: Confidence Intervals
# 第21章：置信区间

## Theme | 主题
From parametric formulas to library functions to nonparametric bootstrap: quantifying uncertainty around estimates.
从参数公式到库函数到非参数自助法：量化估计周围的不确定性。

## Evolution Roadmap | 演变路线图
```
Manual Formula (Binomial CI for p)
├─ Small sample (n=50)
└─ Large sample (n=100)
   └─ Library Function (statsmodels)
      (Normal approx, Wilson score, Clopper-Pearson, etc.)
      └─ Bootstrap CI
         (nonparametric, no distributional assumption)
```

## Progression Logic | 进度逻辑

### Stage 1: Manual Binomial CI (手动二项CI)
**English:** For a proportion p, CI = p ± z * sqrt(p(1-p)/n). Demonstrates core idea: estimate ± critical_value * SE.
**中文:** 对于比例p，CI = p ± z * sqrt(p(1-p)/n)。演示核心思想：估计 ± 临界值 * SE。

### Stage 2: Sample Size Effect (样本量效应)
**English:** Compare CI width for n=50 vs. n=100. Larger n → smaller SE → narrower CI. Illustrates the sqrt(n) scaling law.
**中文:** 比较n=50与n=100的CI宽度。较大的n → 较小的SE → CI较窄。说明sqrt(n)缩放法则。

### Stage 3: Library Functions (库函数)
**English:** Use statsmodels or scipy to compute CI with different methods (Normal approx, Wilson score, Clopper-Pearson). Highlights method selection for different scenarios.
**中文:** 使用statsmodels或scipy通过不同方法计算CI(正态近似、Wilson分数、Clopper-Pearson)。突出不同场景的方法选择。

### Stage 4: Bootstrap CI (自助CI)
**English:** Nonparametric approach: resample data, compute statistic for each sample, extract percentile interval (e.g., 2.5th and 97.5th percentiles).
**中文:** 非参数方法：重采样数据，计算每个样本的统计，提取百分位区间(例如2.5%和97.5%百分位)。

## ML Relevance | ML相关性

1. **Estimate Uncertainty (估计不确定性)**: CI quantifies confidence in sample statistics (mean, proportion, AUC) as population estimates.
2. **Interpretation (解释)**: "95% CI" means if we repeated the experiment many times, 95% of CIs would contain the true parameter.
3. **Model Performance (模型性能)**: Report CI around accuracy, precision, recall to convey uncertainty, not just point estimates.
4. **Nonparametric Flexibility (非参数灵活性)**: Bootstrap CI works for any statistic without distributional assumptions.
