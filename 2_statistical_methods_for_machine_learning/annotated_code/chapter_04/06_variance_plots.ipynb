## Summary / 摘要

**English:** This notebook visualizes the impact of variance on the shape of a Gaussian distribution. We plot two normal distributions with the same mean (0) but different standard deviations ($\sigma = 0.5$ and $\sigma = 1.0$), demonstrating how variance controls the spread of the distribution.

**中文:** 本笔记本可视化方差对高斯分布形状的影响。我们绘制两个均值相同（0）但标准差不同（$\sigma = 0.5$ 和 $\sigma = 1.0$）的正态分布，演示方差如何控制分布的扩展。

**Formula / 公式:**
$$\sigma^2 = \text{Var}(X) = E[(X - \mu)^2]$$

## Step 1 — Import Libraries / 导入库

```python
# Import numerical and visualization libraries / 导入数值和可视化库
from numpy import arange
from matplotlib import pyplot
from scipy.stats import norm
```

## Step 2 — Create X-axis Range / 创建X轴范围

```python
# Generate x-values from -3 to 3 with fine resolution / 生成-3到3的x值，高分辨率
x_axis = arange(-3, 3, 0.001)
```

## Step 3 — Plot First Distribution (Lower Variance) / 绘制第一个分布（较低方差）

```python
# Calculate PDF for N(0, 0.5) - tighter, more peaked distribution / 计算N(0, 0.5)的PDF - 更紧凑、更尖锐的分布
# Smaller standard deviation means tighter concentration around mean / 较小的标准差意味着围绕均值的更紧密浓度
pyplot.plot(x_axis, norm.pdf(x_axis, 0, 0.5), label='σ=0.5')
```

## Step 4 — Plot Second Distribution (Higher Variance) / 绘制第二个分布（较高方差）

```python
# Calculate PDF for N(0, 1.0) - flatter, more spread distribution / 计算N(0, 1.0)的PDF - 更扁平、更分散的分布
# Larger standard deviation means wider spread / 较大的标准差意味着更广泛的扩展
pyplot.plot(x_axis, norm.pdf(x_axis, 0, 1), label='σ=1.0')

# Add labels and legend / 添加标签和图例
pyplot.xlabel('Value / 值')
pyplot.ylabel('Probability Density / 概率密度')
pyplot.title('Impact of Variance on Distribution Shape / 方差对分布形状的影响')
pyplot.legend()
pyplot.show()
```

## Learning Notes / 学习笔记

- **Statistical Concept / 统计学概念:** Variance ($\sigma^2$) measures the spread of a distribution. A larger variance indicates that values are dispersed far from the mean; smaller variance means values cluster tightly around the mean. The standard deviation $\sigma$ is the square root of variance and is on the same scale as the data. For a Gaussian distribution, approximately 68% of values fall within one standard deviation of the mean, 95% within two standard deviations, and 99.7% within three standard deviations (the empirical rule).

- **ML Application / 机器学习应用:** Understanding variance is critical in the bias-variance tradeoff—a fundamental concept in machine learning. A model with high variance overfits to training data, while one with low variance may underfit. Regularization techniques (L1, L2) penalize model complexity to reduce variance. In feature scaling and normalization, high-variance features can dominate low-variance ones, making variance estimation and standardization preprocessing steps essential for model stability and interpretability.

## Complete Code / 完整代码一览

```python
# ===== Section 1: Imports =====
from numpy import arange
from matplotlib import pyplot
from scipy.stats import norm

# ===== Section 2: Create X-axis =====
# Generate x-values for smooth curves / 生成光滑曲线的x值
x_axis = arange(-3, 3, 0.001)

# ===== Section 3: Plot Distributions with Different Variances =====
# Plot tighter distribution (lower variance) / 绘制更紧凑的分布（较低方差）
pyplot.plot(x_axis, norm.pdf(x_axis, 0, 0.5), label='σ=0.5')

# Plot wider distribution (higher variance) / 绘制更广泛的分布（较高方差）
pyplot.plot(x_axis, norm.pdf(x_axis, 0, 1), label='σ=1.0')

# Add labels and legend for clarity / 添加标签和图例以便清晰
pyplot.xlabel('Value / 值')
pyplot.ylabel('Probability Density / 概率密度')
pyplot.title('Impact of Variance on Distribution Shape / 方差对分布形状的影响')
pyplot.legend()
pyplot.show()
```
