## Summary / 摘要

**English:** This notebook calculates the standard deviation, the square root of variance. Standard deviation measures spread in the original units of the data and is widely used for interpretation. For N(50,5), we expect standard deviation $\approx 5$.

**中文:** 本笔记本计算标准差，它是方差的平方根。标准差用原始数据单位测量扩展，并广泛用于解释。对于N(50,5)，我们期望标准差 $\approx 5$。

**Formula / 公式:**
$$\sigma = \sqrt{\sigma^2} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (x_i - \mu)^2}$$

## Step 1 — Import Libraries / 导入库

```python
# Import standard deviation calculation and random generation / 导入标准差计算和随机生成
from numpy import std
from numpy.random import seed, randn
```

## Step 2 — Generate Gaussian Sample / 生成高斯样本

```python
# Set seed for reproducibility / 设置种子以保证可重现性
seed(1)

# Generate 10000 samples from N(50, 5) / 从N(50, 5)生成10000个样本
data = 5 * randn(10000) + 50
```

## Step 3 — Calculate Standard Deviation / 计算标准差

```python
# Calculate standard deviation (square root of variance) / 计算标准差（方差的平方根）
# numpy.std() computes the square root of variance / numpy.std()计算方差的平方根
result = std(data)

# Print the result with 3 decimal places / 用3位小数打印结果
print('Standard Deviation: %.3f' % result)
```

## Learning Notes / 学习笔记

- **Statistical Concept / 统计学概念:** Standard deviation $\sigma$ is the square root of variance and returns to the original scale of measurement. A Gaussian distribution with mean $\mu$ and standard deviation $\sigma$ follows the empirical rule: ~68% of values lie within $\pm 1\sigma$, ~95% within $\pm 2\sigma$, and ~99.7% within $\pm 3\sigma$. This makes standard deviation intuitive for describing uncertainty and confidence intervals. For instance, a 95% confidence interval for a Gaussian mean is approximately $\bar{x} \pm 1.96 \times \sigma/\sqrt{n}$.

- **ML Application / 机器学习应用:** Standard deviation is essential for feature normalization (z-score standardization: $(x - \mu)/\sigma$), which centers features to mean 0 and scales them to unit variance. This preprocessing step improves convergence in gradient descent and enables fair comparison of features with different scales. Confidence intervals based on standard deviation guide uncertainty quantification in predictions. In anomaly detection, points more than 3 standard deviations from the mean are often flagged as outliers.

## Complete Code / 完整代码一览

```python
# ===== Section 1: Imports =====
from numpy import std
from numpy.random import seed, randn

# ===== Section 2: Generate Sample Data =====
# Set seed for reproducibility / 设置种子以保证可重现性
seed(1)

# Generate 10000 samples from N(50, 5) / 从N(50, 5)生成10000个样本
data = 5 * randn(10000) + 50

# ===== Section 3: Calculate Standard Deviation =====
# Compute standard deviation as interpretable spread measure / 计算标准差作为可解释的扩展测度
result = std(data)

# Display the result / 显示结果
print('Standard Deviation: %.3f' % result)
```
