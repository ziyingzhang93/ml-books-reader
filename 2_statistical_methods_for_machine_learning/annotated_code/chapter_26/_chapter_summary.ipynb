# Chapter 26: Five-Number Summary
# 第26章：五数字摘要

## Theme | 主题
Robust descriptive statistics: minimum, Q1, median, Q3, maximum capture distribution shape without assuming normality.
鲁棒描述性统计：最小、Q1、中位数、Q3、最大值捕获分布形状，无需假设正态性。

## Evolution Roadmap | 演变路线图
```
Dataset (observations)
└─ Five-Number Summary
   min, Q1 (25th percentile), median (50th percentile),
   Q3 (75th percentile), max
   └─ Boxplot Visualization
      (box = IQR, whiskers = range, outliers = dots)
```

## Progression Logic | 进度逻辑

### Stage 1: Quantile-Based Summary (基于分位数的摘要)
**English:** Instead of mean/std (Gaussian-dependent), use quantiles (distribution-free). Example: [min, Q1, median, Q3, max] = [1, 4, 5, 8, 10] summarizes 10 values.
**中文:** 而不是均值/标准差(依赖高斯)，使用分位数(无分布)。例如：[min、Q1、中位数、Q3、max] = [1、4、5、8、10]总结10个值。

### Stage 2: Interpretation (解释)
**English:** IQR (Q3 - Q1) = middle 50% of data. Range = Q3 - Q1 + outlier detection: any value > Q3 + 1.5*IQR or < Q1 - 1.5*IQR is flagged as outlier.
**中文:** IQR(Q3 - Q1) = 数据的中间50%。范围 = Q3 - Q1 + 异常值检测：任何值 > Q3 + 1.5*IQR或 < Q1 - 1.5*IQR被标记为异常值。

### Stage 3: Boxplot Visualization (箱线图可视化)
**English:** Box spans Q1 to Q3, line inside is median, whiskers extend to ±1.5*IQR (or min/max), points beyond are outliers. Compact visual summary.
**中文:** 框从Q1跨到Q3，内部线是中位数，须须延伸至±1.5*IQR(或最小/最大)，超越的点是异常值。紧凑的视觉总结。

### Stage 4: Robustness (鲁棒性)
**English:** Five-number summary and boxplot are robust: outliers don't affect percentiles, unlike mean/std. Good for skewed and non-normal data.
**中文:** 五数字摘要和箱线图是鲁棒的：异常值不影响百分位，不像均值/标准差。适合偏斜和非正态数据。

## ML Relevance | ML相关性

1. **EDA & Outlier Detection (EDA与异常值检测)**: Boxplot reveals outliers and data spread at a glance.
2. **Group Comparison (组比较)**: Side-by-side boxplots compare distributions of multiple groups visually.
3. **Preprocessing (预处理)**: IQR-based outlier detection guides data cleaning before model training.
4. **Non-Normal Data (非正态数据)**: Five-number summary applies to any distribution, not just Gaussian.
