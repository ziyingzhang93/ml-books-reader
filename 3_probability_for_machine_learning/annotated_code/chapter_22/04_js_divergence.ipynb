## Summary / 汇总

This notebook calculates Jensen-Shannon divergence from scratch. JS(P||Q) = 0.5*KL(P||M) + 0.5*KL(Q||M) where M=(P+Q)/2. It's symmetric and bounded.

本笔记本从头计算Jensen-Shannon散度。JS(P||Q) = 0.5*KL(P||M) + 0.5*KL(Q||M)，其中M=(P+Q)/2。它是对称的且有界的。

## Step 1 — Define and Calculate Jensen-Shannon Divergence / 定义并计算Jensen-Shannon散度

```python
from math import log2, sqrtfrom numpy import asarray# calculate the kl divergencedef kl_divergence(p, q):    return sum(p[i] * log2(p[i]/q[i]) for i in range(len(p)))# calculate the jensen-shannon divergence (symmetric)def js_divergence(p, q):    # M = (P + Q) / 2: midpoint distribution    m = 0.5 * (p + q)    # JS = 0.5 * KL(P||M) + 0.5 * KL(Q||M)    return 0.5 * kl_divergence(p, m) + 0.5 * kl_divergence(q, m)# define distributionsp = asarray([0.10, 0.40, 0.50])q = asarray([0.80, 0.15, 0.05])# calculate JS(P || Q)js_pq = js_divergence(p, q)print('JS(P || Q) divergence: %.3f bits' % js_pq)print('JS(P || Q) distance: %.3f' % sqrt(js_pq))# calculate JS(Q || P) - should be same as JS(P || Q)js_qp = js_divergence(q, p)print('JS(Q || P) divergence: %.3f bits' % js_qp)print('JS(Q || P) distance: %.3f' % sqrt(js_qp))
```

## Learning Notes / 学习笔记

- **Concept**: Jensen-Shannon divergence is symmetric (JS(P||Q) = JS(Q||P)) and takes values in [0, 1] for binary log. JS distance = √JS divergence provides a true metric satisfying triangle inequality. **概念**: Jensen-Shannon散度是对称的(JS(P||Q) = JS(Q||P))，对于二进制对数的值在[0, 1]中。JS距离 = √JS散度提供满足三角形不等式的真正度量。

- **ML Application**: JS divergence appears naturally in GAN training as it avoids mode coverage issues of KL divergence. It's also used in clustering and distribution matching tasks. **机器学习应用**: JS散度自然出现在GAN训练中，因为它避免了KL散度的模式覆盖问题。它也用于聚类和分布匹配任务。
