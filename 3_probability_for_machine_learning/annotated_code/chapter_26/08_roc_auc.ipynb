## Learning Notes / 学习笔记

- **Concept**: AUC = P(model ranks positive example higher than negative example). It ranges [0,1]: 0.5=random, 1.0=perfect, <0.5=worse than random. **概念**: AUC = P(模型对正例的排名高于负例)。范围[0,1]: 0.5=随机, 1.0=完美, <0.5=差于随机。

- **ML Application**: AUC is a ranking metric popular for comparing classifiers, especially on imbalanced data. It's invariant to threshold selection, making it a universal performance measure. **机器学习应用**: AUC是用于比较分类器的流行排名度量，特别是在不平衡数据上。它对阈值选择是不变的，使其成为通用的性能度量。
