## Summary / 汇总

This notebook implements a classifier that randomly selects a class from the true class distribution. It achieves better performance than 50/50 random but ignores features.

本笔记本实现了一个从真实类分布中随机选择类的分类器。它的性能优于50/50随机，但忽略了特征。

## Step 1 — Randomly Selected Class from Distribution / 从分布中随机选择的类

```python
from numpy import meanfrom numpy.random import randintfrom sklearn.metrics import accuracy_score# predict a randomly selected class from the training distributiondef random_class(y):    return y[randint(len(y))]# define datasetclass0 = [0 for _ in range(25)]class1 = [1 for _ in range(75)]y = class0 + class1# average over many repeatsresults = list()for _ in range(1000):    yhat = [random_class(y) for _ in range(len(y))]    acc = accuracy_score(y, yhat)    results.append(acc)print('Mean: %.3f' % mean(results))
```

## Learning Notes / 学习笔记

- **Concept**: This classifier matches the true class distribution by random sampling from training labels. It achieves ~58% accuracy on average, weighted toward the majority class. **概念**: 该分类器通过从训练标签中随机采样来匹配真实类分布。它平均达到约58%的准确率，向多数类加权。

- **ML Application**: This is still a naive baseline but respects class distribution. It's better than 50/50 random but worse than always predicting the majority class. **机器学习应用**: 这仍然是一个朴素的基线，但尊重类分布。它比50/50随机更好，但不如总是预测多数类更好。
