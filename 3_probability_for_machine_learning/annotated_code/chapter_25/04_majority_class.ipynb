## Summary / 汇总

This notebook implements a majority class classifier using scipy.stats.mode. It achieves 75% accuracy by always predicting the most frequent class.

本笔记本使用scipy.stats.mode实现多数类分类器。通过总是预测最频繁的类，它达到75%的准确率。

## Step 1 — Majority Class Classifier / 多数类分类器

```python
from scipy.stats import modefrom sklearn.metrics import accuracy_score# predict the majority class: always return the modedef majority_class(y):    return mode(y)[0]# define datasetclass0 = [0 for _ in range(25)]class1 = [1 for _ in range(75)]y = class0 + class1# make predictions: same class for all samplesyhat = [majority_class(y) for _ in range(len(y))]# calculate accuracyaccuracy = accuracy_score(y, yhat)print('Accuracy: %.3f' % accuracy)
```

## Learning Notes / 学习笔记

- **Concept**: The majority class classifier always predicts the most frequent class (class 1 in 75%-25% split). It achieves accuracy equal to the majority class proportion. **概念**: 多数类分类器总是预测最频繁的类(75%-25%分割中的1类)。它达到等于多数类比例的准确率。

- **ML Application**: This is the strongest naive baseline. Trained models must outperform this to demonstrate feature learning. It's the default accuracy on imbalanced datasets without proper handling. **机器学习应用**: 这是最强的朴素基线。训练模型必须超越这个以演示特征学习。它是在没有适当处理的不平衡数据集上的默认准确率。
