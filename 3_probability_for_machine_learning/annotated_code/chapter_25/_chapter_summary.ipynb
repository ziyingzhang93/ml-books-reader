## Why Naive Classifiers Matter

### 1. Diagnostic Tool
If your sophisticated model matches naive classifier, something is wrong:
- Not training properly?
- Features not informative?
- Implementation bug?

### 2. Context for Results
"93% accuracy" sounds great until you learn:
- Majority class is 92%
- Your model barely improves
- Naive classifier would suffice

### 3. Different Metrics Reveal Truth
- Accuracy: Majority class looks good
- Precision: Shows model fails on minority
- Recall: Reveals model can't find minority class
- ROC-AUC: All naive classifiers score 0.50

## Scikit-learn DummyClassifier

### Strategies
```python
from sklearn.dummy import DummyClassifier

# Strategy 1: Most frequent class
clf = DummyClassifier(strategy='most_frequent')

# Strategy 2: Class distribution
clf = DummyClassifier(strategy='stratified')

# Strategy 3: Uniform random
clf = DummyClassifier(strategy='uniform')

# Strategy 4: Always specific constant
clf = DummyClassifier(strategy='constant', constant=1)
```

## When to Use Naive Classifiers

### Always
1. **Establish baseline**: Run DummyClassifier first
2. **Check improvements**: Is real model better?
3. **Imbalanced data**: Majority class baseline critical

### Specific Cases
- **Random Guess**: Truly worst case, rarely practical baseline
- **Random Class**: Good for probability calibration checks
- **Majority Class**: Standard baseline, always include

## Handling Class Imbalance

Naive classifiers reveal imbalance problem:
- 95% accuracy but 5% minority class?
- Majority class achieves 95% by ignoring minority

### Solutions
1. **Stratified Evaluation**: Use minority class frequency as baseline
2. **Weighted Metrics**: Precision/recall/F1 on minority class
3. **ROC-AUC**: Robust to imbalance
4. **Resampling**: Oversample minority / undersample majority
5. **Threshold Adjustment**: Change decision boundary

## Key Takeaways
1. Always establish naive baselines
2. Majority class baseline is critical for imbalanced data
3. High accuracy can be misleading without right baseline
4. Different metrics reveal different aspects of performance
5. ROC-AUC = 0.50 for all naive classifiers (good sanity check)
6. Baseline defines minimum standard for real models
7. Use DummyClassifier for reproducible baselines
