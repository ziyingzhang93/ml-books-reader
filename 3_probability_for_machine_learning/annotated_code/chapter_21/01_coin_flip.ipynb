## Summary / 汇总

This notebook calculates the information content of a coin flip outcome. Information h = -log2(p) quantifies the surprise or uncertainty reduction from observing an event.

本笔记本计算硬币翻转结果的信息内容。信息h = -log2(p)量化观察事件的惊讶度或不确定性减少。

## Step 1 — Calculate Information for Coin Flip / 计算硬币翻转的信息

```python
from math import log2# probability of the event (fair coin)p = 0.5# calculate information for event: h(x) = -log2(p)# For p=0.5, each outcome carries 1 bit of informationh = -log2(p)# print the resultprint('p(x)=%.3f, information: %.3f bits' % (p, h))
```

## Learning Notes / 学习笔记

- **Concept**: Information h = -log₂(p) measures how many bits are needed to encode an outcome. Lower probability events carry more information because they're more surprising. **概念**: 信息h = -log₂(p)衡量编码结果所需的比特数。低概率事件携带更多信息，因为它们更令人惊讶。

- **ML Application**: Information theory is the foundation for entropy-based decision trees (ID3, C4.5) and information gain calculations used in feature selection. **机器学习应用**: 信息论是基于熵的决策树(ID3, C4.5)和特征选择中信息增益计算的基础。
