# Chapter 21: Information Theory

## Overview
This chapter introduces **Information Theory**, a mathematical framework for quantifying uncertainty and information. The journey progresses from information of single events to expected information (entropy).

## Key Concepts
- **Information (Surprisal)**: How surprising/informative is an outcome?
- **Entropy**: Expected information across all outcomes
- **Bits vs Nats**: Different units (log₂ vs ln)
- **Maximum Entropy**: Uniform distribution has highest uncertainty
- **Minimum Entropy**: Deterministic events have zero uncertainty

## Information (Surprisal)

### Definition
Information content of an outcome with probability p:
```
I(x) = -log₂(p(x))   [bits]
I(x) = -ln(p(x))     [nats]
```

### Intuition
- High probability event (p=1): I = 0 bits (no surprise, no information)
- Low probability event (p=0.5): I = 1 bit
- Very rare event (p→0): I → ∞ bits

### Example: Coin Flip
```
Fair coin: p(H) = 0.5, I(H) = -log₂(0.5) = 1 bit
Biased coin: p(H) = 0.9, I(H) = -log₂(0.9) ≈ 0.15 bits
```
The biased coin outcome is less surprising (less information).

## Entropy

### Definition
Expected information across all outcomes:
```
H(X) = -Σₓ p(x) × log₂(p(x))
```

### Intuition
- Entropy measures **average uncertainty** in distribution
- Uniform distribution: Maximum entropy (most uncertain)
- Deterministic (one outcome): Entropy = 0 (no uncertainty)
- Binary entropy H(p): Highest at p=0.5

### Example: Die Roll
Fair 6-sided die:
```
H(die) = -6 × (1/6 × log₂(1/6))
       = log₂(6)
       ≈ 2.585 bits
```
Can encode outcome in 2.585 bits on average.

### Example: Biased Die
Die with p(6)=0.9, p(other)=0.02:
```
H(die) ≈ 0.47 bits
```
Much lower entropy: outcome is predictable (almost always 6).

## Properties

### Range
```
0 ≤ H(X) ≤ log₂(k)
```
- k: number of outcomes
- Minimum at deterministic distribution
- Maximum at uniform distribution

### Binary Entropy
```
H(p) = -p × log₂(p) - (1-p) × log₂(1-p)
```
- H(0) = 0 (certain, no uncertainty)
- H(0.5) = 1 (maximum, fair coin)
- H(1) = 0 (certain)

## Connection to Data Compression

### Shannon's Source Coding Theorem
Entropy H(X) is the **average minimum bits** needed to encode symbols from X.

Example: Encode die outcomes
- Naive: 3 bits (need to distinguish 6 outcomes: 2³ = 8)
- Optimal (Huffman): ~2.585 bits average

## Information vs Probability

### Inverse Relationship
```
I(x) = -log(p(x))
```
- More probable → less information
- Less probable → more information
- Makes intuitive sense: expected events carry no information

## Key Takeaways
1. Information quantifies surprise: rare events are informative
2. Entropy is expected information: average uncertainty in distribution
3. Uniform distribution has maximum entropy
4. Deterministic outcomes have zero entropy
5. Entropy connects to data compression: H(X) bits average needed per symbol
6. Foundation for divergence measures (KL, JS) and cross-entropy
7. Used in: decision trees (information gain), deep learning (loss functions)
