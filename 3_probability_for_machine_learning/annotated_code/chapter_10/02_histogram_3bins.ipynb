## Summary

We create histograms with 3, 10, and 30 bins from the same data, demonstrating the resolution tradeoff in density estimation. Fewer bins lose detail, while more bins add noise.

我们从相同数据创建3、10和30个分箱的直方图，演示密度估计中的分辨率权衡。分箱少会丢失细节，分箱多会增加噪声。

## Step 1 — Import and Generate Data / 导入并生成数据

```python
# Import necessary libraries / 导入必要的库
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate data / 生成数据
np.random.seed(42)
mu = 50
sigma = 5
num_samples = 1000
samples = np.random.normal(loc=mu, scale=sigma, size=num_samples)
```

## Step 2 — Compare Histograms with Different Bin Counts / 比较不同分箱数的直方图

```python
# Create subplots for comparison / 创建子图进行比较
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
bin_counts = [3, 10, 30]

# True distribution / 真实分布
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
true_pdf = norm.pdf(x, loc=mu, scale=sigma)

# Create histograms with different bin counts / 创建不同分箱数的直方图
for idx, num_bins in enumerate(bin_counts):
    ax = axes[idx]
    n, bins, _ = ax.hist(samples, bins=num_bins, density=True, alpha=0.7, 
                         edgecolor='black', color='steelblue', linewidth=1.5)
    # Overlay true PDF / 覆盖真实PDF
    ax.plot(x, true_pdf, 'r-', linewidth=2, label='True PDF')
    ax.set_xlabel('Value / 值')
    ax.set_ylabel('Density / 密度')
    ax.set_title(f'Histogram with {num_bins} Bins')
    ax.legend()
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

## Step 3 — Analyze Bin Resolution Effects / 分析分箱分辨率效果

```python
# Calculate MSE for different bin counts / 计算不同分箱数的MSE
print('Bin Resolution Comparison / 分箱分辨率比较')
print('=' * 70)
print(f'{"Num Bins":>10s} {"Bin Width":>15s} {"MSE":>15s} {"Interpretation":>30s}')
print('-' * 70)

for num_bins in [3, 5, 10, 15, 30]:
    n, bins = np.histogram(samples, bins=num_bins, density=True)
    bin_centers = (bins[:-1] + bins[1:]) / 2
    bin_width = bins[1] - bins[0]
    
    histogram_density = n
    true_density = norm.pdf(bin_centers, loc=mu, scale=sigma)
    mse = np.mean((histogram_density - true_density)**2)
    
    interpretation = "Too coarse" if num_bins < 8 else ("Balanced" if num_bins < 20 else "Too noisy")
    print(f'{num_bins:10d} {bin_width:15.3f} {mse:15.6f} {interpretation:>30s}')
```

## Learning Notes / 学习笔记

- **Concept**: The binning choice (bias-variance tradeoff) directly affects estimation quality. Too few bins oversimplify the distribution (high bias), while too many bins create noise (high variance). The optimal choice depends on sample size and data complexity.
  
  **概念**：分箱选择（偏差-方差权衡）直接影响估计质量。分箱太少会过度简化分布（高偏差），分箱太多会产生噪声（高方差）。最优选择取决于样本大小和数据复杂性。

- **ML Application**: Choosing appropriate histogram resolution is crucial for initial EDA. Methods like Freedman-Diaconis, Sturges' rule, or cross-validation provide guidelines for selecting optimal bin width.
  
  **机器学习应用**：为初始EDA选择合适的直方图分辨率至关重要。Freedman-Diaconis、Sturges规则或交叉验证等方法为选择最优分箱宽度提供指导。
