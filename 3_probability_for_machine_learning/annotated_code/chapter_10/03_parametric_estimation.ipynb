## Summary

We fit a normal distribution to sample data by estimating its parameters (mean and standard deviation), then overlay the fitted PDF on a histogram. This parametric approach assumes a known distributional form.

我们通过估计正态分布的参数（均值和标准差）来拟合样本数据，然后将拟合的PDF覆盖在直方图上。这种参数方法假设已知的分布形式。

## Step 1 — Import and Generate Data / 导入并生成数据

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

np.random.seed(42)
mu_true = 50
sigma_true = 5
num_samples = 1000
samples = np.random.normal(loc=mu_true, scale=sigma_true, size=num_samples)
```

## Step 2 — Estimate Parameters / 估计参数

```python
# Estimate mean and standard deviation / 估计均值和标准差
mu_est = np.mean(samples)
sigma_est = np.std(samples, ddof=1)  # Use sample std (ddof=1)

print(f'Parameter Estimation / 参数估计')
print(f'True Parameters / 真实参数:')
print(f'  μ = {mu_true}, σ = {sigma_true}')
print(f'\nEstimated Parameters / 估计参数:')
print(f'  μ̂ = {mu_est:.4f}, σ̂ = {sigma_est:.4f}')
print(f'\nEstimation Errors / 估计误差:')
print(f'  Δμ = {abs(mu_est - mu_true):.4f}')
print(f'  Δσ = {abs(sigma_est - sigma_true):.4f}')
```

## Step 3 — Plot with Fitted PDF / 绘制拟合的PDF

```python
fig, ax = plt.subplots(figsize=(12, 6))

# Histogram / 直方图
ax.hist(samples, bins=20, density=True, alpha=0.6, color='steelblue', 
        edgecolor='black', label='Histogram')

# True PDF / 真实PDF
x = np.linspace(mu_est - 4*sigma_est, mu_est + 4*sigma_est, 1000)
true_pdf = norm.pdf(x, loc=mu_true, scale=sigma_true)
ax.plot(x, true_pdf, 'g-', linewidth=2.5, label=f'True N(μ={mu_true}, σ={sigma_true})')

# Fitted PDF / 拟合PDF
fitted_pdf = norm.pdf(x, loc=mu_est, scale=sigma_est)
ax.plot(x, fitted_pdf, 'r--', linewidth=2.5, label=f'Fitted N(μ={mu_est:.2f}, σ={sigma_est:.2f})')

ax.set_xlabel('Value / 值')
ax.set_ylabel('Density / 密度')
ax.set_title('Parametric Density Estimation: Normal Distribution')
ax.legend()
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

## Learning Notes / 学习笔记

- **Concept**: Parametric estimation assumes a known distributional form and estimates only its parameters. This is efficient when the assumption is correct but can be biased if the true distribution differs significantly from the assumed form.

- **ML Application**: Parametric methods are widely used for computational efficiency and theoretical interpretability. Common examples include Gaussian Naive Bayes, linear regression residuals, and Gaussian mixture models.
