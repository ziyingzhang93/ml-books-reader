# Chapter 16: Bayes Theorem

## Overview
This chapter explores **Bayes Theorem**, the mathematical foundation for updating beliefs with evidence. The journey progresses from simple to increasingly complex real-world applications.

## Core Equation
```
P(A | B) = P(B | A) × P(A) / P(B)
```
- **P(A | B)**: Posterior probability (updated belief)
- **P(B | A)**: Likelihood (probability of evidence given hypothesis)
- **P(A)**: Prior probability (initial belief)
- **P(B)**: Evidence (probability of observing this data)

## Key Insights

### Why Bayes?
Most natural human reasoning is **forward**: given a hypothesis, what evidence would I expect?

But we often need **inverse**: given evidence, what's the probability of the hypothesis?

Bayes theorem bridges forward reasoning (likelihood) to inverse reasoning (posterior).

### Example: Medical Testing
- **Hypothesis A**: Patient has disease
- **Evidence B**: Positive test result
- **Prior P(A)**: Base rate of disease in population
- **Likelihood P(B | A)**: Test sensitivity (correctly identifies disease)
- **Posterior P(A | B)**: Probability patient has disease given positive test

## The Three Forms

### Form 1: Basic Bayes
```
P(A | B) = P(B | A) × P(A) / P(B)
```
When there are two hypotheses (A or not A).

### Form 2: Total Probability Expansion
```
P(A | B) = P(B | A) × P(A) / [P(B | A) × P(A) + P(B | ¬A) × P(¬A)]
```
Expands denominator using law of total probability.
More explicit about considering alternative hypotheses.

### Form 3: Multiple Hypotheses
```
P(Aᵢ | B) = P(B | Aᵢ) × P(Aᵢ) / Σⱼ P(B | Aⱼ) × P(Aⱼ)
```
Generalizes to k > 2 hypotheses.

## Common Applications

### Medical Diagnosis
- Combine test result with base rate of disease
- Account for false positive rate

### Spam Detection
- Combine word frequencies (likelihoods)
- With prior on spam rate

### Fraud Detection
- Combine behavioral signals (likelihood)
- With fraud base rate

### Machine Learning
- Foundation for Naive Bayes classifier
- Foundation for Bayesian inference

## Critical Point: Base Rate

Bayes theorem forces us to consider the prior P(A).
This is critical: even with a perfect test, if the disease is rare, a positive test is still more likely to be false positive.

**Example**: Disease affects 0.1% of population, test is 99% accurate
- Prior P(disease) = 0.001
- P(test | disease) = 0.99 (sensitivity)
- P(test | no disease) = 0.01 (false positive rate)
- Posterior P(disease | test) = 0.0901 (about 9%)

The prior (rare disease) dominates even the good test accuracy!

## Key Takeaways
1. Bayes theorem formalizes how to update beliefs with evidence
2. Prior + Likelihood → Posterior
3. The prior (base rate) is crucial and cannot be ignored
4. Total probability expansion shows all competing hypotheses
5. Bayes is the foundation for probabilistic reasoning in machine learning
