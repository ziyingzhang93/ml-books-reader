## The Naive Bayes Classifier

### Bayes Rule for Classification
```
P(y | x₁, x₂, ..., xₐ) = P(x₁, x₂, ..., xₐ | y) × P(y) / P(x₁, x₂, ..., xₐ)
```
- y: class label (0 or 1)
- x₁, x₂, ..., xₐ: features
- P(y): prior (class frequency)
- P(x₁, x₂, ..., xₐ | y): likelihood (hard to estimate for many features)

### Naive Independence Assumption
Features are **conditionally independent** given the class:
```
P(x₁, x₂, ..., xₐ | y) = P(x₁ | y) × P(x₂ | y) × ... × P(xₐ | y)
```

This is rarely true in practice but works remarkably well!

### Simplified Posterior
```
P(y | x₁, x₂, ..., xₐ) ∝ P(y) × ∏ᵢ P(xᵢ | y)
```
Denominator is constant across classes, so we can ignore it for classification.

## Gaussian Naive Bayes

### Likelihood Assumption
Each feature is normally distributed within each class:
```
P(xᵢ | y) = N(xᵢ | μᵢ,y, σᵢ,y²)
```

### Learning Parameters
From training data, estimate for each feature i and class y:
- μᵢ,y: mean of feature i in class y
- σᵢ,y²: variance of feature i in class y
- P(y): proportion of training samples in class y

### Prediction
```
ŷ = argmax_y P(y) × ∏ᵢ N(xᵢ | μᵢ,y, σᵢ,y²)
```
Choose class with highest posterior probability.

## Why Naive Bayes Works

### The Paradox
- The naive assumption is almost never true in real data
- Yet Naive Bayes often performs surprisingly well
- Sometimes outperforms more sophisticated methods

### Why?
1. **Probabilistic calibration matters more than ranking**: Classifier mainly needs correct ranking of classes
2. **Error cancellation**: Errors from wrong assumption can cancel out
3. **High-dimensional benefit**: With many features, independence can approximate marginal distributions
4. **Low variance estimation**: Simplifying assumptions reduce overfitting

## Advantages
1. **Simple**: Easy to understand and implement
2. **Fast**: Linear time in number of features and training samples
3. **Works with small data**: Doesn't require large datasets
4. **Interpretable**: Can examine feature distributions per class
5. **Probabilistic**: Outputs class probabilities, not just predictions

## Limitations
1. **Naive assumption**: Often violated in practice
2. **Zero likelihood problem**: Unseen feature values cause zero probability
3. **Feature scaling**: Variance estimates can be poor with few samples
4. **Discrete features**: Gaussian assumption may not fit all feature types

## Variants
- **GaussianNB**: For continuous features (assumes Gaussian)
- **MultinomialNB**: For discrete counts (e.g., word frequencies in text)
- **BernoulliNB**: For binary features
- **CategoricalNB**: For categorical features

## Key Takeaways
1. Naive Bayes applies Bayes theorem to classification via independence assumption
2. Despite the naive assumption, it often works well in practice
3. Learning amounts to computing class-conditional feature statistics
4. Prediction is fast: just compute products of likelihoods
5. Good baseline for text classification, email filtering, and other discrete tasks
6. Use scikit-learn's implementation for production use
