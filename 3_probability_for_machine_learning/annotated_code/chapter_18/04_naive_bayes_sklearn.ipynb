## Learning Notes / 学习笔记

- **Concept**: GaussianNB automatically learns the mean and variance for each feature per class. predict_proba() returns class probabilities, while predict() returns the most likely class label. **概念**: GaussianNB自动学习每个类的每个特征的均值和方差。predict_proba()返回类概率，而predict()返回最可能的类标签。

- **ML Application**: Scikit-learn's implementation is optimized and handles numerical stability. It's ideal for quick prototyping and production use where the naive independence assumption is acceptable. **机器学习应用**: Scikit-learn的实现是优化的，并处理数值稳定性。它非常适合快速原型制作和生产使用，其中朴素独立性假设是可接受的。
