# Chapter 13: Odds, Log-Odds, and Likelihood

## Overview
This chapter explores three related but distinct ways to represent **uncertainty and probability**. These representations are foundational for logistic regression and classification.

## Key Concepts
- **Probability**: P(event) ∈ [0, 1], ranges from 0 to 1
- **Odds**: P(event) / P(not event), unbounded [0, ∞)
- **Log-Odds (Logit)**: log(odds), ranges from -∞ to +∞, linear scale
- **Likelihood**: Probability of data given a parameter, not a probability distribution

## Representation Equivalences

### Relationship Between Forms
Given probability p:
- **Odds** = p / (1 - p)
- **Log-Odds** = log(p / (1 - p)) = logit(p)
- **Probability** (inverse) = exp(log_odds) / (1 + exp(log_odds)) = sigmoid(log_odds)

### Why Each Form?

#### Probability (p)
- Intuitive and interpretable (0% to 100%)
- Natural for decision-making ("80% likely to rain")
- Non-linear relationship to predictors (saturates)

#### Odds
- Betting language ("3:1 odds")
- Easier to multiply (combine independent events)
- Still bounded

#### Log-Odds (Logit)
- **Linear scale**: additive relationship to predictors
- Symmetric around 0 (log_odds(0.5) = 0)
- Essential for logistic regression
- Unbounded: no saturation effects

## Why Log-Odds Matter for ML

**Linear Regression**: outcome = β₀ + β₁x

**Logistic Regression**: log-odds(P) = β₀ + β₁x

Log-odds allow probability to have a **linear relationship** with predictors while staying valid [0,1].

## Likelihood
- Not a probability: doesn't integrate to 1
- P(data | θ): probability of observed data given a parameter value
- Essential for: maximum likelihood estimation, model comparison
- Bernoulli likelihood: L(p) = p^k × (1-p)^(n-k) for k successes in n trials

## Key Takeaways
1. Probability, odds, and log-odds are three representations of the same uncertainty
2. Log-odds enable linear relationships in classification models
3. Log-odds are symmetric around 0 (p=0.5)
4. Likelihood is distinct from probability: it's P(data | parameter), not P(parameter | data)
5. These concepts bridge from basic probability to statistical inference and machine learning
