## Learning Notes / 学习笔记

- **Concept**: CalibratedClassifierCV fits a calibration model (sigmoid or isotonic) that maps raw outputs to [0,1]. The method='sigmoid' fits a logistic regression; method='isotonic' fits a non-parametric mapping. **概念**: CalibratedClassifierCV拟合校准模型(sigmoid或isotonic)，将原始输出映射到[0,1]。method='sigmoid'拟合逻辑回归；method='isotonic'拟合非参数映射。

- **ML Application**: Calibration is crucial for probabilistic predictions in production. Models with proper calibration enable confidence-aware decision thresholds and uncertainty quantification. **机器学习应用**: 校准对于生产中的概率预测至关重要。具有适当校准的模型能够实现与置信度相关的决策阈值和不确定性量化。
