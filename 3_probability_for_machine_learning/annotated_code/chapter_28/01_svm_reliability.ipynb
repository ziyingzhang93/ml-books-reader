This notebook plots a reliability diagram for SVM without calibration. The plot shows how well predicted probabilities match observed frequencies.

## Step 1 — Plot Uncalibrated SVM Reliability / 绘制未校准的SVM可靠性

```python
from sklearn.datasets import make_classificationfrom sklearn.svm import SVCfrom sklearn.model_selection import train_test_splitfrom sklearn.calibration import calibration_curvefrom matplotlib import pyplot# generate 2 class datasetX, y = make_classification(n_samples=1000, n_classes=2, weights=[1,1], random_state=1)# split into train/test setstrainX, testX, trainy, testy = train_test_split(X, y, test_size=0.5, random_state=2)# fit a model (SVM doesn't have probability outputs by default)model = SVC(gamma='scale')model.fit(trainX, trainy)# predict decision function (not probabilities)probs = model.decision_function(testX)# reliability diagram: fraction of positives vs mean predicted valuefop, mpv = calibration_curve(testy, probs, n_bins=10, normalize=True)# plot perfectly calibrated (diagonal line)pyplot.plot([0, 1], [0, 1], linestyle='--')# plot model reliabilitypyplot.plot(mpv, fop, marker='.')pyplot.show()
```

## Learning Notes / 学习笔记

- **Concept**: Reliability diagrams plot predicted probability (x-axis) against observed frequency of positive class (y-axis). Diagonal=perfect calibration. Above diagonal=overconfident, below=underconfident. **概念**: 可靠性图绘制预测概率(x轴)对正类观察频率(y轴)。对角线=完美校准。对角线上方=过度自信，下方=不自信。

- **ML Application**: SVM's decision function requires calibration because it's unbounded and doesn't represent true probabilities. CalibratedClassifierCV applies sigmoid transformation to fix this. **机器学习应用**: SVM的决策函数需要校准，因为它是无界的，不代表真实概率。CalibratedClassifierCV应用sigmoid变换来修复这个问题。
