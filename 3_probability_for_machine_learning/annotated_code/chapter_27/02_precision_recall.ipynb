## Learning Notes / 学习笔记

- **Concept**: Precision-Recall curves show precision (TP/(TP+FP)) vs recall (TP/(TP+FN)). For imbalanced data, PR curves are more informative than ROC because they focus on the positive class. **概念**: 精确率-召回率曲线显示精确率(TP/(TP+FP))对召回率(TP/(TP+FN))。对于不平衡数据，PR曲线比ROC更有信息量，因为它们关注正类。

- **ML Application**: PR curves reveal how models handle the minority class. A model perfect on majority class but poor on minority class will have high ROC-AUC but low PR-AUC, making PR-AUC a better metric for imbalanced problems. **机器学习应用**: PR曲线揭示模型如何处理少数类。在多数类上完美但在少数类上表现差的模型将具有高ROC-AUC但低PR-AUC，使PR-AUC成为不平衡问题的更好指标。
