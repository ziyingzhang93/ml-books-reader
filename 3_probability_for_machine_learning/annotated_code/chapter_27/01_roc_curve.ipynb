## Learning Notes / 学习笔记

- **Concept**: Comparing classifiers via ROC curves shows discriminative power. No-skill classifier (constant predictions) produces the diagonal. Better models curve toward top-left. **概念**: 通过ROC曲线比较分类器显示了判别能力。无技能分类器(恒定预测)产生对角线。更好的模型向左上方弯曲。

- **ML Application**: ROC curves provide a threshold-independent visualization for model comparison, especially useful for imbalanced problems where accuracy alone misleads. **机器学习应用**: ROC曲线为模型比较提供了与阈值无关的可视化，特别是对于准确率单独会误导的不平衡问题很有用。
