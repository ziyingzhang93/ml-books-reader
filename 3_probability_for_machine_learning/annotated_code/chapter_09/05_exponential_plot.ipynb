## Summary

We plot the PDF and CDF of the exponential distribution using scipy.stats.expon to visualize its properties and show how it differs from the normal distribution.

我们使用scipy.stats.expon绘制指数分布的PDF和CDF，以可视化其性质并展示它与正态分布的区别。

## Step 1 — Import Libraries / 导入库

```python
# Import necessary libraries / 导入必要的库
from scipy.stats import expon
import numpy as np
import matplotlib.pyplot as plt
```

## Step 2 — Define Parameters / 定义参数

```python
# Define exponential distribution parameter / 定义指数分布参数
scale = 2.0  # Mean / 均值
```

## Step 3 — Plot PDF and CDF / 绘制PDF和CDF

```python
# Create x values / 创建x值
x = np.linspace(0, 4*scale, 1000)

# Calculate PDF and CDF / 计算PDF和CDF
pdf = expon.pdf(x, scale=scale)
cdf = expon.cdf(x, scale=scale)

# Create figure with two subplots / 创建两个子图
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot PDF / 绘制PDF
axes[0].plot(x, pdf, 'b-', linewidth=2, label=f'PDF: Exp(scale={scale})')
axes[0].fill_between(x, pdf, alpha=0.3)
axes[0].axvline(scale, color='r', linestyle='--', linewidth=2, label=f'Mean: {scale}')
axes[0].set_xlabel('Value / 值')
axes[0].set_ylabel('Probability Density / 概率密度')
axes[0].set_title('Exponential Distribution PDF')
axes[0].legend()
axes[0].grid(alpha=0.3)

# Plot CDF / 绘制CDF
axes[1].plot(x, cdf, 'g-', linewidth=2, label=f'CDF: Exp(scale={scale})')
axes[1].axhline(0.5, color='r', linestyle='--', alpha=0.5, linewidth=1)
axes[1].axvline(scale, color='r', linestyle='--', linewidth=2, label=f'Mean: {scale}')
axes[1].set_xlabel('Value / 值')
axes[1].set_ylabel('Cumulative Probability / 累积概率')
axes[1].set_title('Exponential Distribution CDF')
axes[1].legend()
axes[1].grid(alpha=0.3)
axes[1].set_ylim([0, 1.05])

plt.tight_layout()
plt.show()
```

## Step 4 — Compare with Normal / 与正态分布比较

```python
# Compare exponential with normal distribution / 比较指数分布和正态分布
from scipy.stats import norm

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

x = np.linspace(0, 10, 1000)

# PDF comparison / PDF比较
exp_pdf = expon.pdf(x, scale=2)
norm_pdf = norm.pdf(x, loc=2, scale=2)

axes[0].plot(x, exp_pdf, 'b-', linewidth=2, label='Exponential')
axes[0].plot(x, norm_pdf, 'r-', linewidth=2, label='Normal')
axes[0].set_xlabel('Value / 值')
axes[0].set_ylabel('Probability Density / 概率密度')
axes[0].set_title('PDF Comparison')
axes[0].legend()
axes[0].grid(alpha=0.3)

# CDF comparison / CDF比较
exp_cdf = expon.cdf(x, scale=2)
norm_cdf = norm.cdf(x, loc=2, scale=2)

axes[1].plot(x, exp_cdf, 'b-', linewidth=2, label='Exponential')
axes[1].plot(x, norm_cdf, 'r-', linewidth=2, label='Normal')
axes[1].set_xlabel('Value / 值')
axes[1].set_ylabel('Cumulative Probability / 累积概率')
axes[1].set_title('CDF Comparison')
axes[1].legend()
axes[1].grid(alpha=0.3)
axes[1].set_ylim([0, 1.05])

plt.tight_layout()
plt.show()
```

## Learning Notes / 学习笔记

- **Concept**: The exponential distribution's PDF is a decreasing function, meaning values close to zero are most likely. This is fundamentally different from the normal distribution's bell shape. The long right tail reflects rare but possible large values.
  
  **概念**：指数分布的PDF是递减函数，意味着接近零的值最可能。这与正态分布的钟形完全不同。长右尾反映了罕见但可能的大值。

- **ML Application**: Exponential PDF and CDF are used in survival models, predicting event times, and risk assessment in various domains including failure rates and customer churn prediction.
  
  **机器学习应用**：指数PDF和CDF用于生存模型、预测事件时间和各种领域的风险评估，包括失败率和客户流失预测。
