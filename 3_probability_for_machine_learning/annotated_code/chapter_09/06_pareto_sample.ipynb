## Summary

We sample from a Pareto distribution using scipy.stats.pareto. The Pareto distribution models the distribution of wealth, company sizes, and other phenomena following the 80/20 principle (Pareto principle).

我们使用scipy.stats.pareto从帕累托分布中抽样。帕累托分布模拟财富分布、公司规模和遵循80/20原则（帕累托原则）的其他现象。

**Formula:**

$$f(x) = \frac{a m^a}{x^{a+1}}, \quad x \geq m$$

where $a$ is the shape parameter and $m$ is the scale parameter

## Step 1 — Import Libraries / 导入库

```python
# Import necessary libraries / 导入必要的库
from scipy.stats import pareto
import numpy as np
import matplotlib.pyplot as plt
```

## Step 2 — Set Parameters / 设置参数

```python
# Define Pareto distribution parameters / 定义帕累托分布参数
a = 1.5          # Shape parameter (Pareto index) / 形状参数
scale = 1.0      # Minimum value / 最小值
num_samples = 10000  # Number of samples / 样本数
```

## Step 3 — Generate Pareto Samples / 生成帕累托样本

```python
# Sample from Pareto distribution / 从帕累托分布中抽样
samples = pareto.rvs(b=a, scale=scale, size=num_samples)

print(f'Pareto Distribution Sample / 帕累托分布样本')
print(f'Shape parameter (a) / 形状参数: {a}')
print(f'Scale parameter (m) / 尺度参数: {scale}')
print(f'Sample count / 样本数: {num_samples}')
print(f'\nSample statistics / 样本统计:')
print(f'Mean / 均值: {np.mean(samples):.4f}')
print(f'Std Dev / 标准差: {np.std(samples):.4f}')
print(f'Min / 最小值: {np.min(samples):.4f}')
print(f'Max / 最大值: {np.max(samples):.4f}')
print(f'Median / 中位数: {np.median(samples):.4f}')
```

## Step 4 — Visualize with Different Scales / 用不同尺度可视化

```python
# Create visualization / 创建可视化
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Linear scale / 线性尺度
axes[0].hist(samples[samples < 20], bins=50, density=True, alpha=0.7, edgecolor='black')
axes[0].set_xlabel('Value / 值')
axes[0].set_ylabel('Density / 密度')
axes[0].set_title(f'Pareto Distribution (a={a}, Linear Scale) - Showing values < 20')
axes[0].grid(alpha=0.3)

# Log-log scale (to show power law) / 双对数尺度
axes[1].hist(samples, bins=50, density=True, alpha=0.7, edgecolor='black')
axes[1].set_xscale('log')
axes[1].set_yscale('log')
axes[1].set_xlabel('Value (log scale) / 值（对数尺度）')
axes[1].set_ylabel('Density (log scale) / 密度（对数尺度）')
axes[1].set_title(f'Pareto Distribution (a={a}, Log-Log Scale)')
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

## Step 5 — Plot Theoretical PDF / 绘制理论PDF

```python
# Plot theoretical PDF / 绘制理论PDF
fig, ax = plt.subplots(figsize=(12, 6))

# Theoretical PDF / 理论PDF
x = np.linspace(scale, 10, 1000)
pdf = pareto.pdf(x, b=a, scale=scale)

# Plot histogram and PDF / 绘制直方图和PDF
ax.hist(samples[samples < 10], bins=50, density=True, alpha=0.6, edgecolor='black', label='Sample histogram')
ax.plot(x, pdf, 'r-', linewidth=2, label='Theoretical PDF')

ax.set_xlabel('Value / 值')
ax.set_ylabel('Probability Density / 概率密度')
ax.set_title(f'Pareto Distribution vs Sample (a={a})')
ax.legend()
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

## Step 6 — Compare Different Shape Parameters / 比较不同的形状参数

```python
# Plot Pareto distributions with different shape parameters / 绘制不同形状参数的帕累托分布
fig, ax = plt.subplots(figsize=(12, 6))

shape_params = [0.5, 1.0, 1.5, 2.5]
x = np.linspace(1, 5, 1000)

for a_test in shape_params:
    pdf = pareto.pdf(x, b=a_test, scale=1.0)
    ax.plot(x, pdf, linewidth=2, label=f'a={a_test}')

ax.set_xlabel('Value / 值')
ax.set_ylabel('Probability Density / 概率密度')
ax.set_title('Pareto Distributions with Different Shape Parameters')
ax.legend()
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

## Learning Notes / 学习笔记

- **Concept**: The Pareto distribution is a heavy-tailed distribution where a small number of items account for most of the value. The shape parameter $a$ controls how heavy the tail is: smaller $a$ means heavier tail and more extreme values.
  
  **概念**：帕累托分布是重尾分布，其中少数项占大部分价值。形状参数$a$控制尾部有多重：较小的$a$表示尾部更重，极端值更多。

- **ML Application**: Pareto distribution models wealth distribution, company sizes, website traffic, and other phenomena with 80/20 characteristics. Used in risk modeling, anomaly detection, and understanding business metrics.
  
  **机器学习应用**：帕累托分布模拟财富分布、公司规模、网站流量和其他具有80/20特征的现象。用于风险建模、异常检测和理解业务指标。

```python
# Complete Pareto Distribution Sampling / 完整帕累托分布抽样

from scipy.stats import pareto
import numpy as np
import matplotlib.pyplot as plt

# Define Pareto distribution parameters / 定义帕累托分布参数
a = 1.5          # Shape parameter / 形状参数
scale = 1.0      # Minimum value / 最小值
num_samples = 10000  # Number of samples / 样本数

# Sample from Pareto distribution / 从帕累托分布中抽样
samples = pareto.rvs(b=a, scale=scale, size=num_samples)

print(f'Pareto Distribution Sample')
print(f'Shape parameter (a): {a}')
print(f'Scale parameter (m): {scale}')
print(f'Sample count: {num_samples}')
print(f'\nSample statistics:')
print(f'Mean: {np.mean(samples):.4f}')
print(f'Std Dev: {np.std(samples):.4f}')
print(f'Min: {np.min(samples):.4f}')
print(f'Max: {np.max(samples):.4f}')
print(f'Median: {np.median(samples):.4f}')

# Create visualization / 创建可视化
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Linear scale / 线性尺度
axes[0].hist(samples[samples < 20], bins=50, density=True, alpha=0.7, edgecolor='black')
axes[0].set_xlabel('Value / 值')
axes[0].set_ylabel('Density / 密度')
axes[0].set_title(f'Pareto Distribution (a={a}, Linear Scale)')
axes[0].grid(alpha=0.3)

# Log-log scale / 双对数尺度
axes[1].hist(samples, bins=50, density=True, alpha=0.7, edgecolor='black')
axes[1].set_xscale('log')
axes[1].set_yscale('log')
axes[1].set_xlabel('Value (log scale) / 值（对数尺度）')
axes[1].set_ylabel('Density (log scale) / 密度（对数尺度）')
axes[1].set_title(f'Pareto Distribution (a={a}, Log-Log Scale)')
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Plot theoretical PDF / 绘制理论PDF
fig, ax = plt.subplots(figsize=(12, 6))

x = np.linspace(scale, 10, 1000)
pdf = pareto.pdf(x, b=a, scale=scale)

ax.hist(samples[samples < 10], bins=50, density=True, alpha=0.6, edgecolor='black', label='Sample histogram')
ax.plot(x, pdf, 'r-', linewidth=2, label='Theoretical PDF')

ax.set_xlabel('Value / 值')
ax.set_ylabel('Probability Density / 概率密度')
ax.set_title(f'Pareto Distribution vs Sample (a={a})')
ax.legend()
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Plot with different shape parameters / 使用不同的形状参数绘图
fig, ax = plt.subplots(figsize=(12, 6))

shape_params = [0.5, 1.0, 1.5, 2.5]
x = np.linspace(1, 5, 1000)

for a_test in shape_params:
    pdf = pareto.pdf(x, b=a_test, scale=1.0)
    ax.plot(x, pdf, linewidth=2, label=f'a={a_test}')

ax.set_xlabel('Value / 值')
ax.set_ylabel('Probability Density / 概率密度')
ax.set_title('Pareto Distributions with Different Shape Parameters')
ax.legend()
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
