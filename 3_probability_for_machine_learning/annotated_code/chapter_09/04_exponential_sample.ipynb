## Summary

We sample from an exponential distribution using numpy.random.exponential(). The exponential distribution models the time between events in a Poisson process and is widely used in reliability engineering and queuing theory.

我们使用numpy.random.exponential()从指数分布中抽样。指数分布对泊松过程中事件之间的时间建模，广泛用于可靠性工程和排队论。

**Formula:**

$$f(x) = \lambda e^{-\lambda x}, \quad x \geq 0$$

$$E[X] = \frac{1}{\lambda}, \quad \text{Var}[X] = \frac{1}{\lambda^2}$$

## Step 1 — Import Libraries / 导入库

```python
# Import necessary libraries / 导入必要的库
import numpy as np
import matplotlib.pyplot as plt
```

## Step 2 — Set Parameters / 设置参数

```python
# Define exponential distribution parameters / 定义指数分布参数
# scale = 1/lambda (where lambda is the rate parameter)
# scale = 1/λ（其中λ是速率参数）
scale = 2.0          # Mean / 均值
num_samples = 10000  # Number of samples / 样本数
```

## Step 3 — Generate Exponential Samples / 生成指数样本

```python
# Sample from exponential distribution / 从指数分布中抽样
# numpy.random.exponential(scale=scale, size=num_samples)
samples = np.random.exponential(scale=scale, size=num_samples)

print(f'Exponential Distribution Sample / 指数分布样本')
print(f'Scale (mean) / 尺度（均值): {scale}')
print(f'Lambda (rate) / λ (速率): {1/scale:.4f}')
print(f'Sample count / 样本数: {num_samples}')
print(f'\nSample statistics / 样本统计:')
print(f'Mean / 均值: {np.mean(samples):.4f}')
print(f'Std Dev / 标准差: {np.std(samples):.4f}')
print(f'Min / 最小值: {np.min(samples):.4f}')
print(f'Max / 最大值: {np.max(samples):.4f}')
```

## Step 4 — Visualize Distribution / 可视化分布

```python
# Create visualization / 创建可视化
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histogram / 直方图
axes[0].hist(samples, bins=50, density=True, alpha=0.7, edgecolor='black')
axes[0].axvline(np.mean(samples), color='r', linestyle='--', linewidth=2, 
                 label=f'Sample Mean: {np.mean(samples):.2f}')
axes[0].axvline(scale, color='g', linestyle='--', linewidth=2, 
                 label=f'Theoretical Mean: {scale:.2f}')
axes[0].set_xlabel('Value / 值')
axes[0].set_ylabel('Density / 密度')
axes[0].set_title(f'Exponential Distribution (scale={scale})')
axes[0].legend()
axes[0].grid(alpha=0.3)

# Log scale histogram (to see tail better) / 对数尺度直方图
axes[1].hist(samples, bins=50, density=True, alpha=0.7, edgecolor='black')
axes[1].set_xlabel('Value / 值')
axes[1].set_ylabel('Density (log scale) / 密度（对数尺度）')
axes[1].set_yscale('log')
axes[1].set_title('Log Scale View')
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

## Step 5 — Plot Theoretical PDF / 绘制理论PDF

```python
# Plot theoretical PDF / 绘制理论PDF
from scipy.stats import expon

fig, ax = plt.subplots(figsize=(12, 6))

# Theoretical PDF / 理论PDF
x = np.linspace(0, 4*scale, 1000)
pdf = expon.pdf(x, scale=scale)

# Plot histogram and PDF / 绘制直方图和PDF
ax.hist(samples, bins=50, density=True, alpha=0.6, edgecolor='black', label='Sample histogram')
ax.plot(x, pdf, 'r-', linewidth=2, label='Theoretical PDF')
ax.axvline(scale, color='g', linestyle='--', linewidth=2, label=f'Mean: {scale}')

ax.set_xlabel('Value / 值')
ax.set_ylabel('Probability Density / 概率密度')
ax.set_title(f'Exponential Distribution vs Sample (scale={scale})')
ax.legend()
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

## Step 6 — Compare Different Scales / 比较不同的尺度

```python
# Plot multiple exponential distributions / 绘制多个指数分布
fig, ax = plt.subplots(figsize=(12, 6))

scales = [0.5, 1.0, 2.0, 3.0]
x = np.linspace(0, 10, 1000)

for s in scales:
    pdf = expon.pdf(x, scale=s)
    ax.plot(x, pdf, linewidth=2, label=f'scale={s} (λ={1/s:.2f})')

ax.set_xlabel('Value / 值')
ax.set_ylabel('Probability Density / 概率密度')
ax.set_title('Exponential Distributions with Different Scales')
ax.legend()
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

## Learning Notes / 学习笔记

- **Concept**: The exponential distribution models waiting times and lifetimes. Unlike the normal distribution, it is right-skewed with a long tail, meaning most values are small but occasional large values occur. It has the memoryless property: P(X > t+s | X > t) = P(X > s).
  
  **概念**：指数分布模拟等待时间和寿命。与正态分布不同，它是右偏的，有长尾，意味着大多数值很小但偶尔出现大值。它具有无记忆性质：P(X > t+s | X > t) = P(X > s)。

- **ML Application**: Exponential distribution is used in survival analysis, reliability modeling, queue modeling, and anomaly detection where you expect most values to be normal but occasional outliers.
  
  **机器学习应用**：指数分布用于生存分析、可靠性建模、排队建模和异常检测，其中期望大多数值正常但偶尔出现异常值。

## Complete Code / 完整代码一览

```python
# Complete Exponential Distribution Sampling / 完整指数分布抽样

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Define exponential distribution parameters / 定义指数分布参数
scale = 2.0          # Mean / 均值
num_samples = 10000  # Number of samples / 样本数

# Sample from exponential distribution / 从指数分布中抽样
samples = np.random.exponential(scale=scale, size=num_samples)

print(f'Exponential Distribution Sample')
print(f'Scale (mean): {scale}')
print(f'Lambda (rate): {1/scale:.4f}')
print(f'Sample count: {num_samples}')
print(f'\nSample statistics:')
print(f'Mean: {np.mean(samples):.4f}')
print(f'Std Dev: {np.std(samples):.4f}')
print(f'Min: {np.min(samples):.4f}')
print(f'Max: {np.max(samples):.4f}')

# Create visualization / 创建可视化
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histogram / 直方图
axes[0].hist(samples, bins=50, density=True, alpha=0.7, edgecolor='black')
axes[0].axvline(np.mean(samples), color='r', linestyle='--', linewidth=2, 
                 label=f'Sample Mean: {np.mean(samples):.2f}')
axes[0].axvline(scale, color='g', linestyle='--', linewidth=2, 
                 label=f'Theoretical Mean: {scale:.2f}')
axes[0].set_xlabel('Value / 值')
axes[0].set_ylabel('Density / 密度')
axes[0].set_title(f'Exponential Distribution (scale={scale})')
axes[0].legend()
axes[0].grid(alpha=0.3)

# Log scale histogram / 对数尺度直方图
axes[1].hist(samples, bins=50, density=True, alpha=0.7, edgecolor='black')
axes[1].set_xlabel('Value / 值')
axes[1].set_ylabel('Density (log scale) / 密度（对数尺度）')
axes[1].set_yscale('log')
axes[1].set_title('Log Scale View')
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Plot theoretical PDF / 绘制理论PDF
fig, ax = plt.subplots(figsize=(12, 6))

x = np.linspace(0, 4*scale, 1000)
pdf = expon.pdf(x, scale=scale)

ax.hist(samples, bins=50, density=True, alpha=0.6, edgecolor='black', label='Sample histogram')
ax.plot(x, pdf, 'r-', linewidth=2, label='Theoretical PDF')
ax.axvline(scale, color='g', linestyle='--', linewidth=2, label=f'Mean: {scale}')

ax.set_xlabel('Value / 值')
ax.set_ylabel('Probability Density / 概率密度')
ax.set_title(f'Exponential Distribution vs Sample (scale={scale})')
ax.legend()
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Plot multiple exponential distributions / 绘制多个指数分布
fig, ax = plt.subplots(figsize=(12, 6))

scales = [0.5, 1.0, 2.0, 3.0]
x = np.linspace(0, 10, 1000)

for s in scales:
    pdf = expon.pdf(x, scale=s)
    ax.plot(x, pdf, linewidth=2, label=f'scale={s} (λ={1/s:.2f})')

ax.set_xlabel('Value / 值')
ax.set_ylabel('Probability Density / 概率密度')
ax.set_title('Exponential Distributions with Different Scales')
ax.legend()
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
