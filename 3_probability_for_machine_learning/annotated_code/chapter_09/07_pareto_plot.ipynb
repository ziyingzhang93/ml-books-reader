## Summary

We plot the PDF and CDF of the Pareto distribution to visualize its heavy-tailed characteristics and show how it differs from other continuous distributions. This concludes Chapter 09.

我们绘制帕累托分布的PDF和CDF，以可视化其重尾特征，并显示它与其他连续分布的区别。这结束了第09章。

## Step 1 — Import Libraries / 导入库

```python
# Import necessary libraries / 导入必要的库
from scipy.stats import pareto
import numpy as np
import matplotlib.pyplot as plt
```

## Step 2 — Define Parameters / 定义参数

```python
# Define Pareto distribution parameters / 定义帕累托分布参数
a = 1.5      # Shape parameter / 形状参数
scale = 1.0  # Minimum value / 最小值
```

## Step 3 — Plot PDF and CDF / 绘制PDF和CDF

```python
# Create x values / 创建x值
x = np.linspace(scale, 8, 1000)

# Calculate PDF and CDF / 计算PDF和CDF
pdf = pareto.pdf(x, b=a, scale=scale)
cdf = pareto.cdf(x, b=a, scale=scale)

# Create figure with two subplots / 创建两个子图
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot PDF / 绘制PDF
axes[0].plot(x, pdf, 'b-', linewidth=2, label=f'PDF: Pareto(a={a})')
axes[0].fill_between(x, pdf, alpha=0.3)
axes[0].set_xlabel('Value / 值')
axes[0].set_ylabel('Probability Density / 概率密度')
axes[0].set_title('Pareto Distribution PDF')
axes[0].legend()
axes[0].grid(alpha=0.3)

# Plot CDF / 绘制CDF
axes[1].plot(x, cdf, 'g-', linewidth=2, label=f'CDF: Pareto(a={a})')
axes[1].axhline(0.5, color='r', linestyle='--', alpha=0.5, linewidth=1)
axes[1].set_xlabel('Value / 值')
axes[1].set_ylabel('Cumulative Probability / 累积概率')
axes[1].set_title('Pareto Distribution CDF')
axes[1].legend()
axes[1].grid(alpha=0.3)
axes[1].set_ylim([0, 1.05])

plt.tight_layout()
plt.show()
```

## Step 4 — Compare Different Shape Parameters / 比较不同形状参数

```python
# Compare PDF for different shape parameters / 比较不同形状参数的PDF
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

x = np.linspace(1, 8, 1000)
shape_params = [0.5, 1.0, 1.5, 2.5]

# PDF comparison / PDF比较
for a_test in shape_params:
    pdf = pareto.pdf(x, b=a_test, scale=1.0)
    axes[0].plot(x, pdf, linewidth=2, label=f'a={a_test}')

axes[0].set_xlabel('Value / 值')
axes[0].set_ylabel('Probability Density / 概率密度')
axes[0].set_title('PDF Comparison')
axes[0].legend()
axes[0].grid(alpha=0.3)

# CDF comparison / CDF比较
for a_test in shape_params:
    cdf = pareto.cdf(x, b=a_test, scale=1.0)
    axes[1].plot(x, cdf, linewidth=2, label=f'a={a_test}')

axes[1].set_xlabel('Value / 值')
axes[1].set_ylabel('Cumulative Probability / 累积概率')
axes[1].set_title('CDF Comparison')
axes[1].legend()
axes[1].grid(alpha=0.3)
axes[1].set_ylim([0, 1.05])

plt.tight_layout()
plt.show()
```

## Step 5 — Log-Log Plot (Power Law) / 双对数图（幂律）

```python
# Plot on log-log scale to show power law behavior / 在双对数尺度上绘制以显示幂律行为
fig, ax = plt.subplots(figsize=(12, 6))

x = np.linspace(1.01, 20, 1000)
pdf = pareto.pdf(x, b=a, scale=1.0)

# Remove zero values for log plotting / 移除零值以用于对数绘图
x_positive = x[pdf > 0]
pdf_positive = pdf[pdf > 0]

ax.loglog(x_positive, pdf_positive, 'b-', linewidth=2, label=f'Pareto(a={a})')
ax.set_xlabel('Value (log scale) / 值（对数尺度）')
ax.set_ylabel('Probability Density (log scale) / 概率密度（对数尺度）')
ax.set_title(f'Pareto Distribution - Power Law Behavior (a={a})')
ax.legend()
ax.grid(alpha=0.3, which='both')

plt.tight_layout()
plt.show()
```

## Learning Notes / 学习笔记

- **Concept**: The Pareto distribution exhibits power-law behavior, visible as a straight line on a log-log plot. The heavier the tail (smaller $a$), the more probability mass concentrates in the extreme values. This is fundamentally different from the exponential and normal distributions.
  
  **概念**：帕累托分布展现幂律行为，在双对数图上表现为直线。尾部越重（较小的$a$），概率质量越集中在极端值上。这与指数分布和正态分布根本不同。

- **ML Application**: Power-law distributions are found in natural and social phenomena (income, web traffic, earthquake magnitudes). Understanding them is essential for proper statistical modeling, risk assessment, and avoiding underestimation of extreme events.
  
  **机器学习应用**：幂律分布出现在自然和社会现象中（收入、网络流量、地震震级）。理解它们对于正确的统计建模、风险评估和避免低估极端事件至关重要。
