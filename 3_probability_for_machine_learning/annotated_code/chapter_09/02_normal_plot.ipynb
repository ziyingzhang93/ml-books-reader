## Summary

We plot both the Probability Density Function (PDF) and Cumulative Distribution Function (CDF) of the normal distribution using scipy.stats.norm. This allows us to visualize the theoretical distribution and understand its properties.

我们使用scipy.stats.norm绘制正态分布的概率密度函数（PDF）和累积分布函数（CDF）。这允许我们可视化理论分布并理解其性质。

## Step 1 — Import Libraries / 导入库

```python
# Import necessary libraries / 导入必要的库
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

## Step 2 — Define Parameters / 定义参数

```python
# Define normal distribution parameters / 定义正态分布参数
mu = 50      # Mean / 均值
sigma = 5    # Standard deviation / 标准差
```

## Step 3 — Plot PDF and CDF / 绘制PDF和CDF

```python
# Create x values / 创建x值
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Calculate PDF and CDF / 计算PDF和CDF
# norm.pdf(x, loc=mu, scale=sigma) - Probability Density Function
# norm.cdf(x, loc=mu, scale=sigma) - Cumulative Distribution Function
pdf = norm.pdf(x, loc=mu, scale=sigma)
cdf = norm.cdf(x, loc=mu, scale=sigma)

# Create figure with two subplots / 创建两个子图
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot PDF / 绘制PDF
axes[0].plot(x, pdf, 'b-', linewidth=2, label=f'PDF: N(μ={mu}, σ={sigma})')
axes[0].fill_between(x, pdf, alpha=0.3)
axes[0].axvline(mu, color='r', linestyle='--', linewidth=2, label=f'Mean: {mu}')
axes[0].set_xlabel('Value / 值')
axes[0].set_ylabel('Probability Density / 概率密度')
axes[0].set_title('Normal Distribution PDF')
axes[0].legend()
axes[0].grid(alpha=0.3)

# Plot CDF / 绘制CDF
axes[1].plot(x, cdf, 'g-', linewidth=2, label=f'CDF: N(μ={mu}, σ={sigma})')
axes[1].axhline(0.5, color='r', linestyle='--', alpha=0.5, linewidth=1)
axes[1].axvline(mu, color='r', linestyle='--', linewidth=2, label=f'Mean: {mu}')
axes[1].set_xlabel('Value / 值')
axes[1].set_ylabel('Cumulative Probability / 累积概率')
axes[1].set_title('Normal Distribution CDF')
axes[1].legend()
axes[1].grid(alpha=0.3)
axes[1].set_ylim([0, 1.05])

plt.tight_layout()
plt.show()
```

## Step 4 — Highlight Key Regions / 高亮关键区域

```python
# Visualize confidence intervals / 可视化置信区间
fig, ax = plt.subplots(figsize=(12, 6))

# Plot PDF / 绘制PDF
ax.plot(x, pdf, 'b-', linewidth=2, label='PDF')

# Highlight regions for confidence intervals / 高亮置信区间区域
# 68% interval (μ ± σ) / 68%区间
x1 = np.linspace(mu - sigma, mu + sigma, 100)
pdf1 = norm.pdf(x1, loc=mu, scale=sigma)
ax.fill_between(x1, pdf1, alpha=0.5, color='green', label='68% (μ ± σ)')

# 95% interval (μ ± 2σ) / 95%区间
x2 = np.linspace(mu - 2*sigma, mu + 2*sigma, 100)
pdf2 = norm.pdf(x2, loc=mu, scale=sigma)
ax.fill_between(x2, pdf2, alpha=0.3, color='yellow', label='95% (μ ± 2σ)')

# 99% interval (μ ± 3σ) / 99%区间
x3 = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
pdf3 = norm.pdf(x3, loc=mu, scale=sigma)
ax.fill_between(x3, pdf3, alpha=0.2, color='orange', label='99% (μ ± 3σ)')

ax.axvline(mu, color='r', linestyle='--', linewidth=2, label=f'Mean: {mu}')
ax.set_xlabel('Value / 值')
ax.set_ylabel('Probability Density / 概率密度')
ax.set_title(f'Confidence Intervals: Normal Distribution N(μ={mu}, σ={sigma})')
ax.legend()
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

## Step 5 — Compare Different Parameters / 比较不同参数

```python
# Plot multiple normal distributions / 绘制多个正态分布
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Different means, same std / 不同均值，相同标准差
x = np.linspace(30, 70, 1000)
for mu_test in [40, 50, 60]:
    pdf = norm.pdf(x, loc=mu_test, scale=sigma)
    axes[0].plot(x, pdf, linewidth=2, label=f'μ={mu_test}')

axes[0].set_xlabel('Value / 值')
axes[0].set_ylabel('Probability Density / 概率密度')
axes[0].set_title(f'Effect of Mean (σ={sigma} fixed)')
axes[0].legend()
axes[0].grid(alpha=0.3)

# Same mean, different std / 相同均值，不同标准差
x = np.linspace(30, 70, 1000)
for sigma_test in [2, 5, 10]:
    pdf = norm.pdf(x, loc=mu, scale=sigma_test)
    axes[1].plot(x, pdf, linewidth=2, label=f'σ={sigma_test}')

axes[1].set_xlabel('Value / 值')
axes[1].set_ylabel('Probability Density / 概率密度')
axes[1].set_title(f'Effect of Standard Deviation (μ={mu} fixed)')
axes[1].legend()
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

## Learning Notes / 学习笔记

- **Concept**: The PDF describes the relative likelihood of different values, while the CDF describes the probability of being below a certain value. For normal distributions, the PDF is bell-shaped and symmetric around the mean.
  
  **概念**：PDF描述了不同值的相对可能性，而CDF描述了低于某个值的概率。对于正态分布，PDF呈钟形且围绕均值对称。

- **ML Application**: Understanding PDF and CDF is essential for density-based anomaly detection, likelihood calculations in probabilistic models, and interpreting model confidence scores.
  
  **机器学习应用**：理解PDF和CDF对于基于密度的异常检测、概率模型中的似然计算和解释模型置信度评分至关重要。
