## Summary / 汇总

This notebook calculates the entropy of a dataset with 10% class 0 and 90% class 1. Entropy quantifies how mixed the classes are.

本笔记本计算具有10%的0类和90%的1类的数据集的熵。熵量化类的混合程度。

## Step 1 — Calculate Dataset Entropy / 计算数据集熵

```python
from math import log2# proportion of examples in each classclass0 = 10/100class1 = 90/100# calculate entropy: H(S) = -p0*log2(p0) - p1*log2(p1)entropy = -(class0 * log2(class0) + class1 * log2(class1))# print the resultprint('entropy: %.3f bits' % entropy)
```

## Learning Notes / 学习笔记

- **Concept**: Entropy H(S) = -Σ p_i*log₂(p_i) measures dataset purity. Balanced classes (10%, 90%) have lower entropy than (50%, 50%), indicating it's easier to classify homogeneous datasets. **概念**: 熵H(S) = -Σ p_i*log₂(p_i)衡量数据集的纯度。平衡类(10%, 90%)的熵低于(50%, 50%)，表示对同构数据集的分类更容易。

- **ML Application**: Decision trees select splits that maximize information gain, which is parent entropy minus weighted child entropy. Nodes with high entropy need further splitting. **机器学习应用**: 决策树选择最大化信息增益的分割，即父节点熵减去加权子节点熵。高熵节点需要进一步分割。
