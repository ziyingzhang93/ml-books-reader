# Chapter 8: Discrete Distributions

## Overview
This chapter builds understanding of **discrete probability distributions**, focusing on the binomial and multinomial distributions. The journey progresses from simulation to theoretical characterization.

## Key Concepts
- **Binomial Distribution**: Number of successes in n independent trials
- **Probability Mass Function (PMF)**: Probability of each discrete outcome
- **Cumulative Distribution Function (CDF)**: Probability up to and including a value
- **Distribution Moments**: Mean, variance, and other statistical summaries
- **Multinomial Distribution**: Extension to multiple categories (k > 2 outcomes)

## Evolution of Examples

### Binomial Distribution Journey
1. **01_binomial_simulation.py**: Simulate a single binomial trial (n=100, p=0.3)
2. **02_binomial_moments.py**: Calculate mean and variance from theory
3. **03_binomial_mass.py**: Plot the PMF (probability mass function)
4. **04_binomial_cumulative.py**: Plot the CDF (cumulative distribution)

### Multinomial Distribution Journey
5. **05_multinomial_simulate.py**: Simulate multinomial outcomes
6. **06_multinomial_mass.py**: Plot the PMF for multiple categories

## Logic Flow
**Simulate → Calculate → Visualize (Density) → Visualize (Cumulative) → Extend to Multiple Categories**

## Key Takeaways
1. Simulation confirms theoretical moments (mean, variance)
2. PMF shows concentration of probability around the mean
3. CDF shows cumulative probability increasing monotonically
4. Multinomial generalizes binomial to k > 2 outcomes
5. Understanding discrete distributions is foundational for classification, counting, and hypothesis testing
