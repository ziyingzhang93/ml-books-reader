## Summary

We simulate the multinomial distribution using numpy.random.multinomial(), which is a generalization of the binomial distribution to multiple categories. We demonstrate with 3 equally likely categories.

我们使用numpy.random.multinomial()模拟多项分布，这是二项分布到多个类别的推广。我们用3个等可能的类别演示。

**Formula:**

$$P(X_1 = x_1, X_2 = x_2, ..., X_k = x_k) = \frac{n!}{x_1! x_2! \cdots x_k!} p_1^{x_1} p_2^{x_2} \cdots p_k^{x_k}$$

## Step 1 — Import Libraries / 导入库

```python
# Import necessary libraries / 导入必要的库
import numpy as np
import matplotlib.pyplot as plt
```

## Step 2 — Set Parameters / 设置参数

```python
# Define multinomial parameters / 定义多项分布参数
n = 100              # Total number of trials / 总试验次数
k = 3                # Number of categories / 类别数
p = [1/3, 1/3, 1/3]  # Probability of each category (equal) / 每个类别的概率（相等）
num_samples = 10000  # Number of samples to generate / 生成的样本数
```

## Step 3 — Generate Multinomial Samples / 生成多项样本

```python
# Generate multinomial samples / 生成多项分布样本
# numpy.random.multinomial(n, pvals, size)
# Returns array of shape (size, k) where each row sums to n
# 返回形状为(size, k)的数组，每行和为n
samples = np.random.multinomial(n=n, pvals=p, size=num_samples)

print(f'Multinomial Distribution: n={n}, k={k}, p={p}')
print(f'多项分布: n={n}, k={k}, p={p}')
print(f'\nSample shape: {samples.shape}')
print(f'样本形状: {samples.shape}')
print(f'\nFirst 5 samples (each row sums to {n}):/')
print(f'前5个样本（每行的和为{n}）:')
for i in range(5):
    print(f'  Sample {i+1}: {samples[i]} (sum={samples[i].sum()})')
```

## Step 4 — Analyze Sample Statistics / 分析样本统计

```python
# Calculate statistics for each category / 计算每个类别的统计量
means = np.mean(samples, axis=0)
vars = np.var(samples, axis=0)
theoretical_mean = n * np.array(p)
theoretical_var = n * np.array(p) * (1 - np.array(p))

print('\nCategory Statistics / 类别统计:')
print('=' * 70)
print(f'{"Category":>10s} {"Sample Mean":>15s} {"Theoretical":>15s} {"Sample Var":>15s}')
print('-' * 70)

for i in range(k):
    print(f'Category {i+1}  {means[i]:15.2f} {theoretical_mean[i]:15.2f} {vars[i]:15.2f}')
```

## Step 5 — Visualize Distributions / 可视化分布

```python
# Plot histogram for each category / 为每个类别绘制直方图
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for i in range(k):
    ax = axes[i]
    # Plot histogram for category i / 绘制类别i的直方图
    ax.hist(samples[:, i], bins=30, density=True, alpha=0.7, edgecolor='black')
    ax.axvline(means[i], color='r', linestyle='--', linewidth=2, 
               label=f'Sample Mean: {means[i]:.2f}')
    ax.axvline(theoretical_mean[i], color='g', linestyle='--', linewidth=2,
               label=f'Theoretical: {theoretical_mean[i]:.2f}')
    ax.set_xlabel('Count / 计数')
    ax.set_ylabel('Density / 密度')
    ax.set_title(f'Category {i+1} Distribution')
    ax.legend()
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

## Learning Notes / 学习笔记

- **Concept**: The multinomial distribution extends the binomial distribution to multiple categories. It models the frequency of each outcome when sampling n items from k categories. The constraint is that the sum of all categories equals n.
  
  **概念**：多项分布将二项分布扩展到多个类别。它模拟从k个类别中抽样n个项目时每个结果的频率。约束条件是所有类别的总和等于n。

- **ML Application**: Multinomial distribution is fundamental for categorical variables in machine learning, including text classification (word counts), image classification (pixel distributions), and multiclass prediction problems.
  
  **机器学习应用**：多项分布对于机器学习中的分类变量至关重要，包括文本分类（单词计数）、图像分类（像素分布）和多类预测问题。

## Complete Code / 完整代码一览

```python
# Complete Multinomial Simulation / 完整多项分布模拟

import numpy as np
import matplotlib.pyplot as plt

# Define multinomial parameters / 定义多项分布参数
n = 100              # Total number of trials / 总试验次数
k = 3                # Number of categories / 类别数
p = [1/3, 1/3, 1/3]  # Probability of each category (equal) / 每个类别的概率（相等）
num_samples = 10000  # Number of samples to generate / 生成的样本数

# Generate multinomial samples / 生成多项分布样本
samples = np.random.multinomial(n=n, pvals=p, size=num_samples)

print(f'Multinomial Distribution: n={n}, k={k}, p={p}')
print(f'Sample shape: {samples.shape}')
print(f'\nFirst 5 samples:')
for i in range(5):
    print(f'  Sample {i+1}: {samples[i]} (sum={samples[i].sum()})')

# Calculate statistics for each category / 计算每个类别的统计量
means = np.mean(samples, axis=0)
vars = np.var(samples, axis=0)
theoretical_mean = n * np.array(p)
theoretical_var = n * np.array(p) * (1 - np.array(p))

print('\nCategory Statistics:')
print('=' * 70)
print(f'{"Category":>10s} {"Sample Mean":>15s} {"Theoretical":>15s} {"Sample Var":>15s}')
print('-' * 70)

for i in range(k):
    print(f'Category {i+1}  {means[i]:15.2f} {theoretical_mean[i]:15.2f} {vars[i]:15.2f}')

# Plot histogram for each category / 为每个类别绘制直方图
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for i in range(k):
    ax = axes[i]
    ax.hist(samples[:, i], bins=30, density=True, alpha=0.7, edgecolor='black')
    ax.axvline(means[i], color='r', linestyle='--', linewidth=2, 
               label=f'Sample Mean: {means[i]:.2f}')
    ax.axvline(theoretical_mean[i], color='g', linestyle='--', linewidth=2,
               label=f'Theoretical: {theoretical_mean[i]:.2f}')
    ax.set_xlabel('Count / 计数')
    ax.set_ylabel('Density / 密度')
    ax.set_title(f'Category {i+1} Distribution')
    ax.legend()
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
