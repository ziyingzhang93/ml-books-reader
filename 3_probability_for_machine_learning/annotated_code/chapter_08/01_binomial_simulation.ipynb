## Summary

We simulate the binomial distribution using numpy.random.binomial(), which generates random samples from a binomial distribution with parameters n (number of trials) and p (probability of success).

我们使用numpy.random.binomial()来模拟二项分布，它根据参数n（试验次数）和p（成功概率）生成随机样本。

**Formula:**

$$P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}$$

## Step 1 — Import Libraries / 导入库

```python
# Import necessary libraries / 导入必要的库
import numpy as np
import matplotlib.pyplot as plt
```

## Step 2 — Set Parameters / 设置参数

```python
# Define binomial parameters / 定义二项分布参数
n = 100          # Number of trials / 试验次数
p = 0.3          # Probability of success / 成功概率
num_samples = 10000  # Number of samples to generate / 生成的样本数
```

## Step 3 — Generate Binomial Samples / 生成二项样本

```python
# Generate binomial samples / 生成二项分布样本
# numpy.random.binomial(n, p, size) generates samples from Binomial(n, p)
# numpy.random.binomial(n, p, size) 从二项分布(n, p)生成样本
samples = np.random.binomial(n=n, p=p, size=num_samples)

print(f'Generated {num_samples} binomial samples with n={n}, p={p}')
print(f'生成了{num_samples}个二项分布样本，n={n}, p={p}')
```

## Step 4 — Analyze Samples / 分析样本

```python
# Calculate statistics / 计算统计量
sample_mean = np.mean(samples)
sample_var = np.var(samples)
theoretical_mean = n * p           # E[X] = np / E[X] = np
theoretical_var = n * p * (1 - p)  # Var[X] = np(1-p) / Var[X] = np(1-p)

print(f'\nSample mean: {sample_mean:.2f}, Theoretical: {theoretical_mean:.2f}')
print(f'样本均值: {sample_mean:.2f}, 理论值: {theoretical_mean:.2f}')
print(f'Sample variance: {sample_var:.2f}, Theoretical: {theoretical_var:.2f}')
print(f'样本方差: {sample_var:.2f}, 理论值: {theoretical_var:.2f}')
```

## Step 5 — Visualize Distribution / 可视化分布

```python
# Plot histogram of samples / 绘制样本直方图
plt.figure(figsize=(10, 6))
plt.hist(samples, bins=30, density=True, alpha=0.7, edgecolor='black')
plt.axvline(sample_mean, color='r', linestyle='--', linewidth=2, label=f'Sample Mean: {sample_mean:.2f}')
plt.axvline(theoretical_mean, color='g', linestyle='--', linewidth=2, label=f'Theoretical Mean: {theoretical_mean:.2f}')
plt.xlabel('Number of Successes / 成功次数')
plt.ylabel('Density / 密度')
plt.title(f'Binomial Distribution Simulation (n={n}, p={p}) / 二项分布模拟')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

## Learning Notes / 学习笔记

- **Concept**: The binomial distribution models the number of successes in a fixed number of independent trials. It's one of the most important discrete distributions in probability and statistics.
  
  **概念**：二项分布模拟了固定次数独立试验中的成功次数。它是概率和统计中最重要的离散分布之一。

- **ML Application**: Binomial distribution is fundamental for classification problems, A/B testing, click-through rate modeling, and any scenario with binary outcomes.
  
  **机器学习应用**：二项分布对于分类问题、A/B测试、点击率建模和任何二元结果场景都很基础。
