## Summary

We calculate the Cumulative Distribution Function (CDF) of the binomial distribution using scipy.stats.binom.cdf(). The CDF gives the probability that a random variable is less than or equal to a given value.

我们使用scipy.stats.binom.cdf()计算二项分布的累积分布函数（CDF）。CDF给出随机变量小于或等于给定值的概率。

**Formula:**

$$F(k) = P(X \leq k) = \sum_{i=0}^{k} \binom{n}{i} p^i (1-p)^{n-i}$$

## Step 1 — Import Libraries / 导入库

```python
# Import necessary libraries / 导入必要的库
from scipy.stats import binom
import numpy as np
import matplotlib.pyplot as plt
```

## Step 2 — Calculate CDF for Single Distribution / 计算单一分布的CDF

```python
# Define parameters / 定义参数
n = 20
p = 0.3

# Calculate CDF for all possible values / 计算所有可能值的CDF
# CDF(k) = P(X <= k) = sum of PMF from 0 to k
# CDF(k) = P(X <= k) = 从0到k的PMF之和
x = np.arange(0, n+1)
cdf = binom.cdf(k=x, n=n, p=p)

print(f'Binomial CDF: n={n}, p={p}')
print(f'二项CDF: n={n}, p={p}')
print(f'\n{"k":>3s} {"P(X<=k)":>12s}')
print('=' * 17)
for k, cdf_val in zip(x[::2], cdf[::2]):  # Print every other value to save space
    print(f'{k:3d} {cdf_val:12.6f}')
```

## Step 3 — Plot PMF and CDF Together / 同时绘制PMF和CDF

```python
# Create figure with two subplots / 创建两个子图
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

n = 20
p = 0.3
x = np.arange(0, n+1)

# PMF / PMF
from scipy.stats import binom
pmf = binom.pmf(k=x, n=n, p=p)
cdf = binom.cdf(k=x, n=n, p=p)

# Plot PMF / 绘制PMF
axes[0].bar(x, pmf, alpha=0.7, edgecolor='black', color='blue')
axes[0].set_xlabel('Number of Successes / 成功次数')
axes[0].set_ylabel('Probability / 概率')
axes[0].set_title(f'PMF: Binomial(n={n}, p={p})')
axes[0].grid(alpha=0.3)

# Plot CDF / 绘制CDF
axes[1].step(x, cdf, where='mid', linewidth=2, color='red', label='CDF')
axes[1].scatter(x, cdf, color='red', s=50, zorder=5)
axes[1].set_xlabel('Number of Successes / 成功次数')
axes[1].set_ylabel('Cumulative Probability / 累积概率')
axes[1].set_title(f'CDF: Binomial(n={n}, p={p})')
axes[1].grid(alpha=0.3)
axes[1].set_ylim([0, 1.05])

plt.tight_layout()
plt.show()
```

## Step 4 — Calculate Probabilities Using CDF / 使用CDF计算概率

```python
# Calculate different probabilities / 计算不同的概率
n = 20
p = 0.3

# P(X <= 5) / P(X <= 5)
prob_le_5 = binom.cdf(5, n=n, p=p)
print(f'P(X <= 5) = {prob_le_5:.6f}')

# P(X < 5) = P(X <= 4) / P(X < 5) = P(X <= 4)
prob_lt_5 = binom.cdf(4, n=n, p=p)
print(f'P(X < 5) = P(X <= 4) = {prob_lt_5:.6f}')

# P(X > 5) = 1 - P(X <= 5) / P(X > 5) = 1 - P(X <= 5)
prob_gt_5 = 1 - binom.cdf(5, n=n, p=p)
print(f'P(X > 5) = 1 - P(X <= 5) = {prob_gt_5:.6f}')

# P(X >= 5) = 1 - P(X <= 4) / P(X >= 5) = 1 - P(X <= 4)
prob_ge_5 = 1 - binom.cdf(4, n=n, p=p)
print(f'P(X >= 5) = 1 - P(X <= 4) = {prob_ge_5:.6f}')

# P(5 <= X <= 10) = P(X <= 10) - P(X <= 4) / P(5 <= X <= 10) = P(X <= 10) - P(X <= 4)
prob_range = binom.cdf(10, n=n, p=p) - binom.cdf(4, n=n, p=p)
print(f'P(5 <= X <= 10) = {prob_range:.6f}')
```

## Learning Notes / 学习笔记

- **Concept**: The CDF is a monotonically increasing function that goes from 0 to 1. It's useful for calculating tail probabilities and finding quantiles. CDF and PMF contain the same information but in different forms.
  
  **概念**：CDF是从0单调递增到1的函数。它对于计算尾部概率和找到分位数很有用。CDF和PMF包含相同的信息，只是形式不同。

- **ML Application**: CDF is essential for calculating p-values in hypothesis testing, determining confidence intervals, and evaluating tail risks in probabilistic models.
  
  **机器学习应用**：CDF对于计算假设检验中的p值、确定置信区间和评估概率模型中的尾部风险至关重要。
