### Summary / 总结
Learn how to create arrays from lists using `asarray()` and inspect their shape. This is foundational for understanding data structure before applying aggregate functions.

学习如何使用`asarray()`从列表创建数组并检查其形状。这是应用聚合函数前理解数据结构的基础。

### Core Concept / 核心概念
**asarray()**: Converts input to array format (like `array()` but doesn't copy if already an array). Shape reveals dimensions: (2, 3) means 2 rows, 3 columns.

**asarray()**: 将输入转换为数组格式（如`array()`但如果已是数组则不复制）。形状显示维度：(2, 3)表示2行3列。

## Step 1 — Create Array from Nested List / 从嵌套列表创建数组

We create a 2D array from a nested list using `asarray()`. This converts the list structure into a NumPy array.

我们使用`asarray()`从嵌套列表创建二维数组。这将列表结构转换为NumPy数组。

```python3
# 导入asarray函数 / Import asarray function
from numpy import asarray

# 创建嵌套列表 / Create nested list
data = [[1, 2, 3], [4, 5, 6]]
print("Python nested list:")
print(data)
```

## Step 2 — Convert to Array / 转换为数组

We convert the list to a NumPy array using `asarray()`. This enables efficient array operations.

我们使用`asarray()`将列表转换为NumPy数组。这使高效的数组操作成为可能。

```python3
# 使用asarray转换 / Convert using asarray
data = asarray(data)
print("\nNumPy array:")
print(data)
```

## Step 3 — Inspect Shape / 检查形状

We check the array's shape, which tells us it has 2 rows and 3 columns. This is essential information for aggregate operations.

我们检查数组的形状，它告诉我们它有2行3列。这对于聚合操作是重要信息。

```python3
# 显示数组和形状 / Display array and shape
print("\nArray:")
print(data)
print("\nShape:", data.shape)
```

## Learning Notes / 学习笔记

• **Math Essence / 数学本质**: Understanding array structure through shape is fundamental. Shape (m, n) describes a matrix with m rows and n columns. Before aggregating, knowing the shape helps interpret which dimension the operation applies to.

通过形状理解数组结构是基本的。形状(m, n)描述具有m行n列的矩阵。在聚合前，了解形状有助于解释操作适用于哪个维度。

• **ML Application / ML applicaton**: In ML pipelines, checking data shape is the first step. Before applying aggregations like sum, mean, or max, you must know: are there 2 or 3 dimensions? Are we summing along rows or columns? Shape determines aggregation behavior.

在ML管道中，检查数据形状是第一步。在应用sum、mean或max等聚合前，你必须知道：有2个还是3个维度？我们沿着行还是列求和？形状决定聚合行为。

## Complete Code / 完整代码一览

```python3
# --- Array Creation and Shape Inspection / 数组创建和形状检查 ---
from numpy import asarray

# Create array from nested list / 从嵌套列表创建数组
data = [[1, 2, 3], [4, 5, 6]]
print("Python nested list:")
print(data)

# Convert to array / 转换为数组
data = asarray(data)
print("\nNumPy array:")
print(data)

# Check shape / 检查形状
print("\nShape:", data.shape)
```
