## Summary / 总结

Load and examine the wine dataset from scikit-learn. This dataset contains 178 wine samples with 13 chemical features and 3 wine classes, perfect for demonstrating PCA visualization.

从scikit-learn加载并检查葡萄酒数据集。该数据集包含178个葡萄酒样本，13个化学特征和3个葡萄酒类别。

## Dataset Overview / 数据集概览

- **Samples**: 178 wine observations
- **Features**: 13 chemical properties (alcohol, acidity, phenols, etc.)
- **Classes**: 3 wine cultivars

## Step 1 — Import Libraries / 导入库

```python3
# Import dataset loader
# 导入数据集加载器
from sklearn.datasets import load_wine
```

## Step 2 — Load the Dataset / 加载数据集

```python3
# Load wine dataset from sklearn
# Returns a Bunch object with data, target, feature names, and description
# 从sklearn加载葡萄酒数据集
# 返回包含数据、目标、特征名称和描述的Bunch对象
winedata = load_wine()
X, y = winedata['data'], winedata['target']
print(f"Dataset loaded successfully")
```

## Step 3 — Examine Dataset Dimensions / 检查数据集维度

```python3
# Check the shape of X and y
# X: (n_samples, n_features) = (178, 13)
# y: (n_samples,) = (178,)
# 检查X和y的形状
print(f"X shape (samples, features): {X.shape}")
print(f"y shape (targets): {y.shape}")
print(f"\nNumber of samples: {X.shape[0]}")
print(f"Number of features: {X.shape[1]}")
print(f"Number of classes: {len(set(y))}")
```

## Step 4 — Explore Feature Names / 探索特征名称

```python3
# Display feature names
# 显示特征名称
print("Feature names:")
for i, name in enumerate(winedata['feature_names']):
    print(f"  {i}: {name}")
```

## Step 5 — Explore Class Labels / 探索类别标签

```python3
# Display target names and class distribution
# 显示目标名称和类别分布
print("Class labels:")
for i, name in enumerate(winedata['target_names']):
    count = (y == i).sum()
    print(f"  {i}: {name} ({count} samples)")
```

## Step 6 — Examine Data Statistics / 检查数据统计

```python3
# Display data statistics
# 显示数据统计
import numpy as np
print("Data statistics (first 5 features):")
for i in range(5):
    print(f"\n{winedata['feature_names'][i]}:")
    print(f"  Min: {X[:, i].min():.3f}, Max: {X[:, i].max():.3f}")
    print(f"  Mean: {X[:, i].mean():.3f}, Std: {X[:, i].std():.3f}")
```

## Step 7 — Display Sample Data / 显示样本数据

```python3
# Show first few samples
# 显示前几个样本
print("First 3 samples:")
print(f"\nSample 0 (class {y[0]}): {X[0]}")
print(f"\nSample 1 (class {y[1]}): {X[1]}")
print(f"\nSample 2 (class {y[2]}): {X[2]}")
```

- **ML Application**: Understanding data dimensions and distributions is crucial before applying any algorithm. Check feature scaling needs, class balance, and feature relationships before PCA or classification.
  
  **ML应用**：在应用任何算法前理解数据维度和分布至关重要。在PCA或分类前检查特征缩放需求、类别平衡和特征关系。
