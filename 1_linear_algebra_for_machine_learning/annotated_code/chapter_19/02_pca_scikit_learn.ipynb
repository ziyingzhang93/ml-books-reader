## Summary / 总结

Use scikit-learn's PCA class for efficient and practical dimensionality reduction. scikit-learn handles normalization and provides convenient methods for accessing explained variance and transforming data.

使用scikit-learn的PCA类进行高效和实用的降维。scikit-learn处理归一化并提供方便的方法来访问解释方差和变换数据。

## Core Formula / 核心公式

$$X_{transformed} = (X - \mu) @ V^T$$

where $V$ are the principal components (eigenvectors) and $\mu$ is the mean.

## Step 1 — Import Libraries / 导入库

```python
# Import array and PCA from scikit-learn
# 导入数组和scikit-learn中的PCA
from numpy import array
from sklearn.decomposition import PCA
```

## Step 2 — Create Sample Data / 创建样本数据

```python
# Define a simple 3x2 dataset
# 定义一个简单的3x2数据集
A = array([[1, 2],
           [3, 4],
           [5, 6]])
print("Original data A:")
print(A)
```

## Step 3 — Initialize and Fit PCA / 初始化并拟合PCA

```python
# Create PCA object with 2 components (can't have more than min(n_samples, n_features))
# fit() computes the principal components from the data
# 创建有2个成分的PCA对象
# fit()从数据计算主成分
pca = PCA(n_components=2)
pca.fit(A)
print(f"PCA fitted successfully")
print(f"Number of components: {pca.n_components_}")
```

## Step 4 — Access Principal Components / 访问主成分

```python
# components_ contains the principal directions (eigenvectors)
# Each row is one principal component
# components_包含主方向（特征向量）
# 每行是一个主成分
print("Principal component directions (eigenvectors):")
print(pca.components_)
```

## Step 5 — Access Explained Variance / 访问解释方差

```python
# explained_variance_ contains the eigenvalues (variance explained by each PC)
# explained_variance_ratio_ is the proportion of variance explained
# explained_variance_包含特征值（每个PC解释的方差）
# explained_variance_ratio_是解释的方差比例
print("Explained variance by each component:")
print(pca.explained_variance_)

print("\nExplained variance ratio (cumulative):")
cumsum = 0
for i, var_ratio in enumerate(pca.explained_variance_ratio_):
    cumsum += var_ratio
    print(f"  PC{i+1}: {var_ratio:.4f} (cumulative: {cumsum:.4f})")
```

## Step 6 — Transform Data / 变换数据

```python
# transform() projects the data onto the principal components
# Returns n_samples x n_components matrix
# transform()将数据投影到主成分上
# 返回 n_samples x n_components 矩阵
B = pca.transform(A)
print("Transformed data (in PC space):")
print(B)
```

## Step 7 — Inverse Transform / 逆变换

```python
# inverse_transform() reconstructs the original space from PC coordinates
# Can be used for dimensionality reduction and reconstruction
# inverse_transform()从PC坐标重建原始空间
# 可用于降维和重建
A_reconstructed = pca.inverse_transform(B)
print("Reconstructed data (should be close to original):")
print(A_reconstructed)

print("\nReconstruction error (should be near zero):")
import numpy as np
error = np.mean((A - A_reconstructed)**2)
print(f"Mean squared error: {error:.6f}")
```

## Learning Notes / 学习笔记

- **Math Essence**: scikit-learn's PCA uses SVD (Singular Value Decomposition) internally instead of eigendecomposition, which is more numerically stable. It handles all preprocessing (centering) automatically.
  
  **数学本质**：scikit-learn的PCA内部使用SVD（奇异值分解）而不是特征分解，这在数值上更稳定。它自动处理所有预处理（中心化）。

- **ML Application**: Use scikit-learn's PCA for practical applications. Key uses: (1) Visualize high-dimensional data by keeping 2-3 components, (2) Reduce dimensionality while preserving variance (keep components with 95%+ explained variance), (3) Denoise data by discarding low-variance components.
  
  **ML应用**：使用scikit-learn的PCA进行实际应用。关键用途：(1) 通过保留2-3个成分可视化高维数据，(2) 在保留方差的同时降低维度，(3) 通过丢弃低方差成分来去噪数据。

## Complete Code / 完整代码一览

```python
# --- Import Section / 导入部分 ---
from numpy import array
from sklearn.decomposition import PCA

# --- PCA with scikit-learn / 使用scikit-learn的PCA ---
A = array([[1, 2],
           [3, 4],
           [5, 6]])
print(A)

# --- Fit and Transform / 拟合和变换 ---
pca = PCA(2)
pca.fit(A)
print(pca.components_)
print(pca.explained_variance_)

# --- Project Data / 投影数据 ---
B = pca.transform(A)
print(B)
```
