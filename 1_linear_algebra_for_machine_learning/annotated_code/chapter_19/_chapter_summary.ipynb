# Chapter 19 Summary / 第19章总结：Principal Component Analysis (PCA)

## Theme / 主题

PCA is the most widely used dimensionality reduction technique. It finds the directions of maximum variance in data and projects data onto these directions. Understanding PCA from scratch—without a library—teaches the linear algebra concepts (covariance, eigendecomposition). Then using scikit-learn shows the production API. This progression connects theory to practice.

PCA是使用最广泛的降维技术。它找到数据中最大方差的方向，并将数据投影到这些方向上。从头开始理解PCA——不使用库——教导线性代数概念（协方差、特征分解）。然后使用scikit-learn展示生产API。这个进度连接理论和实践。

## Progression Logic / 进度逻辑

PCA is learned through **understanding → application**:

1. **From scratch**: 
   - Center data: subtract mean
   - Covariance: Cov = X^T · X / n
   - Eigendecomposition: find Q (eigenvectors) and Λ (eigenvalues)
   - Eigenvectors = principal components (directions of max variance)
   - Eigenvalues = variance along each component
   - Projection: X_pca = X · Q[:, :k] (keep top k)

2. **With scikit-learn**:
   - `PCA(n_components=k)` does all above automatically
   - `.fit(X)` learns principal components
   - `.transform(X)` projects data
   - `.explained_variance_` = eigenvalues
   - `.components_` = eigenvectors

The insight: **PCA is eigendecomposition of the centered covariance matrix**. Understanding the manual version reveals the beautiful math.

PCA通过**理解→应用**学习：

1. **从头开始**：
   - 中心化数据：减去均值
   - 协方差：Cov = X^T · X / n
   - 特征分解：找到Q（特征向量）和Λ（特征值）
   - 特征向量=主成分（最大方差的方向）
   - 特征值=沿每个分量的方差
   - 投影：X_pca = X · Q[:, :k]（保留前k个）

2. **使用scikit-learn**：
   - `PCA(n_components=k)`自动执行上述所有操作
   - `.fit(X)`学习主成分
   - `.transform(X)`投影数据
   - `.explained_variance_` =特征值
   - `.components_` =特征向量

见解：**PCA是中心化协方差矩阵的特征分解**。理解手动版本揭示了美丽的数学。

## ML Relevance / 机器学习相关性

In machine learning:
- **Dimensionality reduction**:
  - High-dimensional data: 1000+ features → 2, 3, 50 dimensions
  - Visualization: PCA to 2D/3D for plotting
  - Faster training: fewer features → faster algorithms
  - Noise removal: small eigenvalues are mostly noise

- **Data preprocessing**:
  - Fit PCA on training data
  - Transform training data
  - Transform test data with SAME PCA (fitted on training)
  - Critical: never fit PCA on test data (data leakage!)

- **Explained variance**:
  - `explained_variance_ratio_[i]` = λ_i / sum(λ)
  - Cumulative sum tells how much variance retained
  - Choose k where cumulative ratio >= 0.95 (95% variance)

- **Common applications**:
  - Face recognition: Eigenfaces (PCA on face images)
  - Gene expression: compress thousands of genes to interpretable components
  - Image compression: keep top k principal components
  - Outlier detection: points far from principal subspace are anomalies

- **Whitening via PCA**:
  - PCA decorrelates features
  - Whitened data: `X_white = X_pca / sqrt(eigenvalues)`
  - Results in uncorrelated, unit-variance features
  - Improves conditioning for downstream algorithms

- **PCA limitations**:
  - Linear only: captures linear relationships
  - Unsupervised: ignores labels (nonlinear methods like t-SNE might be better)
  - Interpretation: principal components are linear combinations (hard to explain)

**Key insight**: PCA is not just dimensionality reduction—it reveals the intrinsic structure of data. The eigenvectors are the "natural axes" of your data.

在机器学习中：
- **降维**：
  - 高维数据：1000+特征→2、3、50维
  - 可视化：PCA到2D/3D用于绘制
  - 更快的训练：更少特征→更快的算法
  - 噪声去除：小特征值主要是噪声

- **数据预处理**：
  - 在训练数据上拟合PCA
  - 变换训练数据
  - 使用相同的PCA转换测试数据（在训练上拟合）
  - 关键：永远不要在测试数据上拟合PCA（数据泄漏！）

- **解释方差**：
  - `explained_variance_ratio_[i]` = λ_i / sum(λ)
  - 累积和告诉保留了多少方差
  - 选择k，其中累积比率>= 0.95（95%方差）

- **常见应用**：
  - 面部识别：Eigenfaces（面部图像上的PCA）
  - 基因表达：将数千个基因压缩为可解释的分量
  - 图像压缩：保留前k个主成分
  - 异常检测：远离主子空间的点是异常

- **通过PCA的白化**：
  - PCA去相关特征
  - 白化数据：`X_white = X_pca / sqrt(eigenvalues)`
  - 导致不相关的、单位方差的特征
  - 改善下游算法的条件

- **PCA限制**：
  - 仅线性：捕获线性关系
  - 无监督：忽略标签（非线性方法如t-SNE可能更好）
  - 解释：主成分是线性组合（难以解释）

**关键见解**：PCA不仅仅是降维——它揭示了数据的固有结构。特征向量是您数据的"自然轴"。
