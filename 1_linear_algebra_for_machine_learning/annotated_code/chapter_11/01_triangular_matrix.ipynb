## Summary / 总结

Learn to extract lower and upper triangular matrices, which are useful in matrix decompositions and numerical algorithms.

学习如何提取下三角矩阵和上三角矩阵，这在矩阵分解和数值算法中很有用。

## Core Formula / 核心公式

Lower triangular: $L_{ij} = A_{ij}$ if $i \geq j$, else $0$

Upper triangular: $U_{ij} = A_{ij}$ if $i \leq j$, else $0$

## Step 1 — Import Required Functions / 导入所需函数

We import the `array` function and the `tril` and `triu` functions from NumPy.

```python
# Import array to create matrices
# 导入 array 来创建矩阵
from numpy import array

# Import tril to extract lower triangular matrix
# 导入 tril 来提取下三角矩阵
from numpy import tril

# Import triu to extract upper triangular matrix
# 导入 triu 来提取上三角矩阵
from numpy import triu
```

## Step 2 — Create a Matrix / 创建矩阵

We create a square matrix to demonstrate triangular matrix extraction.

```python
# Create a 3x3 matrix M
# 创建一个 3x3 矩阵 M
M = array([[1, 2, 3],[1, 2, 3],[1, 2, 3]])

# Print the original matrix
# 打印原始矩阵
print(M)
```

## Step 3 — Extract Lower Triangular Matrix / 提取下三角矩阵

We use `tril()` to extract the lower triangular part (including the diagonal).

```python
# Extract lower triangular matrix (keep elements where row index >= column index)
# 提取下三角矩阵（保持行索引 >= 列索引的元素）
lower = tril(M)

# Print the lower triangular matrix
# 打印下三角矩阵
print(lower)
```

## Step 4 — Extract Upper Triangular Matrix / 提取上三角矩阵

We use `triu()` to extract the upper triangular part (including the diagonal).

```python
# Extract upper triangular matrix (keep elements where row index <= column index)
# 提取上三角矩阵（保持行索引 <= 列索引的元素）
upper = triu(M)

# Print the upper triangular matrix
# 打印上三角矩阵
print(upper)
```

## Learning Notes / 学习笔记

- **Math Essence / 数学本质**: Triangular matrices have zeros either above or below the diagonal. They are important in matrix decompositions like LU decomposition and enable efficient solving of linear systems.
- **ML Application / 机器学习应用**: Triangular matrices appear in LU decomposition for efficient matrix inversion and in numerical methods for solving systems of linear equations.
