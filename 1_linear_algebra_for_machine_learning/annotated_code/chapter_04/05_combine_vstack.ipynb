### Summary / 总结
Learn how to combine multiple arrays vertically (row-wise) using `vstack()`. This stacks arrays one on top of another.

学习如何使用`vstack()`纵向（按行）组合多个数组。这将数组堆叠在一起。

### Core Concept / 核心概念
**numpy.vstack((arr1, arr2, ...))**: Vertically stacks sequences of arrays (along axis 0). Equivalent to concatenating rows.

**numpy.vstack((arr1, arr2, ...))**: 垂直堆叠数组序列（沿轴0）。等同于连接行。

## Step 1 — Create First Array / 创建第一个数组

We create the first 1D array containing [1, 2, 3]. This will become the first row of our stacked result.

我们创建第一个包含[1, 2, 3]的一维数组。这将成为我们堆叠结果的第一行。

```python
# 导入array和vstack函数 / Import array and vstack functions
from numpy import array
from numpy import vstack

# 创建第一个数组 / Create first array
a1 = array([1, 2, 3])
print("First array:")
print(a1)
print("Shape:", a1.shape)
```

## Step 2 — Create Second Array / 创建第二个数组

We create the second 1D array containing [4, 5, 6]. This will become the second row of our stacked result.

我们创建第二个包含[4, 5, 6]的一维数组。这将成为我们堆叠结果的第二行。

```python
# 创建第二个数组 / Create second array
a2 = array([4, 5, 6])
print("\nSecond array:")
print(a2)
print("Shape:", a2.shape)
```

## Step 3 — Stack Arrays Vertically / 纵向堆叠数组

We use `vstack()` to combine a1 and a2 into a 2D array. The two 1D arrays become rows in the result.

我们使用`vstack()`将a1和a2组合成一个二维数组。两个一维数组成为结果中的行。

```python
# 使用vstack纵向堆叠 / Stack vertically using vstack
a3 = vstack((a1, a2))
print("\nVertically stacked array:")
print(a3)
```

## Step 4 — Inspect Resulting Shape / 检查结果形状

The resulting array a3 is now 2D with shape (2, 3): 2 rows (from the two input arrays) and 3 columns (from the length of each input).

结果数组a3现在是二维的，形状为(2, 3)：2行（来自两个输入数组）和3列（来自每个输入的长度）。

```python
# 检查结果的形状 / Check shape of result
print("\nShape of stacked array:")
print(a3.shape)
```

## Learning Notes / 学习笔记

• **Math Essence / 数学本质**: vstack performs vertical stacking along axis 0 (rows). Mathematically, this is concatenating vectors as rows of a matrix. For 1D arrays [a1, a2], vstack creates matrix [[a1], [a2]].

vstack沿轴0（行）执行垂直堆叠。数学上，这是将向量连接为矩阵的行。对于一维数组[a1, a2]，vstack创建矩阵[[a1], [a2]]。

• **ML Application / ML应用**: In ML, vstack is used to combine multiple data samples into a dataset matrix. When loading batches of data from different sources, vstack assembles them into a single feature matrix for training.

在ML中，vstack用于将多个数据样本组合成数据集矩阵。从不同源加载多批数据时，vstack将它们组合成单个特征矩阵用于训练。

## Complete Code / 完整代码一览

```python
# --- Vertical Stacking / 垂直堆叠 ---
from numpy import array
from numpy import vstack

# Create two 1D arrays / 创建两个一维数组
a1 = array([1, 2, 3])
print("First array:")
print(a1)
print("Shape:", a1.shape)

a2 = array([4, 5, 6])
print("\nSecond array:")
print(a2)
print("Shape:", a2.shape)

# Stack vertically / 纵向堆叠
a3 = vstack((a1, a2))
print("\nVertically stacked array:")
print(a3)
print("Shape:", a3.shape)
```
