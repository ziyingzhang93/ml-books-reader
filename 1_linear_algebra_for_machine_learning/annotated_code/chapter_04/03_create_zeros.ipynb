### Summary / 总结
Learn how to create arrays initialized with zeros using `zeros()`. This is one of the most common array creation methods in scientific computing.

学习如何使用`zeros()`创建用零初始化的数组。这是科学计算中最常见的数组创建方法之一。

### Core Concept / 核心概念
**numpy.zeros(shape)**: Creates an array of given shape filled with zeros. Useful for initialization of matrices and placeholders.

**numpy.zeros(shape)**: 创建给定形状的数组，填充为零。对矩阵初始化和占位符很有用。

## Step 1 — Import and Create Zeros Array / 导入并创建零数组

We import `zeros` from NumPy and create a 3x5 array filled with zeros. This is commonly used to initialize matrices before filling with computed values.

我们从NumPy导入`zeros`并创建一个3x5的零数组。这通常用于在填充计算值之前初始化矩阵。

```python3
# 导入zeros函数 / Import zeros function
from numpy import zeros

# 创建3x5的全零数组 / Create 3x5 array filled with zeros
a = zeros([3, 5])
print("Array of zeros:")
print(a)
print("\nShape:", a.shape)
print("Data type:", a.dtype)
```

## Learning Notes / 学习笔记

• **Math Essence / 数学本质**: Zero matrices are fundamental in linear algebra. The zero vector and zero matrix are identity elements for addition. Creating zeros arrays is essential for initialization in iterative algorithms.

零矩阵是线性代数的基础。零向量和零矩阵是加法的单位元素。创建零数组对迭代算法中的初始化至关重要。

• **ML Application / ML应用**: In machine learning, `zeros()` is used to initialize weight matrices before training, create bias vectors, or initialize gradient accumulators in optimization algorithms.

在机器学习中，`zeros()`用于在训练前初始化权重矩阵、创建偏差向量或初始化优化算法中的梯度累加器。
