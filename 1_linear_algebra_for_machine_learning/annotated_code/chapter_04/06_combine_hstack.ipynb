### Summary / 总结
Learn how to combine multiple arrays horizontally (column-wise) using `hstack()`. This stacks arrays side by side.

学习如何使用`hstack()`水平（按列）组合多个数组。这将数组并排堆叠。

### Core Concept / 核心概念
**numpy.hstack((arr1, arr2, ...))**: Horizontally stacks sequences of arrays (along axis 1). Equivalent to concatenating columns.

**numpy.hstack((arr1, arr2, ...))**: 水平堆叠数组序列（沿轴1）。等同于连接列。

## Step 1 — Create First Array / 创建第一个数组

We create the first 1D array containing [1, 2, 3]. This will become the first column (or extend) of our horizontally stacked result.

我们创建第一个包含[1, 2, 3]的一维数组。这将成为我们水平堆叠结果的第一列（或扩展）。

```python3
# 导入array和hstack函数 / Import array and hstack functions
from numpy import array
from numpy import hstack

# 创建第一个数组 / Create first array
a1 = array([1, 2, 3])
print("First array:")
print(a1)
print("Shape:", a1.shape)
```

## Step 2 — Create Second Array / 创建第二个数组

We create the second 1D array containing [4, 5, 6]. This will be concatenated to the right of the first array.

我们创建第二个包含[4, 5, 6]的一维数组。这将连接到第一个数组的右侧。

```python3
# 创建第二个数组 / Create second array
a2 = array([4, 5, 6])
print("\nSecond array:")
print(a2)
print("Shape:", a2.shape)
```

## Step 3 — Stack Arrays Horizontally / 水平堆叠数组

We use `hstack()` to combine a1 and a2 into a longer 1D array. The elements are concatenated side by side.

我们使用`hstack()`将a1和a2组合成一个更长的一维数组。元素并排连接。

```python3
# 使用hstack水平堆叠 / Stack horizontally using hstack
a3 = hstack((a1, a2))
print("\nHorizontally stacked array:")
print(a3)
```

## Step 4 — Inspect Resulting Shape / 检查结果形状

The resulting array a3 is 1D with shape (6,): it concatenates all 3 elements from a1 followed by 3 elements from a2, creating a vector of length 6.

结果数组a3是一维的，形状为(6,)：它连接a1的所有3个元素，然后是a2的3个元素，创建一个长度为6的向量。

```python3
# 检查结果的形状 / Check shape of result
print("\nShape of stacked array:")
print(a3.shape)
```

## Learning Notes / 学习笔记

• **Math Essence / 数学本质**: hstack performs horizontal stacking along axis 1 (columns). For 1D arrays, it concatenates them end-to-end. For 2D arrays, it concatenates them column-wise. Mathematically: [a1, a2] = [a1 | a2] (side by side).

hstack沿轴1（列）执行水平堆叠。对于一维数组，它们端到端连接。对于二维数组，按列连接。数学上：[a1, a2] = [a1 | a2]（并排）。

• **ML Application / ML应用**: In ML, hstack is used to augment feature matrices by adding new feature columns. When combining features from different sources (e.g., raw features + engineered features), hstack extends the feature matrix horizontally.

在ML中，hstack用于通过添加新特征列来增强特征矩阵。当组合来自不同源的特征（例如，原始特征+工程特征）时，hstack水平扩展特征矩阵。

## Complete Code / 完整代码一览

```python3
# --- Horizontal Stacking / 水平堆叠 ---
from numpy import array
from numpy import hstack

# Create two 1D arrays / 创建两个一维数组
a1 = array([1, 2, 3])
print("First array:")
print(a1)
print("Shape:", a1.shape)

a2 = array([4, 5, 6])
print("\nSecond array:")
print(a2)
print("Shape:", a2.shape)

# Stack horizontally / 水平堆叠
a3 = hstack((a1, a2))
print("\nHorizontally stacked array:")
print(a3)
print("Shape:", a3.shape)
```
