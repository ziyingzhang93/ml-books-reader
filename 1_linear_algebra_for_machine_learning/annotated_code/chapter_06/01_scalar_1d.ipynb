### Summary / 总结
Learn how NumPy broadcasts scalar values with 1D arrays. Broadcasting is a powerful mechanism that allows operations between arrays of different shapes.

学习NumPy如何将标量值与一维数组广播。广播是一个强大的机制，允许不同形状的数组之间的操作。

### Core Concept / 核心概念
**Broadcasting with Scalar**: A scalar (0-dimensional) is automatically expanded to match array dimensions. Example: [1,2,3] + 2 = [3,4,5] (scalar 2 broadcasted to all elements).

**与标量的广播**: 标量（0维）自动扩展以匹配数组维度。示例：[1,2,3] + 2 = [3,4,5]（标量2广播到所有元素）。

## Step 1 — Create 1D Array and Scalar / 创建一维数组和标量

We create a 1D array [1, 2, 3] and a scalar value 2. NumPy will broadcast the scalar to each element.

我们创建一个一维数组[1, 2, 3]和标量值2。NumPy将标量广播到每个元素。

```python3
# 导入array函数 / Import array function
from numpy import array

# 创建一维数组和标量 / Create 1D array and scalar
a = array([1, 2, 3])
b = 2

print("Array a:")
print(a)
print("\nScalar b:")
print(b)
```

## Step 2 — Add Scalar to Array / 将标量添加到数组

We add the scalar 2 to the array. NumPy broadcasts 2 to all elements, resulting in [1+2, 2+2, 3+2] = [3, 4, 5].

我们将标量2添加到数组。NumPy将2广播到所有元素，导致[1+2, 2+2, 3+2] = [3, 4, 5]。

```python3
# 使用广播进行加法 / Add using broadcasting
c = a + b
print("\nResult of a + b (scalar broadcast):")
print(c)
```

## Learning Notes / 学习笔记

• **Math Essence / 数学本质**: Broadcasting applies a scalar operation element-wise to all array elements. Mathematically: if **a** = [a₁, a₂, ..., aₙ] and b is scalar, then **a** + b = [a₁+b, a₂+b, ..., aₙ+b].

广播对所有数组元素逐元素应用标量操作。数学上：如果**a** = [a₁, a₂, ..., aₙ]且b是标量，则**a** + b = [a₁+b, a₂+b, ..., aₙ+b]。

• **ML Application / ML应用**: Broadcasting is essential in ML: normalizing data by subtracting mean (X - mean), scaling features by dividing by standard deviation (X / std), or adding bias terms to all samples (predictions + bias).

广播在ML中至关重要：通过减去平均值来规范化数据(X - mean)、通过除以标准偏差来缩放特征(X / std)或向所有样本添加偏差项(predictions + bias)。
