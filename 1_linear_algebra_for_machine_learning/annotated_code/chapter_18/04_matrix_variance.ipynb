## Summary / 总结

Compute column-wise and row-wise variance for matrices. This is critical for understanding feature distributions in datasets.

计算矩阵的列方差和行方差。这对于理解数据集中的特征分布至关重要。

## Core Formula / 核心公式

**Column variance**: $s_j^2 = \frac{1}{m-1}\sum_{i=1}^{m}(x_{ij} - \bar{x}_j)^2$

**Row variance**: $s_i^2 = \frac{1}{n-1}\sum_{j=1}^{n}(x_{ij} - \bar{x}_i)^2$

## Step 1 — Import Libraries / 导入库

```python3
# Import array and variance function
# 导入数组和方差函数
from numpy import array
from numpy import var
```

## Step 2 — Create a Matrix / 创建矩阵

```python3
# Define a 2x6 matrix
# 定义一个2x6的矩阵
M = array([[1, 2, 3, 4, 5, 6],
           [1, 2, 3, 4, 5, 6]])
print("Matrix M:")
print(M)
```

## Step 3 — Calculate Column Variances / 计算列方差

```python3
# Compute variance along axis=0 (columns) with ddof=1 for sample variance
# 沿axis=0计算方差（列），ddof=1用于样本方差
col_var = var(M, ddof=1, axis=0)
print(f"Column variances: {col_var}")
```

## Step 4 — Calculate Row Variances / 计算行方差

```python3
# Compute variance along axis=1 (rows) with ddof=1 for sample variance
# 沿axis=1计算方差（行），ddof=1用于样本方差
row_var = var(M, ddof=1, axis=1)
print(f"Row variances: {row_var}")
```

## Learning Notes / 学习笔记

- **Math Essence**: Matrix variance computation extends vector variance by applying it along specific axes. Column variances show how each feature spreads; row variances show how each observation spreads.
  
  **数学本质**：矩阵方差计算通过沿特定轴应用向量方差来扩展。列方差显示每个特征的展度；行方差显示每个观测的展度。

- **ML Application**: Column variances are essential for StandardScaler preprocessing. Features with zero variance are constant and should be removed before training.
  
  **ML应用**：列方差对于StandardScaler预处理至关重要。方差为零的特征是常数，应在训练前删除。

## Complete Code / 完整代码一览

```python3
# --- Import Section / 导入部分 ---
from numpy import array
from numpy import var

# --- Matrix Variance Computation / 矩阵方差计算 ---
M = array([[1, 2, 3, 4, 5, 6],
           [1, 2, 3, 4, 5, 6]])
col_var = var(M, ddof=1, axis=0)
print(col_var)

# --- Row Variance Computation / 行方差计算 ---
row_var = var(M, ddof=1, axis=1)
print(row_var)
```
