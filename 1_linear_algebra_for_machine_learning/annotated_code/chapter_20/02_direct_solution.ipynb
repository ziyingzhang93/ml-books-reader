## Summary / 总结

Solve linear regression using the normal equation. This closed-form solution directly computes the optimal coefficients without iteration.

使用正规方程求解线性回归。这个闭形式解直接计算最优系数，无需迭代。

## Core Formula / 核心公式

**Normal Equation**: $$b = (X^T X)^{-1} X^T y$$

This minimizes $||Xb - y||^2$ in one step by setting the gradient to zero.

## Step 1 — Import Libraries / 导入库

```python3
# Import array operations, matrix inversion, and plotting
# 导入数组操作、矩阵求逆和绘图
from numpy import array
from numpy.linalg import inv
from matplotlib import pyplot
```

## Step 2 — Load Dataset / 加载数据集

```python3
# Load the regression dataset
# 加载回归数据集
data = array([[0.05, 0.12],
              [0.18, 0.22],
              [0.31, 0.35],
              [0.42, 0.38],
              [0.5, 0.49]])
X, y = data[:, 0], data[:, 1]
X = X.reshape((len(X), 1))
print(f"X shape: {X.shape}, y shape: {y.shape}")
```

## Step 3 — Compute X^T X / 计算 X^T X

```python3
# Compute the normal equation components
# X.T @ X: (features, n_samples) @ (n_samples, features) -> (features, features)
# 计算正规方程的组成部分
XTX = X.T.dot(X)
print(f"X^T X shape: {XTX.shape}")
print(f"X^T X =\n{XTX}")
```

## Step 4 — Compute (X^T X)^(-1) / 计算 (X^T X)^(-1)

```python3
# Invert the matrix (X^T X)^(-1)
# This assumes X^T X is invertible (has full rank)
# 对矩阵求逆
XTX_inv = inv(XTX)
print(f"(X^T X)^(-1) =\n{XTX_inv}")
```

## Step 5 — Compute X^T y / 计算 X^T y

```python3
# Compute the right-hand side
# X.T @ y: (features, n_samples) @ (n_samples,) -> (features,)
# 计算右侧
XTy = X.T.dot(y)
print(f"X^T y shape: {XTy.shape}")
print(f"X^T y = {XTy}")
```

## Step 6 — Compute Coefficients / 计算系数

```python3
# Solve for b using the normal equation
# b = (X^T X)^(-1) X^T y
# 使用正规方程求解b
b = XTX_inv.dot(XTy)
print(f"Coefficients b: {b}")
print(f"\nInterpretation: y ≈ {b[0]:.4f} * x")
```

## Step 7 — Make Predictions / 进行预测

```python3
# Compute predicted values: yhat = X @ b
# 计算预测值
yhat = X.dot(b)
print(f"Predictions: {yhat}")
print(f"Actual:      {y}")
print(f"\nResiduals (y - yhat): {y - yhat}")
```

## Step 8 — Visualize Results / 可视化结果

```python3
# Plot original data and fitted line
# 绘制原始数据和拟合线
pyplot.scatter(X, y, label='Data')
pyplot.plot(X, yhat, color='red', label='Fitted line')
pyplot.xlabel('X')
pyplot.ylabel('y')
pyplot.legend()
pyplot.title('Linear Regression - Normal Equation Solution')
pyplot.show()
```

## Learning Notes / 学习笔记

- **Math Essence**: The normal equation solves $\nabla_b(||Xb - y||^2) = 0$, yielding the closed-form solution. It's efficient for small to medium datasets but becomes numerically unstable if $X^T X$ is ill-conditioned (nearly singular).
  
  **数学本质**：正规方程解决 $\nabla_b(||Xb - y||^2) = 0$，产生闭形式解。它对于中小数据集很有效，但如果 $X^T X$ 条件数大（接近奇异）则数值上不稳定。

- **ML Application**: The normal equation is rarely used in practice due to numerical stability issues. Instead, use matrix decompositions (QR, SVD) or iterative methods. However, it's essential for understanding the mathematical foundation of linear regression.
  
  **ML应用**：由于数值稳定性问题，正规方程在实践中很少使用。相反，使用矩阵分解或迭代方法。但它对于理解线性回归的数学基础至关重要。

## Complete Code / 完整代码一览

```python3
# --- Import Section / 导入部分 ---
from numpy import array
from numpy.linalg import inv
from matplotlib import pyplot

# --- Load Data / 加载数据 ---
data = array([[0.05, 0.12],
              [0.18, 0.22],
              [0.31, 0.35],
              [0.42, 0.38],
              [0.5, 0.49]])
X, y = data[:, 0], data[:, 1]
X = X.reshape((len(X), 1))

# --- Solve Using Normal Equation / 使用正规方程求解 ---
b = inv(X.T.dot(X)).dot(X.T).dot(y)
print(b)

# --- Make Predictions and Plot / 进行预测和绘图 ---
yhat = X.dot(b)
pyplot.scatter(X, y)
pyplot.plot(X, yhat, color='red')
pyplot.show()
```
