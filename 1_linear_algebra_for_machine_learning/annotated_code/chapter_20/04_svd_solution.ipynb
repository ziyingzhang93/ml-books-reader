## Summary / 总结

Solve linear regression using the pseudo-inverse computed via Singular Value Decomposition. SVD is the most robust method and handles rank-deficient systems gracefully.

使用通过奇异值分解计算的伪逆求解线性回归。SVD是最鲁棒的方法，优雅地处理秩不足系统。

## Core Formula / 核心公式

1. **SVD Decomposition**: $X = U \Sigma V^T$
2. **Pseudo-inverse**: $X^+ = V \Sigma^+ U^T$ (inverse of diagonal singular values)
3. **Solution**: $b = X^+ y$

## Step 1 — Import Libraries / 导入库

```python
# Import pseudo-inverse function
# 导入伪逆函数
from numpy import array
from numpy.linalg import pinv
from matplotlib import pyplot
```

## Step 2 — Load Dataset / 加载数据集

```python
# Load the regression dataset
# 加载回归数据集
data = array([[0.05, 0.12],
              [0.18, 0.22],
              [0.31, 0.35],
              [0.42, 0.38],
              [0.5, 0.49]])
X, y = data[:, 0], data[:, 1]
X = X.reshape((len(X), 1))
print(f"X shape: {X.shape}, y shape: {y.shape}")
```

## Step 3 — Compute Pseudo-Inverse / 计算伪逆

```python
# Compute the pseudo-inverse X+ = V * Sigma+ * U^T
# pinv() performs SVD internally and returns the pseudo-inverse
# This is the most general solution that handles rank-deficient cases
# 计算伪逆 X+ = V * Sigma+ * U^T
# pinv()内部执行SVD并返回伪逆
X_pinv = pinv(X)
print(f"Pseudo-inverse shape: {X_pinv.shape}")
print(f"X+ =\n{X_pinv}")
```

## Step 4 — Verify Pseudo-Inverse Property / 验证伪逆性质

```python
# Property: X * X+ * X ≈ X (approximately for numerical precision)
# 性质：X * X+ * X ≈ X
X_reconstructed = X.dot(X_pinv).dot(X)
print("Original X:")
print(X)
print("\nX * X+ * X (should be close to X):")
print(X_reconstructed)
print(f"\nError: {max((X - X_reconstructed).flatten())}")
```

## Step 5 — Solve for Coefficients / 求解系数

```python
# Solve using the pseudo-inverse: b = X+ * y
# This is equivalent to finding the least-squares solution
# 使用伪逆求解
b = X_pinv.dot(y)
print(f"Coefficients b: {b}")
```

## Step 6 — Make Predictions / 进行预测

```python
# Compute predictions
# 计算预测值
yhat = X.dot(b)
print(f"Predictions: {yhat}")
print(f"Actual:      {y}")
print(f"\nMean Squared Error: {sum((y - yhat)**2) / len(y):.6f}")
```

## Step 7 — Visualize Results / 可视化结果

```python
# Plot original data and fitted line
# 绘制原始数据和拟合线
pyplot.scatter(X, y)
pyplot.plot(X, yhat, color='red')
pyplot.xlabel('X')
pyplot.ylabel('y')
pyplot.title('Linear Regression - SVD Solution')
pyplot.show()
```

## Learning Notes / 学习笔记

- **Math Essence**: The pseudo-inverse is defined as $X^+ = V \Sigma^+ U^T$ where $\Sigma^+$ inverts non-zero singular values and zeros out near-zero singular values. This handles rank-deficient and overdetermined systems naturally without failing on singularity.
  
  **数学本质**：伪逆定义为 $X^+ = V \Sigma^+ U^T$，其中 $\Sigma^+$ 反演非零奇异值。这自然地处理秩不足和超定系统。

- **ML Application**: SVD via pseudo-inverse is the gold standard for linear regression. It's numerically the most stable and handles all edge cases (rank deficiency, ill-conditioning). NumPy's `pinv()` is highly optimized. Use this when absolute robustness is required.
  
  **ML应用**：通过伪逆的SVD是线性回归的黄金标准。它在数值上最稳定，处理所有边界情况。NumPy的 `pinv()` 高度优化。当需要绝对鲁棒性时使用此方法。

## Complete Code / 完整代码一览

```python
# --- Import Section / 导入部分 ---
from numpy import array
from numpy.linalg import pinv
from matplotlib import pyplot

# --- Load Data / 加载数据 ---
data = array([[0.05, 0.12],
              [0.18, 0.22],
              [0.31, 0.35],
              [0.42, 0.38],
              [0.5, 0.49]])
X, y = data[:, 0], data[:, 1]
X = X.reshape((len(X), 1))

# --- SVD Solution via Pseudo-Inverse / 通过伪逆的SVD求解 ---
b = pinv(X).dot(y)
print(b)

# --- Make Predictions and Plot / 进行预测和绘图 ---
yhat = X.dot(b)
pyplot.scatter(X, y)
pyplot.plot(X, yhat, color='red')
pyplot.show()
```
