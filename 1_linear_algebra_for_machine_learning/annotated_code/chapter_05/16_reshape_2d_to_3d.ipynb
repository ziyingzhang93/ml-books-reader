### Summary / 总结
Learn how to reshape 2D arrays into 3D arrays. This is useful for working with image data, time-series sequences, or batches of data with multiple features.

学习如何将二维数组重新整形为三维数组。这对于处理图像数据、时间序列序列或具有多个特征的数据批次很有用。

### Core Concept / 核心概念
**3D Reshape**: Convert 2D array (m, n) to 3D array (m, n, 1). Useful for CNN image batches (batch_size, height, width) or sequence data (samples, timesteps, features).

**三维重新整形**: 将二维数组(m, n)转换为三维数组(m, n, 1)。用于CNN图像批次(batch_size, height, width)或序列数据(samples, timesteps, features)。

## Step 1 — Create 2D Array / 创建二维数组

We create a 2D array with shape (3, 2). We'll reshape it into a 3D array with shape (3, 2, 1).

我们创建一个形状为(3, 2)的二维数组。我们将其重新整形为形状为(3, 2, 1)的三维数组。

```python
# 导入array函数 / Import array function
from numpy import array

# 创建二维数组 / Create 2D array
data = [[11, 22], [33, 44], [55, 66]]
data = array(data)
print("Original 2D array:")
print(data)
print("Shape:", data.shape)
```

## Step 2 — Reshape to 3D / 重新整形为三维

We use .reshape() to add a third dimension. The new shape (3, 2, 1) means 3 "layers", each with 2 rows and 1 column.

我们使用.reshape()来添加第三维。新形状(3, 2, 1)表示3个"层"，每个有2行1列。

```python
# 重新整形为三维 / Reshape to 3D
data = data.reshape((data.shape[0], data.shape[1], 1))
print("\nReshaped 3D array:")
print(data)
print("Shape:", data.shape)
```

## Learning Notes / 学习笔记

• **Math Essence / 数学本质**: Reshaping to 3D is useful for tensor operations. A 2D array (m, n) becomes a 3D tensor (m, n, 1) by adding a dimension. This is fundamental in deep learning where data is 3D or 4D: (samples, height, width, channels) for images.

重新整形为三维对于张量操作很有用。二维数组(m, n)通过添加维度变为三维张量(m, n, 1)。这在深度学习中是基本的，其中数据是三维或四维的：图像的(samples, height, width, channels)。

• **ML Application / ML application**: In deep learning, reshaping is critical: grayscale images from (28, 28) need reshaping to (28, 28, 1) for CNN input. RNN requires shape (samples, timesteps, features). Understanding multi-dimensional arrays is essential for modern ML frameworks.

在深度学习中，重新整形至关重要：来自(28, 28)的灰度图像需要重新整形为(28, 28, 1)以进行CNN输入。RNN需要形状(samples, timesteps, features)。理解多维数组对现代ML框架至关重要。

## Complete Code / 完整代码一览

```python
# --- Reshape 2D to 3D / 从二维重新整形为三维 ---
from numpy import array

# Create 2D array / 创建二维数组
data = [[11, 22], [33, 44], [55, 66]]
data = array(data)
print("Original 2D array:")
print(data)
print("Shape:", data.shape)

# Reshape to 3D / 重新整形为三维
data = data.reshape((data.shape[0], data.shape[1], 1))
print("\nReshaped 3D array:")
print(data)
print("Shape:", data.shape)
```
