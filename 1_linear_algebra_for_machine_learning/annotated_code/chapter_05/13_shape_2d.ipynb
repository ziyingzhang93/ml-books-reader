### Summary / 总结
Learn how to inspect the shape of 2D arrays. For matrices, shape returns (rows, columns), critical for understanding dataset dimensions.

学习如何检查二维数组的形状。对于矩阵，shape返回(行，列)，对于理解数据集维度至关重要。

### Core Concept / 核心概念
**.shape for 2D arrays**: Returns (m, n) tuple where m is number of rows and n is number of columns. For example, shape (3, 2) means 3 rows and 2 columns.

**二维数组的.shape**: 返回(m, n)元组，其中m是行数，n是列数。例如，形状(3, 2)表示3行2列。

## Step 1 — Create 2D Array / 创建二维数组

We create a 2D array (matrix) with 3 rows and 2 columns from a nested list.

我们从嵌套列表创建一个具有3行2列的二维数组（矩阵）。

```python3
# 导入array函数 / Import array function
from numpy import array

# 创建二维数组 / Create 2D array
data = [[11, 22], [33, 44], [55, 66]]
data = array(data)
print("2D array (matrix):")
print(data)
```

## Step 2 — Inspect Shape / 检查形状

We use the .shape attribute on the 2D array. For this 3×2 matrix, shape is (3, 2), indicating 3 rows and 2 columns.

我们在二维数组上使用.shape属性。对于这个3×2矩阵，形状是(3, 2)，表示3行2列。

```python3
# 检查形状 / Check shape
print("\nShape:")
print(data.shape)
print("\nType of shape:")
print(type(data.shape))
```

## Learning Notes / 学习笔记

• **Math Essence / 数学本质**: For a matrix A with m rows and n columns, shape = (m, n). In linear algebra notation: A ∈ ℝ^(m×n). The shape is fundamental for matrix operations: only compatible shapes can be multiplied, added, etc.

对于具有m行n列的矩阵A，shape = (m, n)。在线性代数符号中：A ∈ ℝ^(m×n)。形状对矩阵操作至关重要：只有兼容的形状才能相乘、相加等。

• **ML Application / ML application**: In ML, if your dataset has shape (100, 20), you have 100 training samples with 20 features. Most ML algorithms require shape verification: if a model expects shape (m, n) but receives (n, m), operations fail silently or give wrong results.

在ML中，如果你的数据集形状是(100, 20)，你有100个训练样本，每个20个特征。大多数ML算法需要形状验证：如果模型期望形状(m, n)但接收(n, m)，操作可能静默失败或给出错误结果。
