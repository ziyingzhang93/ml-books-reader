### Summary / 总结
Learn how to extract entire rows from 2D arrays using row indexing. This is essential for accessing complete samples in datasets.

学习如何使用行索引从二维数组提取整行。这对于访问数据集中的完整样本至关重要。

### Core Concept / 核心概念
**Row Slicing**: Access an entire row using [row_index, :] notation. The colon ":" means "all columns". Returns a 1D array.

**行切片**: 使用[row_index, :]符号访问整行。冒号":"表示"所有列"。返回一维数组。

## Step 1 — Create 2D Array / 创建二维数组

We create a 3x2 matrix. Now we'll extract the first row using the syntax [0, :].

我们创建一个3x2矩阵。现在我们将使用语法[0, :]提取第一行。

```python
# 导入array函数 / Import array function
from numpy import array

# 创建二维数组 / Create 2D array
data = array([[11, 22], [33, 44], [55, 66]])
print("Matrix:")
print(data)
```

## Step 2 — Extract First Row / 提取第一行

We use [0, :] to extract the entire first row. The colon ":" selects all columns, returning a 1D array [11, 22].

我们使用[0, :]提取整个第一行。冒号":"选择所有列，返回一维数组[11, 22]。

```python
# 提取第一行（所有列） / Extract first row (all columns)
print("\nFirst row (row 0, all columns):")
print(data[0, :])
print("Shape:", data[0, :].shape)
```

## Learning Notes / 学习笔记

• **Math Essence / 数学本质**: Extracting a row is equivalent to selecting the i-th row of a matrix, returning a vector. In linear algebra, the i-th row of matrix A is denoted as A_i. In NumPy, A[i, :] gives this row.

提取一行等同于选择矩阵的第i行，返回一个向量。在线性代数中，矩阵A的第i行表示为A_i。在NumPy中，A[i, :]给出这一行。

• **ML Application / ML应用**: In ML, extracting a row is critical for accessing individual samples. If your dataset matrix has samples as rows and features as columns, data[i, :] retrieves the i-th training sample as a feature vector.

在ML中，提取一行对于访问单个样本至关重要。如果数据集矩阵将样本作为行，特征作为列，data[i, :]将第i个训练样本作为特征向量检索。

## Complete Code / 完整代码一览

```python
# --- Extract Row from 2D Array / 从二维数组提取行 ---
from numpy import array

# Create 2D array / 创建二维数组
data = array([[11, 22], [33, 44], [55, 66]])
print("Matrix:")
print(data)

# Extract first row / 提取第一行
print("\nFirst row (row 0, all columns):")
print(data[0, :])
print("Shape:", data[0, :].shape)
```
