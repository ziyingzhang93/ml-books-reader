### Summary / 总结
Learn how to separate input features (X) from output labels (y) in a dataset matrix. This is a fundamental operation in preparing data for machine learning.

学习如何在数据集矩阵中分离输入特征(X)和输出标签(y)。这是为机器学习准备数据的基本操作。

### Core Concept / 核心概念
**Feature-Label Separation**: X = data[:, :-1] extracts all rows and all columns except the last (features). y = data[:, -1] extracts the last column (labels).

**特征-标签分离**: X = data[:, :-1]提取所有行和除最后一列外的所有列（特征）。y = data[:, -1]提取最后一列（标签）。

## Step 1 — Create Dataset Matrix / 创建数据集矩阵

We create a 3x3 matrix where the first 2 columns are features and the last column is the target label. This is a typical ML dataset structure.

我们创建一个3x3矩阵，其中前2列是特征，最后一列是目标标签。这是典型的ML数据集结构。

```python
# 导入array函数 / Import array function
from numpy import array

# 创建数据集矩阵 / Create dataset matrix
data = array([[11, 22, 33], [44, 55, 66], [77, 88, 99]])
print("Full dataset (features and labels):")
print(data)
```

## Step 2 — Separate Features (X) / 分离特征(X)

We extract all rows and all columns except the last one. data[:, :-1] returns a 3x2 matrix containing only the input features.

我们提取所有行和除最后一列外的所有列。data[:, :-1]返回一个3x2矩阵，仅包含输入特征。

```python
# 分离输入特征 / Separate input features
X = data[:, :-1]
print("\nInput features (X) - all rows, all columns except last:")
print(X)
print("Shape:", X.shape)
```

## Step 3 — Separate Labels (y) / 分离标签(y)

We extract only the last column, which contains the target labels. data[:, -1] returns a 1D array with the output values.

我们仅提取最后一列，其中包含目标标签。data[:, -1]返回一个包含输出值的一维数组。

```python
# 分离输出标签 / Separate output labels
y = data[:, -1]
print("\nOutput labels (y) - all rows, last column only:")
print(y)
print("Shape:", y.shape)
```

## Learning Notes / 学习笔记

• **Math Essence / 数学本质**: In supervised learning, datasets are partitioned into input matrix **X** (m × n) and output vector **y** (m × 1), where m is samples and n is features. The operation data[:, :-1] and data[:, -1] performs this fundamental decomposition.

在有监督学习中，数据集被划分为输入矩阵**X**(m × n)和输出向量**y**(m × 1)，其中m是样本，n是特征。操作data[:, :-1]和data[:, -1]执行这个基本分解。

• **ML Application / ML应应用**: Separating X and y is the first step in any supervised learning pipeline. X is fed to the model during training, and y provides the ground truth labels. This separation enables model training: model.fit(X, y).

分离X和y是任何有监督学习管道的第一步。X在训练期间被馈送给模型，y提供真实标签。这个分离使模型训练成为可能：model.fit(X, y)。

## Complete Code / 完整代码一览

```python
# --- Separate Features and Labels / 分离特征和标签 ---
from numpy import array

# Create dataset matrix / 创建数据集矩阵
data = array([[11, 22, 33], [44, 55, 66], [77, 88, 99]])
print("Full dataset:")
print(data)

# Separate features and labels / 分离特征和标签
X = data[:, :-1]  # All rows, all columns except last / 所有行，除最后一列外的所有列
y = data[:, -1]   # All rows, last column only / 所有行，仅最后一列

print("\nInput features (X):")
print(X)
print("Shape:", X.shape)

print("\nOutput labels (y):")
print(y)
print("Shape:", y.shape)
```
