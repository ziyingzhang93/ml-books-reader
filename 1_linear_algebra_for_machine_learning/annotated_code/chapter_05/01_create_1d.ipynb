### Summary / 总结
Learn how to create 1D arrays (vectors) from Python lists and inspect their type. Understanding array creation is foundational for all subsequent operations.

学习如何从Python列表创建一维数组（向量）并检查其类型。理解数组创建是所有后续操作的基础。

### Core Concept / 核心概念
**1D Array (Vector)**: A one-dimensional array of scalars, represented as a sequence. In linear algebra notation: **v** = [v₁, v₂, ..., vₙ]

**一维数组（向量）**: 标量的一维数组，表示为一个序列。在线性代数符号中：**v** = [v₁, v₂, ..., vₙ]

## Step 1 — Create Python List / 创建Python列表

We start with a Python list containing five integer values. This list will be converted to a NumPy array.

我们以包含五个整数值的Python列表开始。这个列表将被转换为NumPy数组。

```python
# 导入array函数 / Import array function
from numpy import array

# 创建数据列表 / Create data list
data = [11, 22, 33, 44, 55]
print("Python list:")
print(data)
```

## Step 2 — Convert to NumPy Array / 转换为NumPy数组

We convert the Python list to a NumPy array, which enables efficient numerical operations and linear algebra computations.

我们将Python列表转换为NumPy数组，这使得高效的数值运算和线性代数计算成为可能。

```python
# 转换为NumPy数组 / Convert to NumPy array
data = array(data)
print("\nNumPy 1D array:")
print(data)
```

## Step 3 — Inspect Type and Properties / 检查类型和属性

We verify that data is now a NumPy array (not a Python list). We also check its shape and data type.

我们验证数据现在是一个NumPy数组（不是Python列表）。我们还检查其形状和数据类型。

```python
# 检查类型 / Check type
print("\nType:")
print(type(data))
print("\nShape:", data.shape)
print("Data type:", data.dtype)
```

## Learning Notes / 学习笔记

• **Math Essence / 数学本质**: A vector is a fundamental object in linear algebra. The shape (5,) indicates a 1D array with 5 elements. Vectors are used for representing points in space, directions, or features in ML.

向量是线性代数中的基本对象。形状(5,)表示一个有5个元素的一维数组。向量用于表示空间中的点、方向或ML中的特征。

• **ML Application / ML应用**: In ML, each feature vector represents one sample's attributes. For example, [11, 22, 33, 44, 55] could be 5 features of a single data point used in a machine learning model.

在ML中，每个特征向量表示一个样本的属性。例如，[11, 22, 33, 44, 55]可以是单个数据点的5个特征，用于机器学习模型。

## Complete Code / 完整代码一览

```python
# --- Create 1D Array / 创建一维数组 ---
from numpy import array

# Create Python list / 创建Python列表
data = [11, 22, 33, 44, 55]
print("Python list:")
print(data)

# Convert to NumPy array / 转换为NumPy数组
data = array(data)
print("\nNumPy 1D array:")
print(data)

# Check type and properties / 检查类型和属性
print("\nType:", type(data))
print("Shape:", data.shape)
print("Data type:", data.dtype)
```
