### Summary / 总结
Learn how to slice specific subsets of array elements using start and stop indices. This is crucial for extracting portions of arrays.

学习如何使用start和stop索引切片数组元素的特定子集。这对于提取数组的部分至关重要。

### Core Concept / 核心概念
**Range Slicing**: [start:stop] extracts elements from index start (inclusive) to index stop (exclusive). Example: [0:1] returns only the first element.

**范围切片**: [start:stop]提取从索引start（包含）到索引stop（排他）的元素。示例：[0:1]仅返回第一个元素。

## Step 1 — Create 1D Array / 创建一维数组

We create a 1D array with 5 elements. Now we'll extract a subset using [0:1], which gets only the first element.

我们创建一个有5个元素的一维数组。现在我们将使用[0:1]提取一个子集，它只获取第一个元素。

```python3
# 导入array函数 / Import array function
from numpy import array

# 创建一维数组 / Create 1D array
data = array([11, 22, 33, 44, 55])
print("Original array:")
print(data)
```

## Step 2 — Slice First Element Only / 仅切片第一个元素

We use [0:1] to extract elements from index 0 to 1 (exclusive), which returns only element at index 0. Note the difference: data[0] returns a scalar, but data[0:1] returns a 1D array.

我们使用[0:1]提取从索引0到1（排他）的元素，只返回索引0处的元素。注意差异：data[0]返回标量，但data[0:1]返回一维数组。

```python3
# 使用[0:1]切片第一个元素 / Slice first element using [0:1]
print("\nFirst element only (using [0:1]):")
print(data[0:1])
print("Type:", type(data[0:1]))
print("Shape:", data[0:1].shape)
```

## Learning Notes / 学习笔记

• **Math Essence / 数学本质**: Slicing [start:stop] is a fundamental operation in linear algebra for selecting subvectors. It returns elements from start (inclusive) to stop (exclusive). Important: data[i] and data[i:i+1] are different—one is a scalar, the other is a 1D array.

切片[start:stop]是线性代数中选择子向量的基本操作。它返回从start（包含）到stop（排他）的元素。重要：data[i]和data[i:i+1]不同——一个是标量，另一个是一维数组。

• **ML Application / ML application**: Slicing is used to select training samples (data[0:n_train]), extract feature windows, separate input from labels (data[:, :-1] vs data[:, -1]), or implement mini-batch sampling.

切片用于选择训练样本(data[0:n_train])、提取特征窗口、分离输入与标签(data[:, :-1]对比data[:, -1])或实现小批量采样。
