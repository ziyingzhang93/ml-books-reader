## Step 1 — create and plot an exponential time series

```python
from matplotlib import pyplot
from math import exp
series = [exp(i) for i in range(1,100)]
pyplot.figure(1)
```

---
## Step 2 — line plot

```python
pyplot.subplot(211)
pyplot.plot(series)
```

---
## Step 3 — histogram

```python
pyplot.subplot(212)
pyplot.hist(series)
pyplot.show()
```
